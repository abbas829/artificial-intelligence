## What is a Class? {#what-is-a-class}

A **class** is a blueprint for creating objects. Think of it like a cookie cutter: the class defines the shape, and each object (instance) is an actual cookie made from that cutter. In Python, we define classes using the `class` keyword.

💡 **Key Concept**: Classes encapsulate data (attributes) and functions (methods) that operate on that data into a single, cohesive unit.


```python
# Creating your first class
class DataScientist:
    """A simple class representing a data scientist."""
    
    def __init__(self, name, years_experience):
        # Constructor method - runs when creating a new instance
        self.name = name
        self.years_experience = years_experience
        self.skills = []
    
    def add_skill(self, skill):
        """Add a new skill to the data scientist's toolkit."""
        self.skills.append(skill)
        print(f"{self.name} learned {skill}!")
    
    def introduce(self):
        """Introduce the data scientist."""
        return f"Hi, I'm {self.name} with {self.years_experience} years of experience."

# Creating instances (objects) from the class
ds1 = DataScientist("Alice", 5)
ds2 = DataScientist("Bob", 3)

# Using methods
print(ds1.introduce())
ds1.add_skill("Python")
ds1.add_skill("PyTorch")
print(f"Alice's skills: {ds1.skills}")
print(f"Bob's skills: {ds2.skills}")  # Empty list - separate from Alice's
```

## The `__init__` Constructor and `self` {#the-init-constructor}

The `__init__` method is Python's **constructor**. It automatically runs when you create a new instance. The `self` parameter refers to the instance itself—it's how the object keeps track of its own data.

📌 **Important**: `self` is passed automatically; you don't provide it when calling methods, but you must include it in method definitions.


```python
class Experiment:
    """Class to track ML experiments with initialization."""
    
    def __init__(self, experiment_name, model_type, learning_rate=0.001):
        # self allows us to store parameters as instance attributes
        self.experiment_name = experiment_name
        self.model_type = model_type
        self.learning_rate = learning_rate
        self.epochs_trained = 0
        self.history = {'loss': [], 'accuracy': []}
        
        print(f"Experiment '{experiment_name}' initialized with {model_type}")
    
    def log_epoch(self, loss, accuracy):
        """Log metrics for one epoch."""
        self.history['loss'].append(loss)
        self.history['accuracy'].append(accuracy)
        self.epochs_trained += 1

# Creating experiment instances
exp1 = Experiment("CNN_ImageNet", "ResNet50", 0.01)
exp2 = Experiment("BERT_Sentiment", "Transformer")  # Uses default learning_rate

# Logging some data
exp1.log_epoch(0.5, 0.75)
exp1.log_epoch(0.3, 0.85)
print(f"\nExperiment 1 history: {exp1.history}")
print(f"Epochs trained: {exp1.epochs_trained}")
```

## Attributes: Instance vs Class {#attributes-instance-vs-class}

**Instance attributes** (defined in `__init__` with `self`) are unique to each object. **Class attributes** are shared across all instances of the class—useful for constants or shared state.


```python
class NeuralNetwork:
    """Demonstrating instance vs class attributes."""
    
    # Class attribute - shared by ALL instances
    framework = "PyTorch"
    version = "2.0"
    
    def __init__(self, name, layers):
        # Instance attributes - unique to each instance
        self.name = name
        self.layers = layers
        self.parameters = sum(layers)  # Simple parameter count

# All instances share class attributes
nn1 = NeuralNetwork("SimpleNet", [64, 32, 10])
nn2 = NeuralNetwork("DeepNet", [128, 64, 32, 16, 10])

print(f"nn1 framework: {nn1.framework}")  # Access via instance
print(f"nn2 framework: {nn2.framework}")
print(f"Class access: {NeuralNetwork.framework}")  # Access via class

# But instance attributes are independent
print(f"\nn1 name: {nn1.name}, parameters: {nn1.parameters}")
print(f"nn2 name: {nn2.name}, parameters: {nn2.parameters}")

# Modifying class attribute affects all instances
NeuralNetwork.version = "2.1"
print(f"\nUpdated version for nn1: {nn1.version}")
print(f"Updated version for nn2: {nn2.version}")
```

## Methods: Regular, Class, and Static {#methods-types}

Python offers three types of methods:

1. **Regular methods**: Take `self` as first parameter, operate on instance data
2. **Class methods** (`@classmethod`): Take `cls` as first parameter, operate on class itself
3. **Static methods** (`@staticmethod`): No automatic first parameter, utility functions


```python
class MLUtils:
    """Utility class demonstrating different method types."""
    
    # Class attribute
    supported_formats = ['csv', 'json', 'parquet']
    
    def __init__(self, tool_name):
        self.tool_name = tool_name
    
    # Regular method - needs instance (self)
    def describe(self):
        return f"Tool: {self.tool_name}, supports: {self.supported_formats}"
    
    # Class method - operates on class, not instance
    @classmethod
    def add_format(cls, new_format):
        """Add a new supported format to the class."""
        cls.supported_formats.append(new_format)
        return f"Added {new_format}. Now supports: {cls.supported_formats}"
    
    # Static method - utility function, doesn't need class or instance
    @staticmethod
    def calculate_accuracy(predictions, labels):
        """Calculate accuracy given predictions and labels."""
        if len(predictions) != len(labels):
            raise ValueError("Predictions and labels must have same length")
        correct = sum(1 for p, l in zip(predictions, labels) if p == l)
        return correct / len(predictions)

# Using regular method
utils = MLUtils("DataValidator")
print(utils.describe())

# Using class method (can call on class or instance)
print(MLUtils.add_format("hdf5"))
print(f"Formats after adding: {MLUtils.supported_formats}")

# Using static method (no instance needed!)
preds = [1, 0, 1, 1, 0, 1]
labels = [1, 0, 1, 0, 0, 1]
acc = MLUtils.calculate_accuracy(preds, labels)
print(f"\nAccuracy: {acc:.2%}")
```

## Inheritance and `super()` {#inheritance}

**Inheritance** allows a class to inherit attributes and methods from another class. The child class (subclass) can override methods or add new ones. Use `super()` to call the parent's methods.

💡 This is crucial in ML: `class MyModel(nn.Module)` inherits from PyTorch's base Module class!


```python
class BaseModel:
    """Base class for all ML models."""
    
    def __init__(self, name, input_dim, output_dim):
        self.name = name
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.is_trained = False
    
    def train(self, epochs=10):
        """Base training method."""
        print(f"Training {self.name} for {epochs} epochs...")
        self.is_trained = True
    
    def predict(self, x):
        raise NotImplementedError("Subclasses must implement predict()")

class LinearRegressionModel(BaseModel):
    """Linear Regression inheriting from BaseModel."""
    
    def __init__(self, input_dim, output_dim, learning_rate=0.01):
        # Call parent's __init__ using super()
        super().__init__("LinearRegression", input_dim, output_dim)
        self.learning_rate = learning_rate
        self.weights = None  # Would be initialized properly in real code
    
    def train(self, epochs=10):
        # Override but extend parent's method
        print(f"Initializing weights for Linear Regression...")
        super().train(epochs)  # Call parent's train method
        print(f"Optimizing with learning rate {self.learning_rate}")
    
    def predict(self, x):
        """Implement abstract method."""
        if not self.is_trained:
            raise ValueError("Model must be trained before prediction!")
        return f"Prediction from {self.name} for input shape {len(x)}"

# Using inheritance
model = LinearRegressionModel(input_dim=10, output_dim=1, learning_rate=0.001)
model.train(epochs=5)
print(model.predict([1.0] * 10))

# Check inheritance
print(f"\nIs LinearRegressionModel a subclass of BaseModel? {issubclass(LinearRegressionModel, BaseModel)}")
print(f"Is model instance of BaseModel? {isinstance(model, BaseModel)}")
```

## Encapsulation: Public, Protected, Private {#encapsulation}

**Encapsulation** hides internal state and requires all interaction to be performed through an object's methods. Python uses naming conventions:

- **Public**: `name` - accessible anywhere
- **Protected**: `_name` - internal use, subclass access OK (convention)
- **Private**: `__name` - name mangling, harder to access from outside


```python
class SecureDataset:
    """Demonstrating encapsulation with protected and private attributes."""
    
    def __init__(self, data_path, sensitive_info):
        self.data_path = data_path  # Public
        self._raw_data = None       # Protected (convention: internal use)
        self.__password = sensitive_info  # Private (name mangled)
        self.__access_count = 0     # Private counter
    
    def load_data(self):
        """Public method to load data safely."""
        self._raw_data = f"Loaded data from {self.data_path}"
        self.__access_count += 1
        return self._raw_data
    
    def _preprocess(self):          # Protected method
        """Internal preprocessing - subclasses can override."""
        return self._raw_data.lower() if self._raw_data else None
    
    def __validate(self):           # Private method
        """Strict internal validation."""
        return self.__password is not None
    
    def get_access_count(self):
        """Public getter for private attribute."""
        return self.__access_count

# Testing encapsulation
dataset = SecureDataset("/data/train.csv", "secret123")

# Public access - fine
print(f"Public path: {dataset.data_path}")

# Protected access - possible but discouraged by convention
dataset._raw_data = "Manually set"  # Works, but don't do this!
print(f"Protected data: {dataset._raw_data}")

# Private access - harder (name mangling to _ClassName__attribute)
try:
    print(dataset.__password)  # AttributeError
except AttributeError as e:
    print(f"Error accessing private: {e}")

# But technically accessible via mangled name (don't do this in production!)
print(f"Mangled access: {dataset._SecureDataset__password}")

# Use public methods instead
dataset.load_data()
dataset.load_data()
print(f"Access count (via getter): {dataset.get_access_count()}")
```

## Magic/Dunder Methods {#magic-methods}

**Dunder** (double underscore) methods let you define how objects behave with built-in operations. They're essential for making classes Pythonic!

Common dunder methods:
- `__str__`: Human-readable string representation (print)
- `__repr__`: Unambiguous representation (debugging)
- `__len__`: Length of object (len())
- `__getitem__`: Access by index (obj[key])
- `__call__`: Make object callable (obj())


```python
class CustomDataset:
    """ML Dataset class with magic methods for Pythonic interface."""
    
    def __init__(self, data, labels, transform=None):
        self.data = data
        self.labels = labels
        self.transform = transform
    
    def __len__(self):
        """Allow len(dataset) - returns number of samples."""
        return len(self.data)
    
    def __getitem__(self, idx):
        """Allow dataset[i] - returns sample and label."""
        sample = self.data[idx]
        label = self.labels[idx]
        
        if self.transform:
            sample = self.transform(sample)
        
        return sample, label
    
    def __str__(self):
        """Pretty print for users."""
        return f"CustomDataset(n_samples={len(self)}, n_features={len(self.data[0]) if self.data else 0})"
    
    def __repr__(self):
        """Detailed representation for debugging."""
        return f"CustomDataset(data={self.data!r}, labels={self.labels!r}, transform={self.transform!r})"
    
    def __call__(self, idx):
        """Make dataset callable: dataset(5) returns 5th sample."""
        return self[idx]

# Create dataset
data = [[1, 2], [3, 4], [5, 6], [7, 8]]
labels = [0, 1, 0, 1]
dataset = CustomDataset(data, labels)

# Using magic methods
print(f"String representation: {str(dataset)}")
print(f"Repr: {repr(dataset)}")
print(f"Length: {len(dataset)}")
print(f"First sample (indexing): {dataset[0]}")
print(f"Last sample (calling): {dataset(3)}")

# Iteration works automatically with __len__ and __getitem__
print("\nIterating through dataset:")
for i, (x, y) in enumerate(dataset):
    print(f"  Sample {i}: X={x}, y={y}")
```

### Example 1: Metric Class with Update/Compute Pattern

This pattern is used in PyTorch Lightning, Hugging Face, and professional ML code.


```python
import numpy as np

class AccuracyMetric:
    """Accuracy metric following update/compute pattern."""
    
    def __init__(self, name="Accuracy"):
        self.name = name
        self.reset()
    
    def reset(self):
        """Reset accumulated statistics."""
        self.correct = 0
        self.total = 0
        self.history = []
    
    def update(self, predictions, targets):
        """Update metric with batch of predictions."""
        preds = np.argmax(predictions, axis=1) if predictions.ndim > 1 else predictions
        correct = np.sum(preds == targets)
        self.correct += correct
        self.total += len(targets)
        
        # Store batch accuracy
        batch_acc = correct / len(targets)
        self.history.append(batch_acc)
    
    def compute(self):
        """Compute final aggregated metric."""
        if self.total == 0:
            return 0.0
        return self.correct / self.total
    
    def __str__(self):
        return f"{self.name}: {self.compute():.4f}"

# Simulating training
metric = AccuracyMetric()

# Simulate 5 batches of predictions
np.random.seed(42)
for batch in range(5):
    # Random predictions and targets
    preds = np.random.rand(32, 10)  # 32 samples, 10 classes
    targets = np.random.randint(0, 10, 32)
    
    metric.update(preds, targets)
    print(f"Batch {batch+1}: Current accuracy = {metric.compute():.4f}")

print(f"\nFinal {metric}")
print(f"History length: {len(metric.history)} batches")
```

### Example 2: Config Class for Hyperparameters

Clean configuration management is essential for reproducible ML experiments.


```python
class Config:
    """Configuration class with dict-like and attribute access."""
    
    def __init__(self, **kwargs):
        # Default configuration
        self.learning_rate = 0.001
        self.batch_size = 32
        self.epochs = 100
        self.optimizer = "Adam"
        self.hidden_dims = [256, 128, 64]
        self.dropout = 0.2
        self.seed = 42
        
        # Override with provided kwargs
        for key, value in kwargs.items():
            if hasattr(self, key):
                setattr(self, key, value)
            else:
                raise ValueError(f"Unknown config parameter: {key}")
    
    def __getitem__(self, key):
        """Allow config['key'] access."""
        return getattr(self, key)
    
    def __setitem__(self, key, value):
        """Allow config['key'] = value assignment."""
        setattr(self, key, value)
    
    def __contains__(self, key):
        """Allow 'key' in config check."""
        return hasattr(self, key)
    
    def to_dict(self):
        """Convert to dictionary for saving/loading."""
        return {k: v for k, v in self.__dict__.items() if not k.startswith('_')}
    
    def __str__(self):
        items = [f"  {k}: {v}" for k, v in self.to_dict().items()]
        return "Config:\n" + "\n".join(items)

# Usage
config = Config(learning_rate=0.01, batch_size=64, epochs=50)
print(config)

# Both access styles work
print(f"\nAttribute access: {config.learning_rate}")
print(f"Dict access: {config['batch_size']}")

# Check existence
print(f"\n'seed' in config? {'seed' in config}")
print(f"'momentum' in config? {'momentum' in config}")

# Convert to dict for JSON serialization
import json
config_dict = config.to_dict()
print(f"\nJSON-ready dict: {json.dumps(config_dict, indent=2)}")
```

### Example 3: History Class with Visualization

Track and visualize training metrics over time.


```python
import matplotlib.pyplot as plt

class TrainingHistory:
    """Track and visualize training history."""
    
    def __init__(self):
        self.metrics = {}
    
    def log(self, epoch, **kwargs):
        """Log metrics for an epoch."""
        if 'epoch' not in self.metrics:
            self.metrics['epoch'] = []
        self.metrics['epoch'].append(epoch)
        
        for key, value in kwargs.items():
            if key not in self.metrics:
                self.metrics[key] = []
            self.metrics[key].append(value)
    
    def plot(self, metrics=None, figsize=(10, 6), save_path=None):
        """Plot training history."""
        if not self.metrics or 'epoch' not in self.metrics:
            print("No data to plot!")
            return
        
        # Determine which metrics to plot
        available = [k for k in self.metrics.keys() if k != 'epoch']
        to_plot = metrics if metrics else available
        
        # Create subplots
        n_plots = len(to_plot)
        fig, axes = plt.subplots(1, n_plots, figsize=figsize)
        if n_plots == 1:
            axes = [axes]
        
        epochs = self.metrics['epoch']
        
        for ax, metric_name in zip(axes, to_plot):
            if metric_name in self.metrics:
                ax.plot(epochs, self.metrics[metric_name], 'b-', linewidth=2, label=metric_name)
                ax.set_xlabel('Epoch')
                ax.set_ylabel(metric_name)
                ax.set_title(f'{metric_name} over time')
                ax.grid(True, alpha=0.3)
                ax.legend()
            else:
                ax.text(0.5, 0.5, f"No data for {metric_name}", 
                       ha='center', va='center', transform=ax.transAxes)
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path, dpi=150, bbox_inches='tight')
            print(f"Saved plot to {save_path}")
        
        plt.show()
    
    def get_best(self, metric, mode='min'):
        """Get best value and epoch for a metric."""
        if metric not in self.metrics:
            return None
        
        values = self.metrics[metric]
        if mode == 'min':
            best_val = min(values)
        else:
            best_val = max(values)
        
        best_epoch = self.metrics['epoch'][values.index(best_val)]
        return {'epoch': best_epoch, 'value': best_val}

# Simulate training and plot
history = TrainingHistory()

# Simulate 20 epochs of training
import numpy as np
np.random.seed(42)

for epoch in range(1, 21):
    # Simulated metrics: loss decreases, accuracy increases
    loss = 1.0 * np.exp(-epoch/5) + 0.1 + np.random.normal(0, 0.02)
    acc = 0.5 + 0.45 * (1 - np.exp(-epoch/5)) + np.random.normal(0, 0.01)
    val_loss = loss + 0.05 + np.random.normal(0, 0.01)
    
    history.log(epoch, loss=loss, accuracy=acc, val_loss=val_loss)

# Plot results
history.plot(metrics=['loss', 'accuracy', 'val_loss'], figsize=(15, 4))

# Get best results
best_loss = history.get_best('loss', mode='min')
best_acc = history.get_best('accuracy', mode='max')
print(f"\nBest loss: {best_loss['value']:.4f} at epoch {best_loss['epoch']}")
print(f"Best accuracy: {best_acc['value']:.4f} at epoch {best_acc['epoch']}")
```

### Example 4: Simple DataLoader Class

A simplified version of PyTorch's DataLoader to understand the pattern.


```python
class SimpleDataLoader:
    """Simple batching data loader for ML training."""
    
    def __init__(self, dataset, batch_size=32, shuffle=True, drop_last=False):
        self.dataset = dataset
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.drop_last = drop_last
        self.indices = list(range(len(dataset)))
    
    def __iter__(self):
        """Make the dataloader iterable: for batch in loader."""
        # Shuffle indices if needed
        if self.shuffle:
            np.random.shuffle(self.indices)
        
        self.current_idx = 0
        return self
    
    def __next__(self):
        """Get next batch."""
        if self.current_idx >= len(self.indices):
            raise StopIteration
        
        # Get batch indices
        batch_indices = self.indices[self.current_idx:self.current_idx + self.batch_size]
        
        # Check if we should drop the last incomplete batch
        if len(batch_indices) < self.batch_size and self.drop_last:
            raise StopIteration
        
        # Retrieve actual data
        batch_data = [self.dataset[i] for i in batch_indices]
        
        # Separate samples and labels (assuming dataset returns (x, y) tuples)
        samples = [item[0] for item in batch_data]
        labels = [item[1] for item in batch_data]
        
        self.current_idx += self.batch_size
        
        return np.array(samples), np.array(labels)
    
    def __len__(self):
        """Number of batches per epoch."""
        n_samples = len(self.dataset)
        if self.drop_last:
            return n_samples // self.batch_size
        return (n_samples + self.batch_size - 1) // self.batch_size

# Create a simple dataset
class DummyDataset:
    def __init__(self, n_samples=100, n_features=10):
        self.X = np.random.randn(n_samples, n_features)
        self.y = np.random.randint(0, 2, n_samples)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Test the DataLoader
dataset = DummyDataset(n_samples=50, n_features=5)
loader = SimpleDataLoader(dataset, batch_size=16, shuffle=True, drop_last=False)

print(f"Dataset size: {len(dataset)}")
print(f"Number of batches: {len(loader)}")
print("\nIterating through batches:")

for batch_idx, (X_batch, y_batch) in enumerate(loader):
    print(f"Batch {batch_idx+1}: X shape {X_batch.shape}, y shape {y_batch.shape}")
```

### Solution 2: Class and Static Methods


```python
class Student:
    university = "AI University"
    
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.courses = []
    
    def enroll(self, course):
        self.courses.append(course)
    
    def get_info(self):
        return f"{self.name} ({Student.university}) - {len(self.courses)} courses"
    
    @classmethod
    def set_university(cls, name):
        cls.university = name
    
    @staticmethod
    def is_valid_id(sid):
        return isinstance(sid, int) and sid > 0

# Test
s = Student("Bob", 67890)
print(s.get_info())  # AI University
Student.set_university("Tech Institute")
print(s.get_info())  # Tech Institute
print(Student.is_valid_id(123))  # True
print(Student.is_valid_id(-5))   # False
```

### Solution 3: Magic Methods


```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
    
    def __str__(self):
        return f"{self.title} by {self.author} ({self.year})"
    
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self.year})"

# Test
b = Book("Deep Learning", "Ian Goodfellow", 2016)
print(str(b))   # Deep Learning by Ian Goodfellow (2016)
print(repr(b))  # Book('Deep Learning', 'Ian Goodfellow', 2016)
```

### Solution 5: Dataset Class


```python
class TextDataset:
    def __init__(self, texts, labels):
        assert len(texts) == len(labels), "Mismatched lengths"
        self.texts = texts
        self.labels = labels
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        return self.texts[idx], self.labels[idx]
    
    @property
    def vocab_size(self):
        unique_words = set()
        for text in self.texts:
            unique_words.update(text.split())
        return len(unique_words)

# Test
texts = ["hello world", "machine learning", "deep neural network"]
labels = [0, 1, 1]
ds = TextDataset(texts, labels)
print(f"Length: {len(ds)}")           # 3
print(f"Sample: {ds[0]}")            # ('hello world', 0)
print(f"Vocab size: {ds.vocab_size}")  # 6 unique words
```

### Solution 6: Metric Class


```python
import numpy as np

class PrecisionMetric:
    def __init__(self, num_classes, name="Precision"):
        self.num_classes = num_classes
        self.name = name
        self.reset()
    
    def reset(self):
        self.tp = np.zeros(self.num_classes)
        self.fp = np.zeros(self.num_classes)
    
    def update(self, predictions, targets):
        for c in range(self.num_classes):
            self.tp[c] += np.sum((predictions == c) & (targets == c))
            self.fp[c] += np.sum((predictions == c) & (targets != c))
    
    def compute(self):
        precision_per_class = np.divide(self.tp, self.tp + self.fp, 
                                        out=np.zeros_like(self.tp), 
                                        where=(self.tp + self.fp) != 0)
        macro = np.mean(precision_per_class)
        return {'per_class': precision_per_class, 'macro': macro}

# Test
metric = PrecisionMetric(num_classes=3)
preds = np.array([0, 1, 2, 0, 1, 2])
targets = np.array([0, 1, 1, 0, 2, 2])
metric.update(preds, targets)
result = metric.compute()
print(f"Per-class: {result['per_class']}")
print(f"Macro: {result['macro']:.2f}")
```

### Solution 8: Pipeline Class


```python
class Pipeline:
    def __init__(self, transforms=None):
        self.transforms = transforms or []
    
    def add_transform(self, func):
        self.transforms.append(func)
        return self  # Enable chaining
    
    def __call__(self, data):
        result = data
        for transform in self.transforms:
            result = transform(result)
        return result
    
    def __len__(self):
        return len(self.transforms)

# Test
pipeline = Pipeline()
pipeline.add_transform(str.lower).add_transform(str.strip).add_transform(lambda x: x.split())
result = pipeline("  Hello World Machine Learning  ")
print(result)  # ['hello', 'world', 'machine', 'learning']
print(f"Pipeline length: {len(pipeline)}")  # 3
```

### Solution 9: Encapsulation


```python
class SecureModel:
    def __init__(self, model_name, api_key):
        self.model_name = model_name      # Public
        self._weights = None              # Protected
        self.__api_key = api_key          # Private
    
    def load_weights(self, weights):
        self._weights = weights
    
    def predict(self, x):
        if self._weights is None:
            raise ValueError("Weights not loaded!")
        return f"Prediction from {self.model_name}"
    
    def get_api_key(self):
        return self.__api_key

# Test
model = SecureModel("ResNet", "secret-key-123")
print(model.model_name)          # ResNet
try:
    model.predict("input")
except ValueError as e:
    print(f"Error: {e}")         # Weights not loaded!

model.load_weights([0.1, 0.2])
print(model.predict("input"))    # Prediction from ResNet
print(model.get_api_key())     # secret-key-123
```

### Solution 10: Visualization


```python
import matplotlib.pyplot as plt
import numpy as np

class LivePlotter:
    def __init__(self, metric_names, figsize=(10, 4)):
        self.metric_names = metric_names
        self.figsize = figsize
        self.history = {name: [] for name in metric_names}
        self.steps = []
        self.fig, self.axes = plt.subplots(1, len(metric_names), figsize=figsize)
        if len(metric_names) == 1:
            self.axes = [self.axes]
    
    def update(self, metrics_dict):
        self.steps.append(len(self.steps))
        for name in self.metric_names:
            if name in metrics_dict:
                self.history[name].append(metrics_dict[name])
    
    def draw(self):
        for ax, name in zip(self.axes, self.metric_names):
            ax.clear()
            if self.history[name]:
                ax.plot(self.steps, self.history[name], 'b-', linewidth=2)
                ax.set_title(name)
                ax.set_xlabel('Step')
                ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.pause(0.01)

# Test with simulation
plotter = LivePlotter(['loss', 'accuracy'])
np.random.seed(42)

for step in range(50):
    loss = 1.0 * np.exp(-step/10) + np.random.normal(0, 0.05)
    acc = 0.5 + 0.45 * (1 - np.exp(-step/10)) + np.random.normal(0, 0.02)
    plotter.update({'loss': loss, 'accuracy': acc})
    if step % 10 == 0:
        plotter.draw()

plotter.draw()  # Final draw
plt.show()
```

### Solution 11: DataFrame Class


```python
class DataFrame:
    def __init__(self, data):
        self.data = data  # List of dicts
    
    def select(self, columns):
        new_data = [{k: row[k] for k in columns if k in row} for row in self.data]
        return DataFrame(new_data)
    
    def filter(self, condition_func):
        new_data = [row for row in self.data if condition_func(row)]
        return DataFrame(new_data)
    
    def head(self, n=5):
        for row in self.data[:n]:
            print(row)
        return self  # Return self for potential further chaining
    
    def __len__(self):
        return len(self.data)

# Test
data = [
    {'name': 'Alice', 'age': 25, 'score': 85},
    {'name': 'Bob', 'age': 30, 'score': 90},
    {'name': 'Charlie', 'age': 35, 'score': 78},
    {'name': 'Diana', 'age': 28, 'score': 92}
]

df = DataFrame(data)
print("Filtered and selected:")
df.select(['name', 'score']).filter(lambda r: r['score'] > 80).head()
```
