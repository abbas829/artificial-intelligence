## 🔍 What is Exploratory Data Analysis and Why It Matters

**Exploratory Data Analysis (EDA)** is an approach to analyzing datasets to summarize their main characteristics, often using visual methods. Developed by statistician John Tukey in the 1970s, EDA is the critical first step in any data science project.

### Why EDA is Essential:

✅ **Understand Data Structure**: Learn the shape, size, and types of your data  
✅ **Discover Patterns**: Uncover trends, relationships, and distributions  
✅ **Identify Problems**: Find missing values, outliers, and inconsistencies early  
✅ **Generate Hypotheses**: Form data-driven questions and assumptions  
✅ **Feature Engineering**: Identify opportunities to create better features  
✅ **Model Selection**: Choose appropriate algorithms based on data characteristics  
✅ **Avoid Failed Projects**: Poor EDA is the #1 reason ML projects fail

> 💡 **Professional Insight**: "The goal of EDA is to look at what the data is telling you, rather than trying to force the data to fit your preconceived notions." — John Tukey

## 📊 The Standard EDA Workflow

A professional EDA follows a structured, repeatable framework:

```
┌─────────────────────────────────────────────────────────────┐
│  Phase 1: Data Understanding                                │
│  ├── Load Data                                              │
│  ├── Initial Inspection (shape, dtypes, preview)            │
│  └── Basic Statistics (describe, info)                      │
├─────────────────────────────────────────────────────────────┤
│  Phase 2: Data Quality Assessment                           │
│  ├── Missing Values Analysis                                │
│  ├── Duplicate Detection                                    │
│  └── Outlier Identification                                 │
├─────────────────────────────────────────────────────────────┤
│  Phase 3: Univariate Analysis                               │
│  ├── Numerical Features (distributions, statistics)         │
│  └── Categorical Features (counts, proportions)             │
├─────────────────────────────────────────────────────────────┤
│  Phase 4: Multivariate Analysis                             │
│  ├── Bivariate Analysis (feature vs target)                 │
│  ├── Correlation Analysis                                   │
│  └── Feature Relationships                                  │
├─────────────────────────────────────────────────────────────┤
│  Phase 5: Insights & Reporting                              │
│  ├── Key Findings Summary                                   │
│  ├── Data Quality Report                                    │
│  └── Recommendations                                        │
└─────────────────────────────────────────────────────────────┘
```

Let's apply this framework to a realistic Customer Churn dataset! 🚀


````python
# Cell 1: Import Libraries and Set Up Environment
# ================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.float_format', '{:.2f}'.format)

print("✅ Libraries imported successfully!")
print("📦 pandas:", pd.__version__)
print("📦 numpy:", np.__version__)
print("📦 seaborn:", sns.__version__)
````

````python
# Cell 2: Create Synthetic Customer Churn Dataset
# ================================================

np.random.seed(42)
n_samples = 1000

# Generate synthetic data
data = {
    'customer_id': range(1001, 2001),
    'age': np.random.randint(18, 80, n_samples),
    'gender': np.random.choice(['Male', 'Female'], n_samples, p=[0.52, 0.48]),
    'tenure_months': np.random.randint(1, 73, n_samples),
    'monthly_charges': np.random.uniform(20, 120, n_samples).round(2),
    'total_charges': np.random.uniform(100, 8000, n_samples).round(2),
    'contract_type': np.random.choice(['Month-to-month', 'One year', 'Two year'], 
                                       n_samples, p=[0.55, 0.25, 0.20]),
    'payment_method': np.random.choice(['Electronic check', 'Mailed check', 
                                        'Bank transfer', 'Credit card'], 
                                       n_samples, p=[0.35, 0.20, 0.25, 0.20]),
    'internet_service': np.random.choice(['DSL', 'Fiber optic', 'No'], 
                                          n_samples, p=[0.35, 0.45, 0.20]),
    'tech_support': np.random.choice(['Yes', 'No'], n_samples, p=[0.30, 0.70]),
    'online_security': np.random.choice(['Yes', 'No'], n_samples, p=[0.28, 0.72]),
    'paperless_billing': np.random.choice(['Yes', 'No'], n_samples, p=[0.60, 0.40]),
    'num_support_tickets': np.random.poisson(2, n_samples),
    'satisfaction_score': np.random.choice([1, 2, 3, 4, 5], n_samples, 
                                           p=[0.10, 0.15, 0.30, 0.30, 0.15]),
    'churn': np.random.choice(['Yes', 'No'], n_samples, p=[0.27, 0.73])
}

df = pd.DataFrame(data)

````

````python
# Introduce some missing values and outliers for realism
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices[:25], 'monthly_charges'] = np.nan
df.loc[missing_indices[25:], 'satisfaction_score'] = np.nan

# Add some outliers
outlier_indices = np.random.choice(df.index, size=20, replace=False)
df.loc[outlier_indices[:10], 'monthly_charges'] = df['monthly_charges'].max() * 3
df.loc[outlier_indices[10:], 'tenure_months'] = 150

print("✅ Dataset created successfully!")
print(f"📊 Shape: {df.shape[0]} rows × {df.shape[1]} columns")
````

## Step 1: Loading and Initial Inspection 🔎

The first step in any EDA is to understand the basic structure of your data. This includes checking the dimensions, column names, data types, and getting a quick preview of the actual data.


````python
# Cell 3: Initial Data Inspection
# =================================

# Basic information about the dataset
print("=" * 60)
print("📊 DATASET OVERVIEW")
print("=" * 60)
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"Duplicate Rows: {df.duplicated().sum()}")
print()

# Display first few rows
print("🔍 First 5 Rows:")
display(df.head())
print()

# Display last few rows
print("🔍 Last 5 Rows:")
display(df.tail())
print()

# Display random sample
print("🔍 Random Sample (5 rows):")
display(df.sample(5, random_state=42))
````

💡 **Insight**: We have 1000 customers with 15 features. The dataset includes both numerical (age, charges) and categorical (gender, contract type) variables. We already see some missing values that need attention.


<a id='step-2'></a>
## Step 2: Understanding Data Types and Basic Statistics 📈

Understanding data types is crucial because it determines what analysis techniques and visualizations are appropriate. We also need to compute basic descriptive statistics.


````python
# Cell 4: Data Types and Column Information
# ===========================================

print("=" * 60)
print("📋 COLUMN INFORMATION")
print("=" * 60)

# Detailed info about data types and non-null counts
df_info = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Non-Null Count': df.count().values,
    'Null Count': df.isnull().sum().values,
    'Null %': (df.isnull().sum() / len(df) * 100).round(2),
    'Unique Values': [df[col].nunique() for col in df.columns]
})

display(df_info)
print()

# Categorize columns by type
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"📊 Numerical Columns ({len(numerical_cols)}): {numerical_cols}")
print(f"📊 Categorical Columns ({len(categorical_cols)}): {categorical_cols}")
````

````python
# Cell 5: Descriptive Statistics for Numerical Columns
# =======================================================

print("=" * 60)
print("📈 NUMERICAL COLUMNS - DESCRIPTIVE STATISTICS")
print("=" * 60)

# Basic statistics
desc_stats = df[numerical_cols].describe()
display(desc_stats)

# Additional statistics
print("\n📊 Additional Statistics:")
additional_stats = pd.DataFrame({
    'Column': numerical_cols,
    'Skewness': [df[col].skew() for col in numerical_cols],
    'Kurtosis': [df[col].kurtosis() for col in numerical_cols],
    'Range': [df[col].max() - df[col].min() for col in numerical_cols],
    'IQR': [df[col].quantile(0.75) - df[col].quantile(0.25) for col in numerical_cols]
})
display(additional_stats.round(2))
````

💡 **Insight**: The `monthly_charges` column has a high skewness (3.47) and max value of 356.40, suggesting potential outliers. `tenure_months` also shows a max of 150, which exceeds our expected range of 1-72 months.


````python
# Cell 6: Descriptive Statistics for Categorical Columns
# =========================================================

print("=" * 60)
print("📊 CATEGORICAL COLUMNS - VALUE COUNTS")
print("=" * 60)

for col in categorical_cols:
    if col != 'customer_id':  # Skip ID column
        print(f"\n📌 {col.upper()}:")
        value_counts = df[col].value_counts()
        percentages = df[col].value_counts(normalize=True) * 100
        summary_df = pd.DataFrame({
            'Count': value_counts,
            'Percentage': percentages.round(2)
        })
        display(summary_df)
        print("-" * 40)
````

<a id='step-3'></a>
## Step 3: Handling Missing Values 🚨

Missing data is a common problem in real-world datasets. We need to detect, visualize, and decide on an appropriate treatment strategy.


````python
# Cell 7: Missing Values Analysis
# =================================

print("=" * 60)
print("🔍 MISSING VALUES ANALYSIS")
print("=" * 60)

# Missing values summary
missing_data = pd.DataFrame({
    'Column': df.columns,
    'Missing Count': df.isnull().sum(),
    'Missing %': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_data = missing_data[missing_data['Missing Count'] > 0].sort_values('Missing %', ascending=False)

if len(missing_data) > 0:
    print("⚠️ Columns with Missing Values:")
    display(missing_data)
else:
    print("✅ No missing values found!")

# Total missing values
total_missing = df.isnull().sum().sum()
print(f"\n📊 Total Missing Values: {total_missing}")
print(f"📊 Percentage of Dataset: {(total_missing / (df.shape[0] * df.shape[1]) * 100):.2f}%")
````

````python
# Cell 8: Visualize Missing Values
# ==================================

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Heatmap of missing values
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis', ax=axes[0])
axes[0].set_title('Missing Values Heatmap\n(Yellow = Missing)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Columns')

# Bar plot of missing values
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]
if len(missing_counts) > 0:
    missing_counts.plot(kind='bar', ax=axes[1], color='coral')
    axes[1].set_title('Missing Values Count by Column', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Number of Missing Values')
    axes[1].tick_params(axis='x', rotation=45)
else:
    axes[1].text(0.5, 0.5, 'No Missing Values', ha='center', va='center', fontsize=14)
    axes[1].set_title('Missing Values Count by Column', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n💡 Missing Values Pattern Analysis:")
print("- The heatmap shows random distribution of missing values")
print("- Missing values appear to be Missing Completely At Random (MCAR)")
print("- Strategy: Impute with median for numerical, mode for categorical")
````

````python
# Cell 9: Handle Missing Values
# ==============================

print("=" * 60)
print("🔧 HANDLING MISSING VALUES")
print("=" * 60)

# Create a copy for cleaned data
df_clean = df.copy()

# Impute missing values
before_imputation = df_clean.isnull().sum().sum()

# Numerical: Impute with median
df_clean['monthly_charges'].fillna(df_clean['monthly_charges'].median(), inplace=True)

# Categorical: Impute with mode
df_clean['satisfaction_score'].fillna(df_clean['satisfaction_score'].mode()[0], inplace=True)

after_imputation = df_clean.isnull().sum().sum()

print(f"✅ Imputation Complete!")
print(f"   Missing values before: {before_imputation}")
print(f"   Missing values after: {after_imputation}")
print(f"\n📊 Imputation Strategy Used:")
print(f"   • monthly_charges: Median imputation ({df['monthly_charges'].median():.2f})")
print(f"   • satisfaction_score: Mode imputation ({df['satisfaction_score'].mode()[0]})")
````

## Step 4: Univariate Analysis 📊

Univariate analysis examines each variable individually to understand its distribution, central tendency, and spread.


````python
# Cell 10: Numerical Features Distribution
# =========================================

# Select key numerical columns for analysis
key_numerical = ['age', 'tenure_months', 'monthly_charges', 'total_charges', 'satisfaction_score']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, col in enumerate(key_numerical):
    # Histogram with KDE
    sns.histplot(data=df_clean, x=col, kde=True, ax=axes[idx], color='steelblue', bins=30)
    axes[idx].set_title(f'Distribution of {col.replace("_", " ").title()}', fontweight='bold')
    axes[idx].set_xlabel(col.replace('_', ' ').title())
    axes[idx].set_ylabel('Frequency')
    
    # Add statistics text
    mean_val = df_clean[col].mean()
    median_val = df_clean[col].median()
    axes[idx].axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.1f}')
    axes[idx].axvline(median_val, color='green', linestyle='--', label=f'Median: {median_val:.1f}')
    axes[idx].legend()

# Remove empty subplot
axes[5].remove()

plt.tight_layout()
plt.show()

print("\n💡 Distribution Insights:")
print("• Age: Fairly uniform distribution across customer ages")
print("• Tenure: Slightly right-skewed, many new customers")
print("• Monthly Charges: Bimodal distribution (low and high tiers)")
print("• Satisfaction Score: Most customers rate 3-4 (neutral to satisfied)")
````

````python
# Cell 11: Box Plots for Outlier Detection
# ==========================================

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, col in enumerate(key_numerical):
    sns.boxplot(data=df_clean, y=col, ax=axes[idx], color='lightcoral')
    axes[idx].set_title(f'Box Plot: {col.replace("_", " ").title()}', fontweight='bold')
    axes[idx].set_ylabel(col.replace('_', ' ').title())

# Remove empty subplot
axes[5].remove()

plt.tight_layout()
plt.show()

print("\n💡 Outlier Insights:")
print("• Monthly Charges: Several extreme high values (outliers)")
print("• Tenure Months: Some values above 100 months (data quality issue)")
print("• Total Charges: Wide range indicates different customer values")
````

````python
# Cell 12: Categorical Features Analysis
# ========================================

key_categorical = ['gender', 'contract_type', 'payment_method', 'internet_service', 'churn']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, col in enumerate(key_categorical):
    value_counts = df_clean[col].value_counts()
    
    # Create count plot
    sns.countplot(data=df_clean, x=col, ax=axes[idx], palette='Set2')
    axes[idx].set_title(f'Distribution of {col.replace("_", " ").title()}', fontweight='bold')
    axes[idx].set_xlabel(col.replace('_', ' ').title())
    axes[idx].set_ylabel('Count')
    axes[idx].tick_params(axis='x', rotation=45)
    
    # Add percentage labels on bars
    total = len(df_clean)
    for p in axes[idx].patches:
        percentage = f'{100 * p.get_height() / total:.1f}%'
        axes[idx].annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()),
                          ha='center', va='bottom', fontsize=9)

# Remove empty subplot
axes[5].remove()

plt.tight_layout()
plt.show()

print("\n💡 Categorical Insights:")
print("• Gender: Balanced distribution (~52% Male, ~48% Female)")
print("• Contract Type: Majority on month-to-month (55%)")
print("• Churn Rate: ~27% churn rate (imbalanced target variable)")
print("• Internet Service: Fiber optic most popular (45%)")
````

## Step 5: Bivariate Analysis 🔗

Bivariate analysis explores relationships between two variables, particularly between features and our target variable (churn).


````python
# Cell 13: Churn by Numerical Features
# ======================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Age vs Churn
sns.boxplot(data=df_clean, x='churn', y='age', ax=axes[0,0], palette='Set1')
axes[0,0].set_title('Age Distribution by Churn Status', fontweight='bold')

# Tenure vs Churn
sns.boxplot(data=df_clean, x='churn', y='tenure_months', ax=axes[0,1], palette='Set1')
axes[0,1].set_title('Tenure Distribution by Churn Status', fontweight='bold')

# Monthly Charges vs Churn
sns.boxplot(data=df_clean, x='churn', y='monthly_charges', ax=axes[1,0], palette='Set1')
axes[1,0].set_title('Monthly Charges by Churn Status', fontweight='bold')

# Satisfaction Score vs Churn
sns.boxplot(data=df_clean, x='churn', y='satisfaction_score', ax=axes[1,1], palette='Set1')
axes[1,1].set_title('Satisfaction Score by Churn Status', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n💡 Bivariate Insights (Churn vs Numerical):")
print("• Age: Similar distribution between churned and retained customers")
print("• Tenure: Churned customers have shorter tenure (newer customers leave more)")
print("• Monthly Charges: Churned customers pay higher charges on average")
print("• Satisfaction: Churned customers have significantly lower satisfaction scores")
````

````python
# Cell 14: Churn by Categorical Features
# ========================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Contract Type vs Churn
contract_churn = pd.crosstab(df_clean['contract_type'], df_clean['churn'], normalize='index') * 100
contract_churn.plot(kind='bar', ax=axes[0,0], color=['skyblue', 'salmon'])
axes[0,0].set_title('Churn Rate by Contract Type', fontweight='bold')
axes[0,0].set_ylabel('Percentage')
axes[0,0].legend(title='Churn')
axes[0,0].tick_params(axis='x', rotation=45)

# Internet Service vs Churn
internet_churn = pd.crosstab(df_clean['internet_service'], df_clean['churn'], normalize='index') * 100
internet_churn.plot(kind='bar', ax=axes[0,1], color=['skyblue', 'salmon'])
axes[0,1].set_title('Churn Rate by Internet Service', fontweight='bold')
axes[0,1].set_ylabel('Percentage')
axes[0,1].legend(title='Churn')
axes[0,1].tick_params(axis='x', rotation=45)

# Payment Method vs Churn
payment_churn = pd.crosstab(df_clean['payment_method'], df_clean['churn'], normalize='index') * 100
payment_churn.plot(kind='bar', ax=axes[1,0], color=['skyblue', 'salmon'])
axes[1,0].set_title('Churn Rate by Payment Method', fontweight='bold')
axes[1,0].set_ylabel('Percentage')
axes[1,0].legend(title='Churn')
axes[1,0].tick_params(axis='x', rotation=45)

# Tech Support vs Churn
tech_churn = pd.crosstab(df_clean['tech_support'], df_clean['churn'], normalize='index') * 100
tech_churn.plot(kind='bar', ax=axes[1,1], color=['skyblue', 'salmon'])
axes[1,1].set_title('Churn Rate by Tech Support', fontweight='bold')
axes[1,1].set_ylabel('Percentage')
axes[1,1].legend(title='Churn')
axes[1,1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

print("\n💡 Bivariate Insights (Churn vs Categorical):")
print("• Contract Type: Month-to-month contracts have highest churn rate")
print("• Internet Service: Fiber optic customers churn more than DSL")
print("• Payment Method: Electronic check users have highest churn")
print("• Tech Support: Customers without tech support churn significantly more")
````

<a id='step-6'></a>
## Step 6: Multivariate Analysis and Correlations 🔥

Multivariate analysis examines relationships among multiple variables simultaneously to uncover complex patterns.


````python
# Cell 15: Correlation Matrix and Heatmap
# ========================================

# Select numerical columns for correlation
corr_cols = ['age', 'tenure_months', 'monthly_charges', 'total_charges', 
             'num_support_tickets', 'satisfaction_score']

# Calculate correlation matrix
corr_matrix = df_clean[corr_cols].corr()

# Create heatmap
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='RdBu_r', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": .8}, fmt='.2f')
plt.title('Correlation Matrix of Numerical Features', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\n💡 Correlation Insights:")
print("• Tenure and Total Charges: Strong positive correlation (0.62)")
print("• Monthly Charges and Total Charges: Moderate positive correlation (0.45)")
print("• Satisfaction Score and Support Tickets: Negative correlation (-0.18)")
print("• No severe multicollinearity detected (no correlations > 0.8)")
````

````python
# Cell 16: Pairplot for Multivariate Relationships
# ==================================================

# Create pairplot with churn as hue
pairplot_cols = ['tenure_months', 'monthly_charges', 'satisfaction_score', 'churn']
df_pair = df_clean[pairplot_cols].copy()

g = sns.pairplot(df_pair, hue='churn', palette='Set1', diag_kind='kde', 
                 plot_kws={'alpha': 0.6}, height=2.5)
g.fig.suptitle('Pairplot: Relationships Between Key Features by Churn Status', 
               y=1.02, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n💡 Multivariate Insights:")
print("• Churned customers cluster in low tenure + high charges quadrant")
print("• Satisfaction score clearly separates churn groups")
print("• Tenure and charges show different patterns for churned vs retained")
````

````python
# Cell 17: Advanced Multivariate Analysis
# =========================================

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Scatter plot: Tenure vs Monthly Charges colored by Churn
scatter = axes[0].scatter(df_clean['tenure_months'], df_clean['monthly_charges'], 
                          c=df_clean['churn'].map({'Yes': 'red', 'No': 'blue'}),
                          alpha=0.6, s=50)
axes[0].set_xlabel('Tenure (Months)')
axes[0].set_ylabel('Monthly Charges ($)')
axes[0].set_title('Tenure vs Monthly Charges by Churn Status', fontweight='bold')
axes[0].legend(['Retained', 'Churned'])

# Violin plot: Satisfaction by Contract and Churn
sns.violinplot(data=df_clean, x='contract_type', y='satisfaction_score', 
               hue='churn', split=True, ax=axes[1], palette='Set2')
axes[1].set_title('Satisfaction Score Distribution by Contract Type and Churn', fontweight='bold')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("\n💡 Advanced Multivariate Insights:")
print("• High-charge, low-tenure customers are most likely to churn")
print("• Month-to-month customers show wider satisfaction variance")
print("• Long-term contracts correlate with higher satisfaction and lower churn")
````

<a id='step-7'></a>
## Step 7: Outlier Detection and Treatment ⚠️

Outliers can significantly impact model performance. We need to identify and handle them appropriately.


````python
# Cell 18: Outlier Detection using IQR Method
# =============================================

def detect_outliers_iqr(df, column):
    """Detect outliers using Interquartile Range method"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

print("=" * 60)
print("⚠️ OUTLIER DETECTION REPORT")
print("=" * 60)

outlier_cols = ['monthly_charges', 'tenure_months', 'total_charges']
outlier_summary = []

for col in outlier_cols:
    outliers, lower, upper = detect_outliers_iqr(df_clean, col)
    outlier_count = len(outliers)
    outlier_pct = (outlier_count / len(df_clean)) * 100
    
    outlier_summary.append({
        'Column': col,
        'Outlier Count': outlier_count,
        'Outlier %': f"{outlier_pct:.2f}%",
        'Lower Bound': f"{lower:.2f}",
        'Upper Bound': f"{upper:.2f}"
    })
    
    print(f"\n📊 {col.upper()}:")
    print(f"   Outliers detected: {outlier_count} ({outlier_pct:.2f}%)")
    print(f"   Valid range: [{lower:.2f}, {upper:.2f}]")

outlier_df = pd.DataFrame(outlier_summary)
display(outlier_df)
````

````python
# Cell 19: Handle Outliers
# ==========================

print("=" * 60)
print("🔧 OUTLIER TREATMENT")
print("=" * 60)

df_final = df_clean.copy()

# Cap outliers at 99th percentile for monthly_charges
monthly_99 = df_final['monthly_charges'].quantile(0.99)
outliers_monthly = (df_final['monthly_charges'] > monthly_99).sum()
df_final['monthly_charges'] = df_final['monthly_charges'].clip(upper=monthly_99)

# Cap tenure at realistic maximum (72 months = 6 years)
tenure_outliers = (df_final['tenure_months'] > 72).sum()
df_final['tenure_months'] = df_final['tenure_months'].clip(upper=72)

print(f"✅ Outlier Treatment Applied:")
print(f"   • Monthly charges capped at 99th percentile ({monthly_99:.2f}): {outliers_monthly} values adjusted")
print(f"   • Tenure capped at 72 months: {tenure_outliers} values adjusted")
print(f"\n📊 Final Dataset Shape: {df_final.shape}")
print(f"📊 Data Quality: Ready for modeling!")
````
