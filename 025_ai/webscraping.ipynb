## 1. Ethical Guidelines & Legal Considerations {#ethics}

⚠️ **Before writing any scraping code, understand these ethical guidelines:**

### The Responsible Scraper's Checklist:

1. **Check robots.txt**: Visit `website.com/robots.txt` to see which pages are off-limits
2. **Review Terms of Service**: Some websites explicitly prohibit scraping
3. **Respect Rate Limits**: Add delays between requests (1-3 seconds minimum)
4. **Don't Overload Servers**: Scrape during off-peak hours, limit concurrent requests
5. **Identify Yourself**: Use a descriptive User-Agent string with contact info
6. **Cache Responsibly**: Don't re-scrape the same data unnecessarily
7. **Respect Copyright**: Ensure you have rights to use the collected data
8. **Consider APIs First**: Many sites offer official APIs—use them when available

### Legal Landscape:
- Scraping public data is generally legal in most jurisdictions
- Bypassing authentication or technical barriers may violate laws (CFAA in US)
- EU GDPR and similar privacy laws restrict personal data collection
- Always consult legal counsel for commercial scraping projects

**Golden Rule**: If a website offers an API, use it. If you must scrape, be gentle, transparent, and respectful.

## 2. Setup and Imports {#setup}

Let's import the essential libraries for web scraping. We'll use `requests` for HTTP requests, `BeautifulSoup` for HTML parsing, and `pandas` for data manipulation.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import numpy as np
import json
from urllib.parse import urljoin, urlparse

# Set a descriptive User-Agent to identify our scraper
HEADERS = {
    'User-Agent': 'Python Learning Bot (your-email@example.com) - Educational scraping'
}

print("✅ Libraries imported successfully!")
print(f"Requests version: {requests.__version__}")
print(f"BeautifulSoup version: {BeautifulSoup.__version__ if hasattr(BeautifulSoup, '__version__') else '4.x'}")
print(f"Pandas version: {pd.__version__}")
print("\n⚠️ Remember: Always scrape ethically and respect robots.txt!")
```

## 3. HTTP Requests with Requests Library {#requests}

The first step in web scraping is fetching the webpage. The `requests` library makes this simple and Pythonic. Always check the response status code and handle errors gracefully.

```python
# Example: Fetching a webpage (using quotes.toscrape.com - a scraping-friendly educational site)
url = 'http://quotes.toscrape.com/'

try:
    # Send GET request with headers
    response = requests.get(url, headers=HEADERS, timeout=10)
    
    # Check if request was successful
    response.raise_for_status()
    
    print(f"✅ Success! Status code: {response.status_code}")
    print(f"Content type: {response.headers.get('content-type', 'unknown')}")
    print(f"Content length: {len(response.text)} characters")
    
    # Preview first 500 characters of HTML
    print("\n📄 HTML Preview (first 500 chars):")
    print(response.text[:500])
    
except requests.exceptions.RequestException as e:
    print(f"❌ Error fetching page: {e}")

# ⚠️ Always add a delay after requests
time.sleep(1)
```

## 4. Parsing HTML with BeautifulSoup {#parsing}

**BeautifulSoup** transforms messy HTML into a navigable Python object. It handles broken HTML gracefully and provides intuitive methods for searching and extracting data.

```python
# Parse the HTML content we fetched
soup = BeautifulSoup(response.text, 'html.parser')

# Basic soup properties
print("🔍 BeautifulSoup Object Analysis:")
print(f"Title: {soup.title.string if soup.title else 'No title found'}")
print(f"Title tag: {soup.title}")
print(f"\nFirst paragraph: {soup.find('p')}")

# Get text content (strips all HTML tags)
text_content = soup.get_text(separator='\n', strip=True)
print(f"\n📄 Text content length: {len(text_content)} characters")
print("First 300 characters of text:")
print(text_content[:300])
```

## 5. Navigating the HTML Tree {#navigation}

BeautifulSoup provides several methods to navigate and search the HTML tree:

- **`find()`**: Returns the first matching element
- **`find_all()`**: Returns a list of all matching elements
- **`select()`**: Uses CSS selectors for flexible searching

Let's explore these methods using the quotes website.

```python
# Method 1: find() - Get first occurrence
first_quote = soup.find('span', class_='text')
print("📝 First quote found:")
print(first_quote.get_text() if first_quote else "No quote found")

# Method 2: find_all() - Get all occurrences
all_quotes = soup.find_all('span', class_='text')
print(f"\n📊 Found {len(all_quotes)} quotes on this page")

print("\nFirst 3 quotes:")
for i, quote in enumerate(all_quotes[:3], 1):
    print(f"{i}. {quote.get_text()[:100]}...")

# Method 3: select() - CSS selectors (very powerful!)
authors = soup.select('small.author')  # CSS selector for author names
print(f"\n✍️ Found {len(authors)} authors using CSS selector")

print("\nAuthors on this page:")
for author in authors[:5]:
    print(f"- {author.get_text()}")
```

## 6. Extracting Text and Attributes {#extraction}

Beyond text content, we often need HTML attributes like `href` (links), `src` (images), `class`, or `id`. BeautifulSoup makes this straightforward.

```python
# Extract links and their text
links = soup.find_all('a')
print(f"🔗 Found {len(links)} links on the page")

print("\nSample links (first 5):")
for link in links[:5]:
    href = link.get('href', 'No href')
    text = link.get_text(strip=True)
    # Convert relative URLs to absolute
    full_url = urljoin(url, href)
    print(f"Text: '{text}' -> URL: {full_url}")

# Extract specific attributes from quote elements
quote_divs = soup.find_all('div', class_='quote')

print(f"\n📦 Processing {len(quote_divs)} quote containers:")

for i, div in enumerate(quote_divs[:3], 1):
    # Extract text
    text = div.find('span', class_='text')
    text_content = text.get_text() if text else 'N/A'
    
    # Extract author
    author = div.find('small', class_='author')
    author_name = author.get_text() if author else 'Unknown'
    
    # Extract tags (multiple per quote)
    tags = div.find_all('a', class_='tag')
    tag_list = [tag.get_text() for tag in tags]
    
    print(f"\nQuote {i}:")
    print(f"  Text: {text_content[:80]}...")
    print(f"  Author: {author_name}")
    print(f"  Tags: {', '.join(tag_list) if tag_list else 'None'}")
```

## 7. Handling Different HTML Structures {#structures}

Real-world websites have varied structures. Let's learn to handle common patterns and edge cases gracefully.

```python
# Pattern 1: Nested elements
print("🔍 Navigating nested structures:")
footer = soup.find('div', class_='col-md-4')
if footer:
    # Navigate through children
    for child in footer.find_all(['p', 'a'], recursive=False):
        print(f"Child tag: {child.name}, Text: {child.get_text(strip=True)[:50]}")

# Pattern 2: Parent navigation
first_author = soup.find('small', class_='author')
if first_author:
    parent_div = first_author.find_parent('div', class_='quote')
    print(f"\n👆 Parent div found: {parent_div is not None}")
    
    # Find siblings
    siblings = first_author.find_next_siblings()
    print(f"Siblings found: {len(siblings)}")

# Pattern 3: Handling missing elements safely
def safe_extract(element, selector, attr=None):
    """Safely extract data, return None if not found"""
    found = element.find(selector) if element else None
    if found and attr:
        return found.get(attr)
    elif found:
        return found.get_text(strip=True)
    return None

# Test safe extraction
test_quote = soup.find('div', class_='quote')
safe_text = safe_extract(test_quote, 'span', None)
safe_class = safe_extract(test_quote, 'span', 'class')

print(f"\n🛡️ Safe extraction results:")
print(f"Text found: {safe_text is not None}")
print(f"Class attribute: {safe_class}")
```

## 8. Cleaning Scraped Data {#cleaning}

Raw scraped data is messy—extra whitespace, special characters, and inconsistent formatting. Let's implement cleaning functions.

```python
import re

def clean_text(text):
    """
    Clean scraped text: remove extra whitespace, special chars, normalize
    """
    if not text:
        return ""
    
    # Convert to string if not already
    text = str(text)
    
    # Remove HTML entities and special chars (keep basic punctuation)
    text = re.sub(r'<[^>]+>', '', text)  # Remove any remaining HTML tags
    text = re.sub(r'\s+', ' ', text)      # Normalize whitespace
    text = text.strip()
    
    # Remove unicode quotes and normalize
    text = text.replace('"', '"').replace('"', '"')
    text = text.replace(''', "'").replace(''', "'")
    
    return text

def clean_author_name(name):
    """Clean author names specifically"""
    if not name:
        return "Unknown"
    
    # Remove titles, extra spaces
    name = re.sub(r'^(by|written by)\s+', '', name, flags=re.IGNORECASE)
    name = name.strip()
    
    # Title case
    return name.title()

# Test cleaning functions
raw_samples = [
    "  The   quick   brown   fox  ",
    '<span>"Quote with HTML"</span>',
    "by  albert einstein"
]

print("🧹 Testing cleaning functions:")
for raw in raw_samples:
    cleaned = clean_text(raw)
    print(f"Raw: '{raw}'")
    print(f"Clean: '{cleaned}'\n")

# Apply to scraped quotes
print("📊 Applying to scraped data:")
quotes_data = []
for div in soup.find_all('div', class_='quote')[:5]:
    text = clean_text(safe_extract(div, 'span', None))
    author = clean_author_name(safe_extract(div, 'small', None))
    quotes_data.append({'quote': text, 'author': author})

for item in quotes_data[:3]:
    print(f"Author: {item['author']}")
    print(f"Quote: {item['quote'][:80]}...\n")
```

## 9. Building a Scraper Function {#functions}

Let's encapsulate our scraping logic into a reusable, robust function with error handling and rate limiting.

```python
def scrape_quotes_page(url, headers=None, delay=1):
    """
    Scrape quotes from a single page with ethical delays and error handling
    
    Args:
        url: Page URL to scrape
        headers: Request headers (User-Agent, etc.)
        delay: Seconds to wait after request (rate limiting)
    
    Returns:
        list: List of dictionaries containing quote data
    """
    if headers is None:
        headers = HEADERS
    
    quotes_list = []
    
    try:
        # Respectful delay before request
        print(f"⏳ Waiting {delay}s before requesting {url}...")
        time.sleep(delay)
        
        # Fetch page
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        
        # Parse
        soup = BeautifulSoup(response.text, 'html.parser')
        quote_divs = soup.find_all('div', class_='quote')
        
        print(f"✅ Found {len(quote_divs)} quotes on {url}")
        
        # Extract data
        for div in quote_divs:
            text = clean_text(safe_extract(div, 'span', None))
            author = clean_author_name(safe_extract(div, 'small', None))
            
            # Get tags
            tags = [tag.get_text() for tag in div.find_all('a', class_='tag')]
            
            # Get author link for potential follow-up
            author_link = div.find('a')
            author_url = urljoin(url, author_link.get('href')) if author_link else None
            
            quotes_list.append({
                'quote': text,
                'author': author,
                'tags': tags,
                'author_url': author_url,
                'source_url': url
            })
            
    except requests.exceptions.RequestException as e:
        print(f"❌ Error scraping {url}: {e}")
    except Exception as e:
        print(f"❌ Unexpected error: {e}")
    
    return quotes_list

# Test the function
test_url = 'http://quotes.toscrape.com/'
scraped_data = scrape_quotes_page(test_url, delay=1)

print(f"\n📦 Scraped {len(scraped_data)} quotes successfully")
print("\nSample entry:")
if scraped_data:
    print(json.dumps(scraped_data[0], indent=2))
```

## 11. Handling Pagination {#pagination}

Most websites spread content across multiple pages. Let's learn to navigate pagination to scrape complete datasets.

```python
def scrape_multiple_pages(base_url, max_pages=3, delay=2):
    """
    Scrape quotes from multiple pages with pagination
    
    Args:
        base_url: Starting URL
        max_pages: Maximum pages to scrape (safety limit)
        delay: Delay between pages
    """
    all_quotes = []
    current_url = base_url
    page_count = 0
    
    while current_url and page_count < max_pages:
        page_count += 1
        print(f"\n📄 Scraping page {page_count}: {current_url}")
        
        # Scrape current page
        page_quotes = scrape_quotes_page(current_url, delay=delay)
        all_quotes.extend(page_quotes)
        
        # Find next page link
        try:
            response = requests.get(current_url, headers=HEADERS, timeout=10)
            soup = BeautifulSoup(response.text, 'html.parser')
            
            next_link = soup.find('li', class_='next')
            if next_link and next_link.find('a'):
                next_href = next_link.find('a').get('href')
                current_url = urljoin(base_url, next_href)
            else:
                print("🏁 No more pages found.")
                current_url = None
                
        except Exception as e:
            print(f"❌ Error finding next page: {e}")
            current_url = None
    
    print(f"\n✅ Total quotes scraped: {len(all_quotes)} from {page_count} pages")
    return all_quotes

# Scrape first 2 pages (be respectful!)
all_quotes_data = scrape_multiple_pages('http://quotes.toscrape.com/', max_pages=2, delay=2)

# Convert to DataFrame
df_all = pd.DataFrame(all_quotes_data)
df_all['tags'] = df_all['tags'].apply(lambda x: ', '.join(x) if x else '')

print(f"\n📊 Final dataset shape: {df_all.shape}")
print(f"Unique authors: {df_all['author'].nunique()}")
print(f"Most common authors:")
print(df_all['author'].value_counts().head())
```

## 12. Real AI/ML Examples {#ml-examples}

Let's apply our scraping skills to realistic ML data collection scenarios: sentiment analysis datasets, NLP text corpora, and structured tabular data.

```python
# Example 1: Building a Sentiment Analysis Dataset from Reviews
# Simulating product review scraping (using quotes as proxy for text data)

def build_text_classification_dataset(quotes_data):
    """
    Transform scraped quotes into a format suitable for text classification/NLP
    """
    # Create labels based on quote length as a proxy sentiment (demo purposes)
    dataset = []
    for item in quotes_data:
        text = item['quote']
        
        # Create features for ML
        features = {
            'text': text,
            'author': item['author'],
            'text_length': len(text),
            'word_count': len(text.split()),
            'has_question': '?' in text,
            'has_exclamation': '!' in text,
            'category': item['tags'][0] if item['tags'] else 'general'
        }
        
        # Simple sentiment proxy (in real scenario, you'd use actual sentiment)
        features['sentiment_proxy'] = 'positive' if '!' in text else 'neutral'
        
        dataset.append(features)
    
    return pd.DataFrame(dataset)

nlp_df = build_text_classification_dataset(all_quotes_data[:10])
print("🤖 NLP Dataset Preview (first 3 rows):")
print(nlp_df[['text', 'text_length', 'word_count', 'category']].head(3))

print(f"\n📊 Dataset statistics:")
print(f"Avg text length: {nlp_df['text_length'].mean():.1f} chars")
print(f"Avg word count: {nlp_df['word_count'].mean():.1f} words")
print(f"Categories: {nlp_df['category'].value_counts().to_dict()}")
```

```python
# Example 2: Extracting Structured Tabular Data
# Many websites have tables—let's scrape them effectively

def extract_tables(soup):
    """Extract all tables from a page into DataFrames"""
    tables = soup.find_all('table')
    dataframes = []
    
    for i, table in enumerate(tables):
        try:
            # Extract headers
            headers = []
            header_row = table.find('thead')
            if header_row:
                headers = [th.get_text(strip=True) for th in header_row.find_all('th')]
            
            # Extract rows
            rows = []
            for tr in table.find_all('tr'):
                row_data = [td.get_text(strip=True) for td in tr.find_all(['td', 'th'])]
                if row_data and any(row_data):  # Skip empty rows
                    rows.append(row_data)
            
            # Create DataFrame
            if rows:
                if headers and len(headers) == len(rows[0]):
                    df = pd.DataFrame(rows[1:], columns=headers)  # Skip header row in data
                else:
                    df = pd.DataFrame(rows)
                dataframes.append(df)
                
        except Exception as e:
            print(f"Error processing table {i}: {e}")
    
    return dataframes

# Note: quotes.toscrape.com doesn't have tables, so this is a template
print("📋 Table extraction function ready!")
print("Usage: dfs = extract_tables(soup)")
print("This handles thead/tbody structures and varying column counts.")
```

```python
# Example 3: Complete ML Data Collection Pipeline
def collect_ml_dataset(seed_urls, target_schema, max_pages_per_site=2):
    """
    Complete pipeline for collecting ML training data
    
    Args:
        seed_urls: List of URLs to start scraping
        target_schema: Dict defining what fields to extract
        max_pages_per_site: Safety limit per site
    """
    collected_data = []
    
    for url in seed_urls:
        print(f"\n🌐 Processing: {url}")
        
        try:
            # Respect robots.txt (simplified check)
            parsed = urlparse(url)
            robots_url = f"{parsed.scheme}://{parsed.netloc}/robots.txt"
            print(f"⚠️ Remember to check: {robots_url}")
            
            # Scrape with limits
            site_data = scrape_multiple_pages(url, max_pages=max_pages_per_site, delay=2)
            
            # Transform to target schema
            for item in site_data:
                record = {}
                for field, source in target_schema.items():
                    if callable(source):
                        record[field] = source(item)
                    else:
                        record[field] = item.get(source, None)
                collected_data.append(record)
                
        except Exception as e:
            print(f"❌ Failed to process {url}: {e}")
    
    return pd.DataFrame(collected_data)

# Define schema for sentiment analysis dataset
schema = {
    'text': 'quote',
    'label': lambda x: 'inspirational' if 'life' in x.get('quote', '').lower() else 'general',
    'source': 'source_url',
    'author': 'author',
    'metadata': lambda x: {'tags': x.get('tags', [])}
}

print("🎯 ML Dataset Collection Pipeline defined!")
print("Schema:", schema)
print("\nThis pattern allows you to:")
print("- Collect from multiple sources")
print("- Transform to consistent format")
print("- Add derived features (labels, metadata)")
print("- Maintain provenance (source tracking)")
```

### Solution 3: Clean the Text
Text cleaning function to normalize quotes.

```python
import re

def clean_quote_text(text):
    """Remove quotes and normalize whitespace"""
    if not text:
        return ""
    
    # Remove surrounding quotes
    text = text.strip().strip('"').strip("'")
    
    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()

# Apply to all quotes
for item in quotes_list:
    item['quote'] = clean_quote_text(item['quote'])

print("Sample cleaned quote:")
print(quotes_list[0]['quote'][:100])

# Expected: Clean text without surrounding quotes or extra spaces
```

### Solution 4: Reusable Scraper Function
Complete function with error handling and delays.

```python
def scrape_quotes(url, headers=None, delay=1):
    """Scrape quotes from a URL with error handling"""
    if headers is None:
        headers = {'User-Agent': 'Python Learning Bot'}
    
    quotes = []
    
    try:
        time.sleep(delay)
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, 'html.parser')
        
        for div in soup.find_all('div', class_='quote'):
            text_elem = div.find('span', class_='text')
            author_elem = div.find('small', class_='author')
            
            if text_elem and author_elem:
                quotes.append({
                    'quote': clean_quote_text(text_elem.get_text()),
                    'author': author_elem.get_text()
                })
        
        print(f"✅ Scraped {len(quotes)} quotes from {url}")
        
    except requests.exceptions.RequestException as e:
        print(f"❌ Error: {e}")
    
    return quotes

# Test
test_quotes = scrape_quotes('http://quotes.toscrape.com/')

# Expected: Function returns list of quotes with error handling
```

### Solution 6: Handle Pagination (2 Pages)
Scrape multiple pages with next link detection.

```python
def scrape_two_pages(start_url):
    """Scrape exactly 2 pages with pagination"""
    all_quotes = []
    current_url = start_url
    
    for page in range(1, 3):
        print(f"Scraping page {page}: {current_url}")
        
        response = requests.get(current_url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extract quotes
        for div in soup.find_all('div', class_='quote'):
            text = div.find('span', class_='text').get_text()
            author = div.find('small', class_='author').get_text()
            all_quotes.append({'quote': text, 'author': author})
        
        # Find next page
        next_li = soup.find('li', class_='next')
        if next_li and next_li.find('a'):
            next_href = next_li.find('a').get('href')
            current_url = urljoin(start_url, next_href)
        else:
            break
        
        time.sleep(2)
    
    return all_quotes

two_page_quotes = scrape_two_pages('http://quotes.toscrape.com/')
print(f"\nTotal quotes from 2 pages: {len(two_page_quotes)}")

# Expected: Approximately 20 quotes (10 per page) from 2 pages
```

### Solution 10: Error Handling Challenge
Robust scraper with comprehensive error handling.

```python
def robust_scraper(url, max_retries=3):
    """Scraper with comprehensive error handling"""
    quotes = []
    
    for attempt in range(max_retries):
        try:
            time.sleep(1)
            response = requests.get(url, headers=headers, timeout=5)
            
            # Handle 404
            if response.status_code == 404:
                print(f"❌ Page not found: {url}")
                return []
            
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')
            
            for div in soup.find_all('div', class_='quote'):
                # Handle missing elements
                text_elem = div.find('span', class_='text')
                author_elem = div.find('small', class_='author')
                
                if not text_elem:
                    print("⚠️ Warning: Missing quote text")
                    continue
                
                text = text_elem.get_text()
                if not text.strip():
                    print("⚠️ Warning: Empty quote text")
                    continue
                
                author = author_elem.get_text() if author_elem else 'Unknown'
                
                quotes.append({
                    'quote': text,
                    'author': author
                })
            
            print(f"✅ Success on attempt {attempt + 1}")
            break
            
        except requests.exceptions.Timeout:
            print(f"⏱️ Timeout on attempt {attempt + 1}, retrying...")
        except requests.exceptions.RequestException as e:
            print(f"❌ Request error: {e}")
            break
        except Exception as e:
            print(f"❌ Unexpected error: {e}")
            break
    
    return quotes

# Test with valid URL
test_results = robust_scraper('http://quotes.toscrape.com/')
print(f"\nScraped {len(test_results)} quotes with robust handling")

# Expected: Handles all edge cases gracefully with appropriate messages
```
