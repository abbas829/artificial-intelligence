## 1. Setup & Imports

Let's import the essential libraries and configure our plotting environment for beautiful, publication-ready visualizations.

```python
# Core data and visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set aesthetic parameters for consistent, beautiful plots
sns.set_style('whitegrid')
sns.set_palette('husl')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 11

# Load example datasets
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')

print('✅ Libraries imported successfully!')
print(f'Tips dataset: {tips.shape[0]} rows, {tips.shape[1]} columns')
print(f'Titanic dataset: {titanic.shape[0]} rows, {titanic.shape[1]} columns')
print(f'Iris dataset: {iris.shape[0]} rows, {iris.shape[1]} columns')
```

<a id='categorical'></a>
## 2. Categorical Plots Overview

Seaborn's categorical plots help us visualize the relationship between categorical variables and numeric variables. The `catplot()` function is a versatile interface that can create various categorical plot types.

```python
# 📊 catplot() overview - the unified interface for categorical plots
# Let's explore different 'kind' parameters: box, violin, boxen, strip, swarm, bar, point

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Box plot - shows quartiles and outliers
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0,0], palette='Set2')
axes[0,0].set_title('📦 Box Plot: Total Bill by Day', fontsize=14, fontweight='bold')

# Violin plot - shows distribution shape
sns.violinplot(data=tips, x='day', y='total_bill', ax=axes[0,1], palette='Set2')
axes[0,1].set_title('🎻 Violin Plot: Distribution Shape by Day', fontsize=14, fontweight='bold')

# Boxen plot - enhanced box plot for larger datasets
sns.boxenplot(data=tips, x='day', y='total_bill', ax=axes[1,0], palette='Set2')
axes[1,0].set_title('📦 Boxen Plot: Detailed Quantiles', fontsize=14, fontweight='bold')

# Bar plot - shows central tendency with confidence intervals
sns.barplot(data=tips, x='day', y='total_bill', ax=axes[1,1], palette='Set2')
axes[1,1].set_title('📊 Bar Plot: Mean Total Bill by Day', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print('💡 Key Insight: Box plots show outliers, violin plots reveal distribution shape,')
print('   boxen plots provide more quantile details, and bar plots emphasize central tendency.')
```

### 🎨 Box, Violin, and Boxen Plots - Comparing Distributions

These plots are essential for comparing distributions across categories. **Box plots** highlight outliers and quartiles, **violin plots** show the full distribution shape (including multimodality), and **boxen plots** provide a more detailed view of quantiles for larger datasets.

```python
# Advanced comparison with hue parameter
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Box plot with hue - comparing across two categorical variables
sns.boxplot(data=titanic, x='class', y='age', hue='sex', ax=axes[0], palette='coolwarm')
axes[0].set_title('📦 Age Distribution by Class & Gender', fontsize=13, fontweight='bold')
axes[0].legend(title='Gender', loc='upper right')

# Violin plot with split for better comparison
sns.violinplot(data=titanic, x='class', y='age', hue='sex', split=True, ax=axes[1], palette='coolwarm')
axes[1].set_title('🎻 Split Violin: Age by Class & Gender', fontsize=13, fontweight='bold')
axes[1].legend(title='Gender', loc='upper right')

# Boxen plot for detailed quantile analysis
sns.boxenplot(data=titanic.dropna(), x='class', y='fare', hue='survived', ax=axes[2], palette='RdYlGn')
axes[2].set_title('📦 Fare Distribution by Class & Survival', fontsize=13, fontweight='bold')
axes[2].legend(title='Survived', loc='upper right')

plt.tight_layout()
plt.show()

print('✅ ML Application: Use these plots to identify class imbalances, outliers that need handling,')
print('   and feature distributions that might affect model performance.')
```

### 🐝 Strip Plots & Swarm Plots - Showing Individual Data Points

While box and violin plots summarize distributions, **strip plots** and **swarm plots** show every individual data point. This is crucial for understanding sample sizes and spotting clustering patterns that summary statistics might hide.

```python
# Individual data point visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Strip plot - simple scatter by category
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', 
              ax=axes[0], palette='viridis', alpha=0.7, size=6, jitter=True)
axes[0].set_title('🐝 Strip Plot: Individual Bills by Day', fontsize=14, fontweight='bold')
axes[0].legend(title='Gender')

# Swarm plot - non-overlapping points for clearer view
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', 
              ax=axes[1], palette='viridis', size=6)
axes[1].set_title('🐝 Swarm Plot: Non-Overlapping Distribution', fontsize=14, fontweight='bold')
axes[1].legend(title='Gender')

plt.tight_layout()
plt.show()

print('💡 Tip: Swarm plots are great for smaller datasets. For large datasets (>1000 points),')
print('   consider using strip plots with transparency or sampling.')
```

### 📊 Bar Plots & Point Plots - Central Tendency & Trends

**Bar plots** display means (or other estimators) with confidence intervals, perfect for comparing group averages. **Point plots** connect estimates with lines, making them excellent for visualizing trends across ordered categories.

```python
# Central tendency visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar plot with confidence intervals
sns.barplot(data=titanic, x='class', y='survived', hue='sex', 
            ax=axes[0], palette='Spectral', capsize=0.1)
axes[0].set_title('📊 Survival Rate by Class & Gender', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Survival Probability')
axes[0].legend(title='Gender')

# Point plot - excellent for showing trends
sns.pointplot(data=titanic, x='class', y='survived', hue='sex', 
              ax=axes[1], palette='Spectral', markers=['o', 's'], linestyles=['-', '--'])
axes[1].set_title('📈 Survival Trends Across Classes', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Survival Probability')
axes[1].legend(title='Gender')

plt.tight_layout()
plt.show()

print('🎯 ML Insight: Point plots clearly show the interaction effect between class and gender')
print('   on survival - valuable for feature interaction engineering!')
```

### 🔢 Count Plots - Frequency Analysis

**Count plots** are specialized bar plots for showing the frequency of categorical variables. Essential for understanding class distributions and identifying imbalanced datasets — a critical step before ML model training.

```python
# Frequency analysis with count plots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Basic count plot
sns.countplot(data=titanic, x='class', hue='survived', ax=axes[0], palette='RdYlBu')
axes[0].set_title('🔢 Passenger Count by Class & Survival', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
axes[0].legend(title='Survived', labels=['No', 'Yes'])

# Horizontal count plot for better readability with many categories
sns.countplot(data=tips, y='day', hue='time', ax=axes[1], palette='Set1')
axes[1].set_title('🔢 Restaurant Visits by Day & Time', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Count')
axes[1].legend(title='Meal Time')

plt.tight_layout()
plt.show()

print('⚠️ Data Science Alert: Notice the class imbalance in survival rates!')
print('   This would require stratified sampling or class weighting in ML models.')
```

## 3. Relational Plots

Relational plots visualize the relationship between two or more numeric variables. They are fundamental for understanding correlations, clusters, and patterns in your data — insights that directly inform feature selection and model architecture decisions.

```python
# Scatter and line plots with multiple encodings
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Basic scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[0,0], 
                color='steelblue', alpha=0.7, s=80)
axes[0,0].set_title('💠 Scatter Plot: Bill vs Tip', fontsize=13, fontweight='bold')

# Scatter with hue (color encoding)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', 
                ax=axes[0,1], palette='deep', s=80, alpha=0.8)
axes[0,1].set_title('💠 Scatter with Hue: Time of Day', fontsize=13, fontweight='bold')
axes[0,1].legend(title='Meal Time')

# Scatter with hue and size (multivariate)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', size='size', 
                ax=axes[1,0], palette='husl', alpha=0.7, sizes=(50, 300))
axes[1,0].set_title('💠 Multivariate: Day (color) & Party Size (size)', fontsize=13, fontweight='bold')
axes[1,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Line plot with confidence intervals
# Aggregate tips by bill amount bins for clearer trend
tips['bill_bin'] = pd.cut(tips['total_bill'], bins=10)
sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex', 
             ax=axes[1,1], palette='Dark2', marker='o', ci=95)
axes[1,1].set_title('📈 Line Plot: Tip Trend by Gender', fontsize=13, fontweight='bold')
axes[1,1].legend(title='Gender')

plt.tight_layout()
plt.show()

print('🎨 Visualization Power: Notice how adding hue and size encodings reveals')
print('   4-dimensional relationships in a single 2D plot!')
```

### 🎭 Faceted Relational Plots with relplot()

The `relplot()` function creates **faceted plots** — multiple subplots showing different subsets of data. This is incredibly powerful for understanding how relationships vary across categories, a technique often used in segmentation analysis and cohort studies.

```python
# Faceted relational plots
# Using relplot to create a grid of scatter plots

g = sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='time',      # Separate columns for lunch/dinner
    row='smoker',    # Separate rows for smoker status
    style='sex',     # Different markers for gender
    palette='Set2',
    alpha=0.7,
    height=4,
    aspect=1.2
)

g.fig.suptitle('🎭 Faceted Scatter: Tips by Time, Smoking Status & Gender', 
               fontsize=16, fontweight='bold', y=1.02)
g.set_axis_labels('Total Bill ($)', 'Tip Amount ($)')
plt.show()

print('🔍 Analysis Insight: Faceting reveals that the bill-tip relationship varies')
print('   by meal time and smoking status - important segmentation variables!')
```

## 4. Real AI/ML Examples

Let's apply these visualization techniques to realistic machine learning scenarios: analyzing target variables, model performance, and feature importance across categories.

```python
# Example 1: Target Variable Analysis (Titanic Survival)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Survival by passenger class and gender
sns.barplot(data=titanic, x='class', y='survived', hue='sex', 
            ax=axes[0,0], palette='coolwarm', ci=95)
axes[0,0].set_title('🎯 Survival Rate by Class & Gender', fontsize=13, fontweight='bold')
axes[0,0].set_ylabel('Survival Probability')

# Age distribution by survival status
sns.violinplot(data=titanic.dropna(), x='survived', y='age', 
               ax=axes[0,1], palette='RdYlGn')
axes[0,1].set_title('👶 Age Distribution by Survival', fontsize=13, fontweight='bold')
axes[0,1].set_xticklabels(['Did Not Survive', 'Survived'])

# Fare vs Age colored by survival
sns.scatterplot(data=titanic.dropna(), x='age', y='fare', hue='survived', 
                ax=axes[1,0], palette='RdYlGn', alpha=0.6, s=60)
axes[1,0].set_title('💰 Fare vs Age by Survival Status', fontsize=13, fontweight='bold')
axes[1,0].legend(title='Survived', labels=['No', 'Yes'])

# Count plot showing class imbalance
sns.countplot(data=titanic, x='class', hue='survived', ax=axes[1,1], palette='RdYlBu')
axes[1,1].set_title('📊 Class Distribution & Survival Counts', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()

print('🤖 ML Takeaway: Clear patterns emerge - gender, class, and age are strong predictors.')
print('   The class imbalance in the target variable requires careful handling.')
```

```python
# Example 2: Model Performance Simulation Across Categories
# Simulating model accuracy scores across different algorithms and datasets

np.random.seed(42)
n_samples = 100

performance_data = pd.DataFrame({
    'Algorithm': ['Random Forest'] * n_samples + ['XGBoost'] * n_samples + 
                 ['SVM'] * n_samples + ['Logistic Regression'] * n_samples,
    'Dataset': (['Training'] * (n_samples//2) + ['Test'] * (n_samples//2)) * 4,
    'Accuracy': list(np.random.normal(0.92, 0.03, n_samples)) + 
                list(np.random.normal(0.94, 0.02, n_samples)) + 
                list(np.random.normal(0.89, 0.04, n_samples)) + 
                list(np.random.normal(0.87, 0.03, n_samples))
})

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Box plot comparing algorithms
sns.boxplot(data=performance_data, x='Algorithm', y='Accuracy', hue='Dataset', 
            ax=axes[0], palette='Set3')
axes[0].set_title('🤖 Model Performance Comparison', fontsize=14, fontweight='bold')
axes[0].set_ylim(0.75, 1.0)
axes[0].tick_params(axis='x', rotation=15)

# Point plot showing trends
sns.pointplot(data=performance_data, x='Algorithm', y='Accuracy', hue='Dataset', 
              ax=axes[1], palette='Set2', markers=['o', 's'], linestyles=['-', '--'])
axes[1].set_title('📈 Performance Trends: Training vs Test', fontsize=14, fontweight='bold')
axes[1].tick_params(axis='x', rotation=15)
axes[1].set_ylim(0.75, 1.0)

plt.tight_layout()
plt.show()

print('📊 Model Selection Insight: XGBoost shows highest accuracy with low variance,')
print('   while the gap between training and test scores indicates overfitting levels.')
```

```python
# Example 3: Feature Importance by Category
# Simulating feature importance scores across different customer segments

features = ['Age', 'Income', 'Tenure', 'Usage', 'Support_Calls']
segments = ['Premium', 'Standard', 'Basic']

importance_data = []
for segment in segments:
    if segment == 'Premium':
        importances = [0.15, 0.35, 0.25, 0.20, 0.05]
    elif segment == 'Standard':
        importances = [0.20, 0.30, 0.20, 0.25, 0.05]
    else:
        importances = [0.25, 0.20, 0.15, 0.30, 0.10]
    
    for feat, imp in zip(features, importances):
        # Add some noise for realism
        for _ in range(50):
            importance_data.append({
                'Feature': feat,
                'Segment': segment,
                'Importance': imp + np.random.normal(0, 0.02)
            })

importance_df = pd.DataFrame(importance_data)

# Create faceted bar plot
g = sns.catplot(
    data=importance_df,
    x='Feature',
    y='Importance',
    hue='Segment',
    kind='bar',
    palette='viridis',
    height=6,
    aspect=1.5,
    legend_out=False
)

g.fig.suptitle('🔍 Feature Importance by Customer Segment', fontsize=16, fontweight='bold', y=1.02)
g.set_axis_labels('Feature', 'Importance Score')
g.set_xticklabels(rotation=30, ha='right')
plt.tight_layout()
plt.show()

print('🎯 Business Insight: Different features drive predictions for different segments!')
print('   Consider building separate models or using segment as a feature.')
```

### ✅ Exercise 1 Solution: Basic Box Plot

**Task:** Create a box plot showing the distribution of `sepal_length` for each `species` in the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length', palette='Set2')
plt.title('📦 Sepal Length Distribution by Species', fontsize=14, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.show()

# Expected: Clear separation between setosa and other species
print('✅ Result: Setosa has distinctly shorter sepals, showing good class separability.')
```

### ✅ Exercise 2 Solution: Violin Plot with Hue

### ✅ Exercise 4 Solution: Bar Plot with Confidence Intervals

**Task:** Create a bar plot showing mean tip percentage by day with 95% confidence intervals.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100

plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='tip_pct', ci=95, palette='husl', capsize=0.1)
plt.title('📊 Mean Tip Percentage by Day (with 95% CI)', fontsize=14, fontweight='bold')
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Tip Percentage (%)', fontsize=12)
plt.show()

# Expected: Sunday may show higher tip percentage with visible error bars
print('✅ Result: Error bars show confidence intervals. Sunday tips average around 16-17%.')
```

### ✅ Exercise 9 Solution: Combined Categorical Analysis

**Task:** Create side-by-side box and violin plots with multiple encodings.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Box plot: Age by Class and Survival
sns.boxplot(data=titanic.dropna(), x='class', y='age', hue='survived', 
            ax=axes[0], palette='coolwarm')
axes[0].set_title('📦 Age Distribution by Class & Survival', fontsize=13, fontweight='bold')
axes[0].legend(title='Survived', labels=['No', 'Yes'])

# Violin plot: Fare by Survival and Class
sns.violinplot(data=titanic.dropna(), x='survived', y='fare', hue='class', 
               ax=axes[1], palette='coolwarm', split=False)
axes[1].set_title('🎻 Fare Distribution by Survival & Class', fontsize=13, fontweight='bold')
axes[1].set_xticklabels(['Did Not Survive', 'Survived'])
axes[1].legend(title='Class', loc='upper right')

plt.tight_layout()
plt.show()

# Expected: Left shows age patterns, right shows fare differences
print('✅ Result: Survivors paid higher fares (right plot). Age varies by class (left plot).')
```

### ✅ Exercise 10 Solution: Custom Palette and Styling

**Task:** Create a count plot with custom colors and styling.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style('darkgrid')

tips = sns.load_dataset('tips')

# Custom palette
custom_palette = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

plt.figure(figsize=(10, 6))
sns.countplot(data=tips, x='day', palette=custom_palette)
plt.title('🔢 Restaurant Traffic by Day (Custom Styling)', fontsize=14, fontweight='bold')
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)

# Add value labels on bars
for i, v in enumerate(tips['day'].value_counts().sort_index()):
    plt.text(i, v + 1, str(v), ha='center', fontweight='bold')

plt.show()

# Reset style
sns.set_style('whitegrid')

print('✅ Result: Saturday is the busiest day. Custom colors make the plot visually distinctive.')
```

### ✅ Exercise 11 Solution: Advanced Faceting

**Task:** Create an advanced faceted violin plot with catplot().

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

g = sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='time',
    row='smoker',
    col='sex',
    kind='violin',
    split=True,
    palette='muted',
    height=4,
    aspect=1.2
)

g.fig.suptitle('🎭 Advanced Faceting: Bills by Day, Time, Smoking & Gender', 
               fontsize=16, fontweight='bold', y=1.02)
g.set_axis_labels('Day of Week', 'Total Bill ($)')
plt.tight_layout()
plt.show()

# Expected: 8 subplots (2 smoker x 2 sex x 2 time shown in split violins)
print('✅ Result: 8-panel grid reveals complex interactions. Smokers show different patterns by gender.')
```
