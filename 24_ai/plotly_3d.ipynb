## 2. 3D Scatter Plots {#3d-scatter}

3D scatter plots are the foundation of three-dimensional visualization. They're perfect for visualizing relationships between three continuous variables, or for showing the first three principal components of dimensionality-reduced data.

```python
# Create sample 3D data - synthetic dataset with 3 features
np.random.seed(42)
n_points = 200

df_3d = pd.DataFrame({
    'x': np.random.normal(0, 1, n_points),
    'y': np.random.normal(0, 1, n_points),
    'z': np.random.normal(0, 1, n_points),
    'category': np.random.choice(['Group A', 'Group B', 'Group C'], n_points),
    'size': np.abs(np.random.normal(10, 5, n_points))
})

# Basic 3D scatter plot
fig = px.scatter_3d(
    df_3d, 
    x='x', y='y', z='z',
    color='category',
    size='size',
    title='🎯 Basic 3D Scatter Plot with Color and Size',
    labels={'x': 'Feature X', 'y': 'Feature Y', 'z': 'Feature Z'},
    opacity=0.7
)

fig.update_layout(
    scene=dict(
        xaxis_title='Feature X',
        yaxis_title='Feature Y',
        zaxis_title='Feature Z'
    ),
    width=800,
    height=600
)

fig.show()
```

```python
# 3D scatter with Iris dataset - a classic ML dataset
iris = load_iris()
df_iris = pd.DataFrame(
    iris.data, 
    columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
)
df_iris['species'] = [iris.target_names[i] for i in iris.target]
df_iris['species_num'] = iris.target

# Visualize three key features in 3D
fig = px.scatter_3d(
    df_iris,
    x='sepal_length',
    y='petal_length', 
    z='petal_width',
    color='species',
    symbol='species',
    size='sepal_width',
    title='🌸 Iris Dataset: 3D Visualization by Species',
    hover_data=['sepal_width'],
    color_discrete_sequence=px.colors.qualitative.Bold
)

fig.update_layout(
    scene=dict(
        xaxis_title='Sepal Length (cm)',
        yaxis_title='Petal Length (cm)',
        zaxis_title='Petal Width (cm)',
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.2))
    ),
    legend_title_text='Species',
    width=900,
    height=700
)

fig.show()
```

## 3. 3D Line and Surface Plots {#3d-line-surface}

Beyond scatter plots, Plotly excels at 3D line plots (great for trajectories or time series in 3D space) and surface plots (perfect for visualizing mathematical functions or 2D data landscapes).

```python
# 3D Line plot - visualizing a trajectory or parametric curve
t = np.linspace(0, 4*np.pi, 500)
x = np.sin(t) * np.cos(3*t)
y = np.sin(t) * np.sin(3*t)
z = np.cos(t)

fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='lines',
    line=dict(
        color=t,  # Color based on parameter t
        colorscale='Viridis',
        width=4
    ),
    marker=dict(
        size=2,
        color=t,
        colorscale='Viridis',
        showscale=True,
        colorbar=dict(title='Parameter t')
    ),
    name='Parametric Curve'
)])

fig.update_layout(
    title='🌀 3D Line Plot: Parametric Curve with Color Gradient',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y', 
        zaxis_title='Z',
        aspectmode='cube'
    ),
    width=800,
    height=700
)

fig.show()
```

```python
# 3D Surface plot - mathematical function visualization
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Create a complex surface: combination of sine waves
Z = np.sin(np.sqrt(X**2 + Y**2)) * np.cos(X) * np.sin(Y)

fig = go.Figure(data=[go.Surface(
    x=X, y=Y, z=Z,
    colorscale='RdBu',
    contours=dict(
        z=dict(show=True, usecolormap=True, project_z=True),
        x=dict(show=True, usecolormap=True, project_x=True),
        y=dict(show=True, usecolormap=True, project_y=True)
    ),
    colorbar=dict(title='Function Value')
)])

fig.update_layout(
    title='🌊 3D Surface Plot: Mathematical Function with Contours',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.0))
    ),
    width=900,
    height=700
)

fig.show()
```

## 4. Customizing 3D Plots {#customizing-3d}

Plotly provides extensive customization options for 3D plots. Let's explore how to control colors, symbols, hover information, and layout details to create publication-quality visualizations.

```python
# Highly customized 3D scatter with advanced features
np.random.seed(123)
n = 150

# Create multi-dimensional data
df_custom = pd.DataFrame({
    'pc1': np.random.normal(0, 1, n),
    'pc2': np.random.normal(0, 1, n),
    'pc3': np.random.normal(0, 1, n),
    'cluster': np.random.choice(['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'], n),
    'confidence': np.random.uniform(0.5, 1.0, n),
    'sample_id': [f'Sample_{i:03d}' for i in range(n)]
})

# Custom color palette
color_map = {
    'Cluster 1': '#FF6B6B',
    'Cluster 2': '#4ECDC4', 
    'Cluster 3': '#45B7D1',
    'Cluster 4': '#96CEB4'
}

fig = px.scatter_3d(
    df_custom,
    x='pc1',
    y='pc2',
    z='pc3',
    color='cluster',
    color_discrete_map=color_map,
    size='confidence',
    size_max=15,
    hover_name='sample_id',
    hover_data={
        'pc1': ':.2f',
        'pc2': ':.2f',
        'pc3': ':.2f',
        'confidence': ':.1%',
        'cluster': False  # Hide since it's in color
    },
    title='🎨 Customized 3D Plot: PCA Results with Confidence Scores',
    template='plotly_white'
)

# Advanced layout customization
fig.update_layout(
    scene=dict(
        xaxis=dict(
            title='Principal Component 1',
            gridcolor='lightgray',
            backgroundcolor='rgb(245, 245, 245)'
        ),
        yaxis=dict(
            title='Principal Component 2',
            gridcolor='lightgray',
            backgroundcolor='rgb(245, 245, 245)'
        ),
        zaxis=dict(
            title='Principal Component 3',
            gridcolor='lightgray',
            backgroundcolor='rgb(245, 245, 245)'
        ),
        camera=dict(
            eye=dict(x=1.8, y=1.8, z=1.2),
            center=dict(x=0, y=0, z=0)
        ),
        aspectmode='cube'
    ),
    legend=dict(
        title=dict(text='Cluster Assignment'),
        yanchor='top',
        y=0.99,
        xanchor='left',
        x=0.01,
        bgcolor='rgba(255, 255, 255, 0.8)'
    ),
    width=1000,
    height=800,
    margin=dict(l=0, r=0, t=50, b=0)
)

# Update marker symbols
fig.update_traces(
    marker=dict(
        line=dict(width=1, color='DarkSlateGrey'),
        opacity=0.8
    ),
    selector=dict(mode='markers')
)

fig.show()
```

## 5. Interactive Features and Camera Control {#interactive-3d}

One of Plotly's greatest strengths is interactivity. Users can rotate, zoom, pan, and reset the view. As developers, we can set initial camera angles and provide the best default view of our data.

```python
# Demonstrating camera angles and interactive controls
# Create a spiral galaxy-like dataset
np.random.seed(42)
n_arms = 3
n_points_per_arm = 100
points = []

for arm in range(n_arms):
    theta = np.linspace(0, 4*np.pi, n_points_per_arm)
    r = np.linspace(0.5, 3, n_points_per_arm)
    arm_offset = arm * 2 * np.pi / n_arms
    
    x = r * np.cos(theta + arm_offset) + np.random.normal(0, 0.1, n_points_per_arm)
    y = r * np.sin(theta + arm_offset) + np.random.normal(0, 0.1, n_points_per_arm)
    z = np.random.normal(0, 0.3, n_points_per_arm) + r * 0.2
    
    for i in range(n_points_per_arm):
        points.append({
            'x': x[i], 'y': y[i], 'z': z[i],
            'arm': f'Arm {arm + 1}',
            'distance': r[i]
        })

df_spiral = pd.DataFrame(points)

# Create figure with specific camera positioning
fig = px.scatter_3d(
    df_spiral,
    x='x', y='y', z='z',
    color='arm',
    size='distance',
    color_discrete_sequence=px.colors.sequential.Plasma[:3],
    title='🌌 Interactive 3D: Spiral Structure (Try rotating!)',
    hover_data=['distance']
)

# Multiple camera preset buttons
fig.update_layout(
    scene=dict(
        xaxis_title='X Coordinate',
        yaxis_title='Y Coordinate',
        zaxis_title='Z Coordinate',
        camera=dict(eye=dict(x=0, y=0, z=2.5)),  # Top-down view initially
        aspectmode='data'
    ),
    updatemenus=[{
        'type': 'buttons',
        'showactive': False,
        'buttons': [
            {
                'label': '⬆️ Top View',
                'method': 'relayout',
                'args': [{'scene.camera.eye': dict(x=0, y=0, z=2.5)}]
            },
            {
                'label': '➡️ Side View',
                'method': 'relayout',
                'args': [{'scene.camera.eye': dict(x=2.5, y=0, z=0)}]
            },
            {
                'label': '🔄 Default',
                'method': 'relayout',
                'args': [{'scene.camera.eye': dict(x=1.5, y=1.5, z=1.2)}]
            }
        ],
        'x': 0.1,
        'y': 0.95
    }],
    width=900,
    height=750
)

fig.show()
```

## 7. Map Projections and Styling {#map-styling}

Different map projections serve different purposes. Let's explore various projection types and styling options to make your maps both accurate and visually appealing.

```python
# Compare different map projections
projections = ['equirectangular', 'mercator', 'orthographic', 'natural earth', 'kavrayskiy7']

# Sample data - earthquake magnitude simulation
np.random.seed(42)
earthquake_data = []
for _ in range(50):
    lat = np.random.uniform(-60, 70)
    lon = np.random.uniform(-180, 180)
    mag = np.random.exponential(2) + 3
    earthquake_data.append({
        'lat': lat,
        'lon': lon,
        'magnitude': min(mag, 9.0),
        'depth_km': np.random.exponential(50)
    })

df_quakes = pd.DataFrame(earthquake_data)

# Create subplot with different projections
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=3,
    subplot_titles=[f'Projection: {p}' for p in projections],
    specs=[[{'type': 'geo'}, {'type': 'geo'}, {'type': 'geo'}],
           [{'type': 'geo'}, {'type': 'geo'}, {'type': 'geo'}]]
)

for i, proj in enumerate(projections):
    row = i // 3 + 1
    col = i % 3 + 1
    
    scatter = px.scatter_geo(
        df_quakes,
        lat='lat',
        lon='lon',
        size='magnitude',
        color='depth_km',
        color_continuous_scale='Hot',
        size_max=15
    ).data[0]
    
    fig.add_trace(scatter, row=row, col=col)
    
    fig.update_layout({
        f'geo{i+1}': dict(
            projection_type=proj,
            showland=True,
            landcolor='rgb(230, 230, 230)',
            showocean=True,
            oceancolor='rgb(220, 235, 255)',
            showcoastlines=True,
            coastlinecolor='rgb(150, 150, 150)',
            showframe=False
        )
    })

fig.update_layout(
    title_text='🗺️ Map Projection Comparison: Earthquake Data Visualization',
    showlegend=False,
    width=1200,
    height=800,
    coloraxis=dict(colorscale='Hot', colorbar=dict(title='Depth (km)', y=0.5))
)

fig.show()
```

## 8. Animation in 3D and Maps {#animation}

Animation adds a time dimension to your visualizations, allowing you to show how data evolves. This is particularly powerful for showing ML model training progress, time-series predictions, or geographic trends over time.

```python
# Animated 3D scatter - showing clustering convergence over iterations
np.random.seed(42)

# Generate data that moves over time (simulating clustering convergence)
n_points = 50
n_frames = 8
frames_data = []

for frame in range(n_frames):
    # Simulate points converging to 3 centers
    t = frame / (n_frames - 1)  # 0 to 1
    centers = np.array([[2, 2, 2], [-2, -2, 2], [0, 2, -2]])
    
    for i in range(n_points):
        target = centers[i % 3]
        start = np.random.normal(0, 3, 3)
        pos = start * (1 - t) + target * t + np.random.normal(0, 0.3 * (1-t), 3)
        
        frames_data.append({
            'x': pos[0],
            'y': pos[1],
            'z': pos[2],
            'cluster': f'Cluster {i % 3 + 1}',
            'point_id': i,
            'iteration': frame + 1
        })

df_animation = pd.DataFrame(frames_data)

# Create animated 3D plot
fig = px.scatter_3d(
    df_animation,
    x='x', y='y', z='z',
    color='cluster',
    animation_frame='iteration',
    animation_group='point_id',
    size_max=10,
    color_discrete_sequence=px.colors.qualitative.Set1,
    title='🎬 Animated 3D: Simulated Clustering Convergence Over Iterations',
    range_x=[-5, 5],
    range_y=[-5, 5],
    range_z=[-5, 5]
)

fig.update_layout(
    scene=dict(
        xaxis_title='X Position',
        yaxis_title='Y Position',
        zaxis_title='Z Position',
        aspectmode='cube'
    ),
    width=900,
    height=700,
    updatemenus=[{
        'type': 'buttons',
        'showactive': False,
        'buttons': [
            {
                'label': '▶️ Play',
                'method': 'animate',
                'args': [None, {'frame': {'duration': 800, 'redraw': True}, 
                               'fromcurrent': True, 'mode': 'immediate'}]
            },
            {
                'label': '⏸️ Pause',
                'method': 'animate',
                'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 
                                 'mode': 'immediate', 'transition': {'duration': 0}}]
            }
        ],
        'x': 0.1,
        'y': 0.05
    }]
)

fig.show()
```

```python
# Animated choropleth - showing data changes over years
years = list(range(2015, 2024))
countries = ['United States', 'China', 'Germany', 'Brazil', 'India', 'Japan']

# Simulate growth data over years
growth_data = []
for year in years:
    for country in countries:
        base = {'United States': 65, 'China': 45, 'Germany': 55, 
                'Brazil': 35, 'India': 25, 'Japan': 50}[country]
        growth = (year - 2015) * np.random.uniform(2, 5)
        value = base + growth + np.random.normal(0, 3)
        
        growth_data.append({
            'country': country,
            'year': year,
            'renewable_energy_pct': max(0, min(100, value)),
            'continent': {'United States': 'North America', 'China': 'Asia', 'Germany': 'Europe',
                         'Brazil': 'South America', 'India': 'Asia', 'Japan': 'Asia'}[country]
        })

df_energy = pd.DataFrame(growth_data)

fig = px.choropleth(
    df_energy,
    locations='country',
    locationmode='country names',
    color='renewable_energy_pct',
    animation_frame='year',
    color_continuous_scale='Greens',
    range_color=[0, 100],
    title='🌱 Animated Map: Renewable Energy Adoption (%) Over Time',
    labels={'renewable_energy_pct': 'Renewable Energy %'},
    hover_data=['continent']
)

fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='natural earth'
    ),
    coloraxis_colorbar=dict(
        title='Renewable<br>Energy %',
        ticksuffix='%'
    ),
    width=1000,
    height=600,
    sliders=[{
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 16},
            'prefix': 'Year: ',
            'visible': True,
            'xanchor': 'right'
        },
        'transition': {'duration': 300, 'easing': 'cubic-in-out'},
        'pad': {'b': 10, 't': 50},
        'len': 0.9,
        'x': 0.1,
        'y': 0,
        'steps': [
            {
                'args': [[str(year)], {'frame': {'duration': 300, 'redraw': True},
                                      'mode': 'immediate', 'transition': {'duration': 300}}],
                'label': str(year),
                'method': 'animate'
            } for year in years
        ]
    }]
)

fig.show()
```

```python
# Real ML Example 1: 3D PCA Visualization of High-Dimensional Data
from sklearn.datasets import load_wine

# Load wine dataset (13 features)
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

# Standardize and apply PCA to reduce to 3 components
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame for plotting
df_pca = pd.DataFrame(
    X_pca,
    columns=['PC1', 'PC2', 'PC3']
)
df_pca['class'] = [wine.target_names[i] for i in y]
df_pca['class_num'] = y

# Add original feature values to hover data
for i, name in enumerate(feature_names):
    df_pca[name] = X[:, i]

# Create comprehensive 3D PCA plot
fig = px.scatter_3d(
    df_pca,
    x='PC1',
    y='PC2',
    z='PC3',
    color='class',
    symbol='class',
    color_discrete_sequence=px.colors.qualitative.Dark24,
    title='🍷 ML Visualization: Wine Dataset in 3D PCA Space',
    hover_data={
        'class': False,
        'PC1': ':.2f',
        'PC2': ':.2f',
        'PC3': ':.2f',
        'alcohol': ':.1f',
        'color_intensity': ':.1f',
        'proline': ':.0f'
    },
    opacity=0.8
)

# Add explained variance to axis labels
explained_var = pca.explained_variance_ratio_
fig.update_layout(
    scene=dict(
        xaxis_title=f'PC1 ({explained_var[0]:.1%} variance)',
        yaxis_title=f'PC2 ({explained_var[1]:.1%} variance)',
        zaxis_title=f'PC3 ({explained_var[2]:.1%} variance)',
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.0))
    ),
    legend_title_text='Wine Class',
    width=950,
    height=750,
    annotations=[
        dict(
            text=f'Total Variance Explained: {sum(explained_var):.1%}',
            showarrow=False,
            x=0.02,
            y=0.98,
            xref='paper',
            yref='paper',
            font=dict(size=12, color='gray')
        )
    ]
)

fig.show()

print(f"Explained variance by component: {explained_var}")
print(f"Total variance captured: {sum(explained_var):.2%}")
```

```python
# Real ML Example 2: K-Means Clustering in 3D with Decision Boundaries
from sklearn.cluster import KMeans

# Generate synthetic clustering data
X_blobs, y_true = make_blobs(
    n_samples=300,
    centers=4,
    n_features=3,
    cluster_std=0.8,
    random_state=42
)

# Fit K-Means
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
y_pred = kmeans.fit_predict(X_blobs)

# Create DataFrame
df_clusters = pd.DataFrame({
    'x': X_blobs[:, 0],
    'y': X_blobs[:, 1],
    'z': X_blobs[:, 2],
    'cluster': [f'Cluster {i+1}' for i in y_pred],
    'true_label': y_true
})

# Create the plot
fig = px.scatter_3d(
    df_clusters,
    x='x', y='y', z='z',
    color='cluster',
    symbol='cluster',
    color_discrete_sequence=px.colors.qualitative.Bold,
    title='🎯 ML Visualization: K-Means Clustering Results (k=4)',
    opacity=0.7
)

# Add cluster centers
centers = kmeans.cluster_centers_
fig.add_trace(go.Scatter3d(
    x=centers[:, 0],
    y=centers[:, 1],
    z=centers[:, 2],
    mode='markers',
    marker=dict(
        size=12,
        color='black',
        symbol='x',
        line=dict(width=2, color='white')
    ),
    name='Centroids',
    hovertemplate='<b>Centroid</b><br>X: %{x:.2f}<br>Y: %{y:.2f}<br>Z: %{z:.2f}<extra></extra>'
))

fig.update_layout(
    scene=dict(
        xaxis_title='Feature 1',
        yaxis_title='Feature 2',
        zaxis_title='Feature 3',
        aspectmode='cube'
    ),
    legend=dict(
        yanchor='top',
        y=0.99,
        xanchor='left',
        x=0.01
    ),
    width=900,
    height=700
)

fig.show()

print(f"Inertia (within-cluster sum of squares): {kmeans.inertia_:.2f}")
print(f"Number of iterations: {kmeans.n_iter_}")
```

```python
# Real ML Example 3: Geospatial House Price Analysis
np.random.seed(42)

# Simulate real estate data for a city
n_properties = 200

# Generate coordinates around city center (simulating San Francisco area)
center_lat, center_lon = 37.7749, -122.4194
latitudes = center_lat + np.random.normal(0, 0.08, n_properties)
longitudes = center_lon + np.random.normal(0, 0.08, n_properties)

# Features affecting price
distance_to_center = np.sqrt((latitudes - center_lat)**2 + (longitudes - center_lon)**2)
sqft = np.random.lognormal(7, 0.5, n_properties)  # Square footage
bedrooms = np.random.poisson(3, n_properties) + 1
age = np.random.exponential(20, n_properties)  # Years old

# Price calculation (with some noise)
price = (500000 + 
         sqft * 300 + 
         bedrooms * 50000 - 
         age * 2000 - 
         distance_to_center * 2000000 +
         np.random.normal(0, 100000, n_properties))

price = np.maximum(price, 200000)  # Minimum price

df_real_estate = pd.DataFrame({
    'latitude': latitudes,
    'longitude': longitudes,
    'price': price,
    'price_formatted': [f'${p:,.0f}' for p in price],
    'sqft': sqft.astype(int),
    'bedrooms': bedrooms,
    'age': age.astype(int),
    'price_per_sqft': (price / sqft).astype(int),
    'distance_to_downtown': distance_to_center
})

# Create interactive map
fig = px.scatter_mapbox(
    df_real_estate,
    lat='latitude',
    lon='longitude',
    color='price',
    size='sqft',
    color_continuous_scale='RdYlGn',
    range_color=[df_real_estate['price'].quantile(0.1), 
                 df_real_estate['price'].quantile(0.9)],
    hover_name='price_formatted',
    hover_data={
        'bedrooms': True,
        'sqft': ':,',
        'age': ':.0f years',
        'price_per_sqft': ':$,.0f',
        'latitude': False,
        'longitude': False,
        'price': False,
        'distance_to_downtown': ':.2f°'
    },
    zoom=11,
    height=700,
    title='🏠 ML Geospatial Analysis: Real Estate Prices by Location'
)

fig.update_layout(
    mapbox_style='carto-positron',
    margin=dict(r=0, t=50, l=0, b=0),
    coloraxis_colorbar=dict(
        title='Price ($)',
        tickformat='$,.0f'
    )
)

fig.show()

# Show correlation insights
print("📊 Key Insights from Geospatial Analysis:")
print(f"Average price: ${price.mean():,.0f}")
print(f"Price correlation with distance to center: {np.corrcoef(price, distance_to_center)[0,1]:.3f}")
print(f"Price correlation with square footage: {np.corrcoef(price, sqft)[0,1]:.3f}")
```

### Solution 1: Basic 3D Scatter
Create a 3D scatter plot of 100 random points with x, y, z coordinates. Color the points by their distance from the origin.

```python
# Solution 1
import plotly.express as px
import numpy as np
import pandas as pd

np.random.seed(42)
n = 100
df = pd.DataFrame({
    'x': np.random.normal(0, 1, n),
    'y': np.random.normal(0, 1, n),
    'z': np.random.normal(0, 1, n)
})
df['distance'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)

fig = px.scatter_3d(df, x='x', y='y', z='z', color='distance',
                    title='3D Points Colored by Distance from Origin',
                    color_continuous_scale='Viridis')
fig.show()
print("✅ Expected: Interactive 3D scatter with color gradient from center")
```

### Solution 3: Synthetic High-Dimensional Data
Generate a dataset with 5 features, apply PCA to reduce to 3D, and visualize.

```python
# Solution 3
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
n_samples = 200
# Generate 5D data with 3 underlying clusters
X = np.random.randn(n_samples, 5)
true_labels = np.random.choice(['A', 'B', 'C'], n_samples)
for i, label in enumerate(['A', 'B', 'C']):
    mask = true_labels == label
    X[mask] += np.random.randn(5) * 2  # Shift each cluster

# PCA
X_scaled = StandardScaler().fit_transform(X)
X_pca = PCA(n_components=3).fit_transform(X_scaled)

df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
df_pca['class'] = true_labels

fig = px.scatter_3d(df_pca, x='PC1', y='PC2', z='PC3', color='class',
                    title='5D Data Projected to 3D via PCA',
                    symbol='class')
fig.show()
print("✅ Expected: 3 clusters visible in reduced 3D space")
```

### Solution 6: 3D Surface
Create a 3D surface plot of Z = sin(X) * cos(Y).

```python
# Solution 6
import plotly.graph_objects as go

x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='RdBu')])
fig.update_layout(title='Surface: Z = sin(X) × cos(Y)',
                  scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
                  width=800, height=600)
fig.show()
print("✅ Expected: Wavy surface with red/blue color gradient")
```

### Solution 7: Animated 3D
Create an animated 3D scatter plot showing points moving in a circular pattern.

```python
# Solution 7
n_points = 50
n_frames = 10
frames = []

for frame in range(n_frames):
    angle = 2 * np.pi * frame / n_frames
    for i in range(n_points):
        radius = 1 + i / n_points
        theta = angle + i * 0.2
        frames.append({
            'x': radius * np.cos(theta),
            'y': radius * np.sin(theta),
            'z': np.sin(angle) * 0.5,
            'point': i,
            'frame': frame
        })

df_anim = pd.DataFrame(frames)
fig = px.scatter_3d(df_anim, x='x', y='y', z='z', animation_frame='frame',
                    animation_group='point', color='point',
                    color_continuous_scale='Rainbow',
                    title='Animated Circular Motion in 3D',
                    range_x=[-2, 2], range_y=[-2, 2], range_z=[-1, 1])
fig.show()
print("✅ Expected: Animated spiral/circular motion, use play button")
```

### Solution 10: ML Clustering Visualization
Generate 3D clustering data, fit K-Means, and visualize with centers.

```python
# Solution 10
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=3, n_features=3, 
                  cluster_std=0.8, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10).fit(X)
labels = kmeans.labels_

df_clust = pd.DataFrame(X, columns=['x', 'y', 'z'])
df_clust['cluster'] = [f'C{i+1}' for i in labels]

fig = px.scatter_3d(df_clust, x='x', y='y', z='z', color='cluster',
                    title='K-Means Clustering (k=3)')

# Add centers
centers = kmeans.cluster_centers_
fig.add_trace(go.Scatter3d(
    x=centers[:,0], y=centers[:,1], z=centers[:,2],
    mode='markers', marker=dict(size=15, color='black', symbol='x'),
    name='Centers'
))
fig.show()
print("✅ Expected: 3 colored clusters with black X centroids")
```

### Solution 11: Mini Project - Global Temperature Analysis
Create comprehensive temperature visualization with animation and maps.

```python
# Solution 11 - Mini Project
years = list(range(2015, 2024))
countries = ['USA', 'China', 'Brazil', 'Russia', 'Australia', 'India']
base_temps = {'USA': 12, 'China': 14, 'Brazil': 25, 'Russia': -5, 'Australia': 22, 'India': 25}

temp_data = []
for year in years:
    for country in countries:
        warming = (year - 2015) * 0.15  # Simulated warming trend
        temp = base_temps[country] + warming + np.random.normal(0, 1)
        temp_data.append({
            'country': country,
            'year': year,
            'temperature': temp,
            'anomaly': temp - base_temps[country]
        })

df_temp = pd.DataFrame(temp_data)

# Animated choropleth
fig_map = px.choropleth(df_temp, locations='country', locationmode='country names',
                        color='temperature', animation_frame='year',
                        color_continuous_scale='RdBu_r',
                        range_color=[-10, 30],
                        title='Global Temperature Trends (2015-2023)')
fig_map.update_layout(height=600)
fig_map.show()

print("✅ Expected: Animated heat map showing warming trends over years")
print("🎓 Mini Project Complete! You've combined animation + geospatial + time series!")
```
