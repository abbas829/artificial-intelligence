## 1. Introduction to Automated Exploratory Data Analysis ⚡

### What is Automated EDA?
Automated EDA tools analyze datasets programmatically and generate comprehensive reports containing:
- **Descriptive statistics** (mean, median, mode, quantiles)
- **Distribution visualizations** (histograms, KDE plots)
- **Missing value analysis** (patterns, counts, heatmaps)
- **Correlation matrices** (Pearson, Spearman, Cramer's V)
- **Data quality alerts** (duplicates, zeros, constants)
- **Interactive visualizations** (scatter plots, pair plots)

### Popular Automated EDA Tools:
1. **ydata-profiling** (formerly pandas-profiling) - Most comprehensive
2. **Sweetviz** - Great for dataset comparisons
3. **AutoViz** - One-line visualization library
4. **D-Tale** - Interactive web-based exploration
5. **Lux** - Visualization recommendation system

### When to Use Automated EDA:
- ✅ **Initial data assessment** (first look at new dataset)
- ✅ **Data quality validation** (checking for issues)
- ✅ **Stakeholder reporting** (shareable HTML reports)
- ✅ **Pipeline monitoring** (automated data drift detection)
- ✅ **Quick comparisons** (train vs test sets)

### When to Use Manual EDA:
- 🔍 **Deep pattern exploration** (custom visualizations)
- 🔍 **Domain-specific insights** (business logic integration)
- 🔍 **Feature engineering** (creating new variables)
- 🔍 **Outlier investigation** (understanding anomalies)
- 🔍 **Hypothesis testing** (statistical validation)

```python
# 2. VARIABLES SECTION - Detailed statistics for each column
print("\n2️⃣ VARIABLES SECTION (Sample - Age column)")
print("-"*80)

age_stats = df['age'].describe()
print("Descriptive Statistics:")
print(age_stats)

print(f"\nAdditional Metrics:")
print(f"   Distinct values: {df['age'].nunique()}")
print(f"   Missing: {df['age'].isnull().sum()} ({df['age'].isnull().mean()*100:.1f}%)")
print(f"   Infinite values: {np.isinf(df['age']).sum()}")
print(f"   Mean: {df['age'].mean():.2f}")
print(f"   Median: {df['age'].median():.2f}")
print(f"   Mode: {df['age'].mode().values[0] if not df['age'].mode().empty else 'N/A'}")
print(f"   Standard deviation: {df['age'].std():.2f}")
print(f"   Variance: {df['age'].var():.2f}")
print(f"   Skewness: {df['age'].skew():.2f}")
print(f"   Kurtosis: {df['age'].kurtosis():.2f}")
print(f"   Range: {df['age'].max() - df['age'].min():.2f}")
print(f"   Interquartile range (IQR): {df['age'].quantile(0.75) - df['age'].quantile(0.25):.2f}")

# Categorical example
print(f"\n\nCategorical Example - Embarked:")
print(f"   Distinct values: {df['embarked'].nunique()}")
print(f"   Missing: {df['embarked'].isnull().sum()}")
print("   Value counts:")
for idx, (val, count) in enumerate(df['embarked'].value_counts().head().items(), 1):
    print(f"      {idx}. {val}: {count} ({count/len(df)*100:.1f}%)")
```

```python
# 3. INTERACTIONS & CORRELATIONS SECTION
print("\n3️⃣ INTERACTIONS & CORRELATIONS SECTION")
print("-"*80)

# Correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

print("Pearson Correlation Matrix (numeric features):")
print(correlation_matrix.round(3))

# High correlations
print(f"\n🔍 HIGH CORRELATIONS (|r| > 0.5):")
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.5:
            high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_val))

if high_corr_pairs:
    for var1, var2, corr in high_corr_pairs:
        print(f"   • {var1} ↔ {var2}: {corr:.3f}")
else:
    print("   No high correlations found (threshold: 0.5)")

# Correlation with target
if 'survived' in correlation_matrix.columns:
    target_corr = correlation_matrix['survived'].drop('survived').sort_values(key=abs, ascending=False)
    print(f"\n🎯 CORRELATION WITH TARGET (survived):")
    for feature, corr in target_corr.items():
        print(f"   • {feature}: {corr:.3f}")
```

```python
# 4. MISSING VALUES SECTION
print("\n4️⃣ MISSING VALUES SECTION")
print("-"*80)

missing_stats = df.isnull().sum()
missing_stats = missing_stats[missing_stats > 0].sort_values(ascending=False)

if len(missing_stats) > 0:
    print("Missing Values by Column:")
    print(f"{'Column':<15} {'Count':<8} {'Percentage':<10} {'Pattern'}")
    print("-"*50)
    for col, count in missing_stats.items():
        pct = count / len(df) * 100
        # Determine pattern
        missing_pattern = "MCAR" if pct < 5 else "MAR" if pct < 20 else "MNAR"
        print(f"{col:<15} {count:<8} {pct:<10.1f} {missing_pattern}")
    
    print(f"\nTotal missing values: {missing_stats.sum():,}")
    print(f"Total complete rows: {(~df.isnull().any(axis=1)).sum():,} ({(~df.isnull().any(axis=1)).mean()*100:.1f}%)")
else:
    print("✅ No missing values detected in this dataset!")

# Missing value correlation
print(f"\nMissing Value Co-occurrence:")
missing_cols = missing_stats.index.tolist()
if len(missing_cols) > 1:
    missing_corr = df[missing_cols].isnull().corr()
    print(missing_corr.round(3))
else:
    print("   Insufficient columns with missing data for correlation analysis")
```

```python
# 5. ALERTS SECTION (Data Quality Warnings)
print("\n5️⃣ ALERTS SECTION (Data Quality Issues)")
print("-"*80)

alerts = []

# Check for high cardinality
for col in df.select_dtypes(include=['object']).columns:
    if df[col].nunique() > 50:
        alerts.append(f"🔸 High cardinality: '{col}' has {df[col].nunique()} distinct values")

# Check for high missing percentage
for col in df.columns:
    missing_pct = df[col].isnull().mean() * 100
    if missing_pct > 20:
        alerts.append(f"⚠️  High missing: '{col}' has {missing_pct:.1f}% missing values")

# Check for zeros in numeric columns
for col in df.select_dtypes(include=[np.number]).columns:
    zero_pct = (df[col] == 0).mean() * 100
    if zero_pct > 50 and col != 'survived':
        alerts.append(f"⚠️  Many zeros: '{col}' has {zero_pct:.1f}% zeros (zero-inflated)")

# Check for duplicates
if df.duplicated().sum() > 0:
    alerts.append(f"⚠️  Duplicates: {df.duplicated().sum()} duplicate rows found")

# Check for constant values
for col in df.columns:
    if df[col].nunique() == 1:
        alerts.append(f"⚠️  Constant value: '{col}' has only one unique value")

# Check for high skewness
for col in df.select_dtypes(include=[np.number]).columns:
    if abs(df[col].skew()) > 3:
        alerts.append(f"🔸 High skewness: '{col}' has skewness of {df[col].skew():.2f}")

if alerts:
    print(f"Found {len(alerts)} data quality alerts:\n")
    for alert in alerts:
        print(alert)
else:
    print("✅ No major data quality issues detected!")

print("\n💡 TIP: Automated alerts help catch issues early, but always verify manually!")
```

## 5. Analyzing a Tabular Dataset with Pandas Profiling 📈

Let's apply the profiling to a more complex dataset to see advanced features.

```python
# Create a more complex dataset for advanced profiling
# Customer churn dataset simulation
np.random.seed(45)
n_customers = 2000

churn_df = pd.DataFrame({
    'customer_id': range(1000, 3000),
    'tenure_months': np.random.randint(1, 72, n_customers),
    'monthly_charges': np.round(np.random.uniform(20, 120, n_customers), 2),
    'total_charges': np.round(np.random.uniform(100, 8000, n_customers), 2),
    'age': np.random.randint(18, 80, n_customers),
    'contract_type': np.random.choice(['Month-to-month', 'One year', 'Two year'], n_customers, p=[0.5, 0.3, 0.2]),
    'payment_method': np.random.choice(['Credit card', 'Bank transfer', 'Electronic check', 'Mailed check'], n_customers),
    'internet_service': np.random.choice(['DSL', 'Fiber optic', 'No'], n_customers, p=[0.4, 0.4, 0.2]),
    'tech_support': np.random.choice(['Yes', 'No'], n_customers),
    'online_security': np.random.choice(['Yes', 'No'], n_customers),
    'paperless_billing': np.random.choice(['Yes', 'No'], n_customers),
    'num_support_tickets': np.random.poisson(2, n_customers),
    'avg_call_duration': np.round(np.random.exponential(10, n_customers), 2),
    'churn': np.random.choice(['Yes', 'No'], n_customers, p=[0.27, 0.73])
})

# Introduce some data quality issues for the profiler to catch
# Missing values
missing_idx = np.random.choice(churn_df.index, size=100, replace=False)
churn_df.loc[missing_idx[:50], 'age'] = np.nan
churn_df.loc[missing_idx[50:80], 'payment_method'] = np.nan
churn_df.loc[missing_idx[80:], 'monthly_charges'] = np.nan

# Outliers
outlier_idx = np.random.choice(churn_df.index, size=20, replace=False)
churn_df.loc[outlier_idx, 'monthly_charges'] = churn_df.loc[outlier_idx, 'monthly_charges'] * 5

# Zeros (potential data quality issue)
zero_idx = np.random.choice(churn_df.index, size=50, replace=False)
churn_df.loc[zero_idx, 'avg_call_duration'] = 0

print("📊 CUSTOMER CHURN DATASET CREATED")
print("="*60)
print(f"Shape: {churn_df.shape[0]:,} rows × {churn_df.shape[1]} columns")
print(f"\nFeatures include:")
print("   • Demographics: age, customer_id")
print("   • Financial: monthly_charges, total_charges, tenure_months")
print("   • Services: contract_type, internet_service, tech_support")
print("   • Behavior: num_support_tickets, avg_call_duration")
print("   • Target: churn (Yes/No)")
print(f"\nIntentionally introduced data quality issues for profiling:")
print(f"   • Missing values: {churn_df.isnull().sum().sum()} cells")
print(f"   • Outliers: 20 extreme monthly charges")
print(f"   • Zeros: 50 zero call durations")
```

```python
# Generate comprehensive profile for churn dataset
churn_profile = ProfileReport(
    churn_df,
    title="Customer Churn - Automated EDA Report",
    dataset={
        "description": "Telecom customer churn prediction dataset with data quality issues",
        "author": "Python & AI Learning Path",
        "source": "Synthetic dataset for educational purposes"
    },
    explorative=True,
    correlations={
        "pearson": {"calculate": True},
        "spearman": {"calculate": True},
        "kendall": {"calculate": False},
        "phi_k": {"calculate": True},
        "cramers": {"calculate": True}
    },
    missing_diagrams={
        "heatmap": True,
        "dendrogram": True,
        "matrix": True
    }
)

print("✅ Churn dataset profile generated!")
print("📊 This report includes:")
print("   • Multiple correlation methods (Pearson, Spearman, Phik, Cramers)")
print("   • Missing value visualizations (heatmap, dendrogram)")
print("   • Data quality alerts for introduced issues")
print("\n💡 Note: Phik and Cramers V are especially useful for categorical correlations")
```

```python
# Display the churn report
try:
    churn_profile.to_widgets()
except Exception as e:
    print("⚠️ Unable to render widgets, falling back to notebook iframe:")
    print(e)
    churn_profile.to_notebook_iframe()

print("🔍 CHURN DATASET REPORT LOADED ABOVE")
print("Look for:")
print("   • Alerts about missing values in age, payment_method, monthly_charges")
print("   • High skewness warnings for charges")
print("   • Correlation between churn and contract_type/tenure")
print("   • Missing value patterns in the matrix view")

print("🔍 CHURN DATASET REPORT LOADED ABOVE")
print("Look for:")
print("   • Alerts about missing values in age, payment_method, monthly_charges")
print("   • High skewness warnings for charges")
print("   • Correlation between churn and contract_type/tenure")
print("   • Missing value patterns in the matrix view")
```

## 6. Interpreting Key Insights from the Report 💡

Let's extract and interpret the most valuable insights from automated reports.

```python
# Extract key insights programmatically from the profile object
print("🔍 KEY INSIGHTS EXTRACTION")
print("="*80)

# Access report description (if available in the object)
# Note: Some properties may vary by ydata-profiling version

# 1. Dataset overview insights
print("\n1️⃣ DATASET OVERVIEW INSIGHTS")
print("-"*80)
print(f"Total observations: {len(churn_df):,}")
print(f"Total variables: {churn_df.shape[1]}")
print(f"Memory footprint: {churn_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"Duplicate rows: {churn_df.duplicated().sum()}")

# 2. Variable-level insights
print("\n2️⃣ VARIABLE-LEVEL INSIGHTS")
print("-"*80)

# Numeric variables summary
numeric_cols = churn_df.select_dtypes(include=[np.number]).columns
print(f"\nNumeric Variables Summary ({len(numeric_cols)} total):")
for col in numeric_cols[:5]:  # Show first 5
    stats = churn_df[col].describe()
    print(f"   {col}:")
    print(f"      Range: [{stats['min']:.2f}, {stats['max']:.2f}]")
    print(f"      Mean ± Std: {stats['mean']:.2f} ± {churn_df[col].std():.2f}")
    print(f"      Missing: {churn_df[col].isnull().sum()} ({churn_df[col].isnull().mean()*100:.1f}%)")
    if abs(churn_df[col].skew()) > 1:
        print(f"      ⚠️  Highly skewed (skewness: {churn_df[col].skew():.2f})")

# Categorical variables summary
cat_cols = churn_df.select_dtypes(include=['object']).columns
print(f"\nCategorical Variables Summary ({len(cat_cols)} total):")
for col in cat_cols[:5]:
    print(f"   {col}:")
    print(f"      Unique values: {churn_df[col].nunique()}")
    print(f"      Most common: {churn_df[col].value_counts().index[0]} ({churn_df[col].value_counts().iloc[0]/len(churn_df)*100:.1f}%)")
    if churn_df[col].isnull().sum() > 0:
        print(f"      ⚠️  Missing: {churn_df[col].isnull().sum()}")
```

```python
# 3. Target variable insights
print("\n3️⃣ TARGET VARIABLE ANALYSIS (Churn)")
print("-"*80)

target_dist = churn_df['churn'].value_counts()
target_pct = churn_df['churn'].value_counts(normalize=True) * 100

print("Distribution:")
for val, count in target_dist.items():
    print(f"   {val}: {count:,} ({target_pct[val]:.1f}%)")

print(f"\nClass balance ratio: {target_dist.max()/target_dist.min():.2f}:1")
print(f"Interpretation: {'Balanced' if target_dist.max()/target_dist.min() < 2 else 'Moderately imbalanced' if target_dist.max()/target_dist.min() < 5 else 'Highly imbalanced'}")

# 4. Correlation insights
print("\n4️⃣ CORRELATION INSIGHTS")
print("-"*80)

# Encode target for correlation
churn_encoded = churn_df.copy()
churn_encoded['churn_encoded'] = (churn_encoded['churn'] == 'Yes').astype(int)

# Numeric correlations with target
numeric_corr = churn_encoded[numeric_cols.tolist() + ['churn_encoded']].corr()['churn_encoded'].drop('churn_encoded')
numeric_corr = numeric_corr.sort_values(key=abs, ascending=False)

print("Top correlations with churn (numeric features):")
for feature, corr in numeric_corr.head(5).items():
    direction = "positively" if corr > 0 else "negatively"
    strength = "strongly" if abs(corr) > 0.5 else "moderately" if abs(corr) > 0.3 else "weakly"
    print(f"   • {feature}: {corr:.3f} ({strength} {direction} correlated)")

# Categorical associations (using groupby)
print(f"\nCategorical associations with churn:")
for col in ['contract_type', 'internet_service', 'tech_support']:
    if col in churn_df.columns:
        crosstab = pd.crosstab(churn_df[col], churn_df['churn'], normalize='index')
        churn_rate = crosstab['Yes']
        highest = churn_rate.idxmax()
        lowest = churn_rate.idxmin()
        print(f"   • {col}:")
        print(f"      Highest churn: {highest} ({churn_rate[highest]:.1%})")
        print(f"      Lowest churn: {lowest} ({churn_rate[lowest]:.1%})")
```

```python
# 5. Data quality insights
print("\n5️⃣ DATA QUALITY INSIGHTS")
print("-"*80)

quality_issues = []

# Missing values
missing_by_col = churn_df.isnull().sum()
missing_cols = missing_by_col[missing_by_col > 0]
if len(missing_cols) > 0:
    quality_issues.append(f"Missing values in {len(missing_cols)} columns")
    print("Missing Value Summary:")
    for col, count in missing_cols.items():
        print(f"   • {col}: {count} ({count/len(churn_df)*100:.1f}%)")

# Outliers (using IQR method)
print(f"\nOutlier Detection (IQR method):")
for col in ['monthly_charges', 'total_charges', 'tenure_months']:
    if col in churn_df.columns:
        Q1 = churn_df[col].quantile(0.25)
        Q3 = churn_df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = churn_df[(churn_df[col] < Q1 - 1.5*IQR) | (churn_df[col] > Q3 + 1.5*IQR)]
        if len(outliers) > 0:
            print(f"   • {col}: {len(outliers)} outliers ({len(outliers)/len(churn_df)*100:.1f}%)")
            quality_issues.append(f"Outliers in {col}")

# Zero values
print(f"\nZero Value Analysis:")
for col in ['avg_call_duration', 'num_support_tickets']:
    if col in churn_df.columns:
        zero_count = (churn_df[col] == 0).sum()
        if zero_count > 0:
            print(f"   • {col}: {zero_count} zeros ({zero_count/len(churn_df)*100:.1f}%)")
            if zero_count / len(churn_df) > 0.3:
                quality_issues.append(f"Many zeros in {col} (zero-inflated)")

# Cardinality warnings
print(f"\nHigh Cardinality Check:")
for col in churn_df.select_dtypes(include=['object']).columns:
    if churn_df[col].nunique() > 50:
        print(f"   ⚠️  {col}: {churn_df[col].nunique()} unique values")
        quality_issues.append(f"High cardinality in {col}")

print(f"\n✅ Total quality issues identified: {len(quality_issues)}")
print("💡 Recommendation: Address these issues before modeling")
```

## 7. Limitations of Automated EDA Tools ⚠️

Understanding when automated tools fall short is crucial for responsible data science.

## 9. Creating a Custom EDA Workflow 🛠️

Let's build a reusable function that combines automated and manual EDA.

```python
def comprehensive_eda(df, target_col=None, dataset_name="Dataset", 
                     generate_html=True, html_filename=None):
    """
    Comprehensive EDA combining automated profiling and manual analysis.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The dataset to analyze
    target_col : str, optional
        Name of the target variable
    dataset_name : str
        Name of the dataset for reporting
    generate_html : bool
        Whether to save HTML report
    html_filename : str, optional
        Filename for HTML report
    
    Returns:
    --------
    dict
        Dictionary containing key insights and quality metrics
    """
    
    print(f"🔍 COMPREHENSIVE EDA: {dataset_name}")
    print("="*80)
    
    results = {
        'dataset_name': dataset_name,
        'shape': df.shape,
        'timestamp': pd.Timestamp.now().isoformat()
    }
    
    # PHASE 1: Automated Profiling
    print("\n⚡ PHASE 1: Generating Automated Profile...")
    profile = ProfileReport(
        df, 
        title=f"{dataset_name} - Automated EDA",
        explorative=True,
        minimal=False
    )
    
    if generate_html:
        if html_filename is None:
            html_filename = f"{dataset_name.lower().replace(' ', '_')}_eda_report.html"
        profile.to_file(html_filename)
        print(f"   ✓ HTML report saved: {html_filename}")
    
    # PHASE 2: Manual Deep Dive
    print("\n🔎 PHASE 2: Manual Analysis...")
    
    # 2.1 Data Quality Assessment
    print("\n   2.1 Data Quality Metrics:")
    quality_metrics = {
        'total_missing': df.isnull().sum().sum(),
        'missing_pct': df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100,
        'duplicate_rows': df.duplicated().sum(),
        'memory_mb': df.memory_usage(deep=True).sum() / 1024**2
    }
    results['quality_metrics'] = quality_metrics
    
    print(f"      • Missing values: {quality_metrics['total_missing']:,} ({quality_metrics['missing_pct']:.2f}%)")
    print(f"      • Duplicate rows: {quality_metrics['duplicate_rows']:,}")
    print(f"      • Memory usage: {quality_metrics['memory_mb']:.2f} MB")
    
    # 2.2 Target Analysis (if provided)
    if target_col and target_col in df.columns:
        print(f"\n   2.2 Target Variable Analysis ({target_col}):")
        target_series = df[target_col]
        
        if target_series.dtype == 'object' or target_series.nunique() < 10:
            # Categorical target
            value_counts = target_series.value_counts()
            results['target_distribution'] = value_counts.to_dict()
            print(f"      • Type: Categorical ({target_series.nunique()} classes)")
            for val, count in value_counts.items():
                print(f"      • {val}: {count:,} ({count/len(df)*100:.1f}%)")
        else:
            # Numeric target
            print(f"      • Type: Numeric")
            print(f"      • Mean: {target_series.mean():.2f}")
            print(f"      • Range: [{target_series.min():.2f}, {target_series.max():.2f}]")
    
    # 2.3 Feature Importance (correlation with target)
    if target_col and target_col in df.columns:
        print(f"\n   2.3 Top Feature Associations:")
        
        # Numeric correlations
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        if target_col in numeric_cols:
            correlations = df[numeric_cols].corr()[target_col].drop(target_col).sort_values(key=abs, ascending=False)
            results['top_correlations'] = correlations.head(5).to_dict()
            print("      Top numeric correlations:")
            for feat, corr in correlations.head(3).items():
                print(f"      • {feat}: {corr:.3f}")
    
    # 2.4 Distribution Analysis
    print(f"\n   2.4 Distribution Summary:")
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    high_skew = []
    for col in numeric_cols:
        if col != target_col:
            skewness = df[col].skew()
            if abs(skewness) > 2:
                high_skew.append((col, skewness))
    
    if high_skew:
        print(f"      • Highly skewed features ({len(high_skew)}):")
        for col, skew in high_skew[:3]:
            print(f"        - {col}: {skew:.2f}")
    
    # 2.5 Recommendations
    print(f"\n   2.5 Recommendations:")
    recommendations = []
    
    if quality_metrics['missing_pct'] > 5:
        recommendations.append("Address missing values before modeling")
    if quality_metrics['duplicate_rows'] > 0:
        recommendations.append("Remove or investigate duplicate rows")
    if len(high_skew) > 0:
        recommendations.append("Consider log transformation for skewed features")
    
    if not recommendations:
        recommendations.append("Data quality is good - proceed with modeling")
    
    results['recommendations'] = recommendations
    for i, rec in enumerate(recommendations, 1):
        print(f"      {i}. {rec}")
    
    print("\n" + "="*80)
    print("✅ EDA Complete!")
    
    return results
```

### Exercise 8 Solution: Reusable Pipeline

```python
# Solution 8: Stakeholder-ready EDA pipeline
def generate_stakeholder_report(df, dataset_name, target_col=None, output_file=None):
    """
    Generate a stakeholder-friendly EDA summary report.
    
    Returns a formatted text report and saves to file if specified.
    """
    lines = []
    lines.append("="*80)
    lines.append(f"EXECUTIVE SUMMARY: {dataset_name}")
    lines.append("="*80)
    lines.append(f"Generated: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M')}")
    lines.append("")
    
    # Data Quality Score
    missing_pct = df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100
    duplicate_pct = df.duplicated().sum() / len(df) * 100
    quality_score = max(0, 100 - missing_pct - duplicate_pct)
    
    lines.append(f"📊 DATA QUALITY SCORE: {quality_score:.1f}/100")
    lines.append(f"   • Missing data: {missing_pct:.2f}%")
    lines.append(f"   • Duplicate rows: {duplicate_pct:.2f}%")
    lines.append("")
    
    # Dataset Overview
    lines.append("📋 DATASET OVERVIEW")
    lines.append(f"   • Total records: {df.shape[0]:,}")
    lines.append(f"   • Total features: {df.shape[1]}")
    lines.append(f"   • Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
    lines.append("")
    
    # Key Statistics
    lines.append("📈 KEY STATISTICS")
    numeric_cols = df.select_dtypes(include=[np.number]).columns[:5]  # Top 5
    for col in numeric_cols:
        lines.append(f"   • {col}:")
        lines.append(f"     - Average: {df[col].mean():.2f}")
        lines.append(f"     - Range: [{df[col].min():.2f}, {df[col].max():.2f}]")
    lines.append("")
    
    # Target Analysis
    if target_col and target_col in df.columns:
        lines.append("🎯 TARGET VARIABLE ANALYSIS")
        target_series = df[target_col]
        if target_series.dtype == 'object' or target_series.nunique() < 10:
            lines.append(f"   • Target: {target_col}")
            for val, count in target_series.value_counts().items():
                lines.append(f"     - {val}: {count:,} ({count/len(df)*100:.1f}%)")
        lines.append("")
    
    # Recommendations
    lines.append("💡 RECOMMENDATIONS")
    if missing_pct > 5:
        lines.append("   1. Address missing values before modeling")
    if duplicate_pct > 0:
        lines.append("   2. Investigate and remove duplicate records")
    lines.append("   3. Review automated HTML report for detailed visualizations")
    lines.append("   4. Validate data quality issues with domain experts")
    lines.append("")
    
    lines.append("="*80)
    lines.append("For detailed analysis, see accompanying HTML report.")
    lines.append("="*80)
    
    report_text = "\n".join(lines)
    
    if output_file:
        with open(output_file, 'w') as f:
            f.write(report_text)
        print(f"Report saved to: {output_file}")
    
    return report_text

# Test the function
report = generate_stakeholder_report(
    churn_df, 
    "Customer Churn Analysis",
    target_col='churn',
    output_file="churn_stakeholder_summary.txt"
)

print(report)
```

### Exercise 9 Solution: Train vs Test Comparison

```python
# Solution 9: Train vs Test comparison using automated EDA
from sklearn.model_selection import train_test_split

# Split churn data into train/test
train_df, test_df = train_test_split(churn_df, test_size=0.3, random_state=42, stratify=churn_df['churn'])

print("⚖️ TRAIN vs TEST COMPARISON")
print("="*80)
print(f"Train set: {len(train_df):,} samples ({len(train_df)/len(churn_df)*100:.1f}%)")
print(f"Test set: {len(test_df):,} samples ({len(test_df)/len(churn_df)*100:.1f}%)")

# Generate profiles for both
train_profile = ProfileReport(train_df, title="Train Set", minimal=True)
test_profile = ProfileReport(test_df, title="Test Set", minimal=True)

# Compare key statistics
comparison_metrics = ['monthly_charges', 'tenure_months', 'age']

print(f"\n📊 DISTRIBUTION COMPARISON:")
print(f"{'Metric':<20} {'Train Mean':<15} {'Test Mean':<15} {'Difference':<15}")
print("-"*65)

for metric in comparison_metrics:
    if metric in churn_df.columns:
        train_mean = train_df[metric].mean()
        test_mean = test_df[metric].mean()
        diff_pct = abs(train_mean - test_mean) / train_mean * 100
        print(f"{metric:<20} {train_mean:<15.2f} {test_mean:<15.2f} {diff_pct:<15.2f}%")

# Check for data leakage indicators
print(f"\n🔍 DATA LEAKAGE CHECK:")
print(f"   • Duplicate rows across sets: {len(set(train_df.index) & set(test_df.index))}")
print(f"   • Target distribution match: {abs(train_df['churn'].value_counts(normalize=True)['Yes'] - test_df['churn'].value_counts(normalize=True)['Yes']):.3f} diff")

if diff_pct < 5:
    print(f"\n✅ Train/Test split looks good - distributions are similar")
else:
    print(f"\n⚠️  Warning: Significant distribution shift detected!")

# Save comparison reports
train_profile.to_file("train_set_profile.html")
test_profile.to_file("test_set_profile.html")
print(f"\n💾 Individual reports saved for detailed comparison")
```
