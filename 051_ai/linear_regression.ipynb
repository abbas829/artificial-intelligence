## 1. Introduction to Linear Regression and Its Importance

Linear Regression models the relationship between a dependent variable (target) and one or more independent variables (features) by fitting a linear equation.

**Simple Linear Regression:** $y = \beta_0 + \beta_1 x + \epsilon$

**Why it matters:**
- ✅ Foundation for neural networks and deep learning
- ✅ Interpretable and explainable
- ✅ Fast training and prediction
- ✅ Baseline for complex models
- ✅ Used in finance, healthcare, real estate, and more


````python
# 📦 Import essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(51)

print("✅ Libraries imported successfully!")
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
````

## 2. Mathematical Foundation

### 🧮 The Hypothesis Function
For simple linear regression: $h_\theta(x) = \theta_0 + \theta_1 x$

For multiple features: $h_\theta(x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_n x_n = \theta^T x$

### 📉 Cost Function (Mean Squared Error)
$$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$$

### 🔄 Gradient Descent
Repeat until convergence:
$$\theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta)$$

Where:
- $m$ = number of training examples
- $\alpha$ = learning rate
- $\theta_j$ = parameters to optimize


````python
# 🎨 Create synthetic data for demonstration
m = 100  # number of samples
X = np.linspace(0, 10, m).reshape(-1, 1)
true_theta_0 = 2
true_theta_1 = 3
y = true_theta_0 + true_theta_1 * X.flatten() + np.random.normal(0, 2, m)

# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.6, color='blue', label='Training Data')
plt.plot(X, true_theta_0 + true_theta_1 * X.flatten(), 'r-', linewidth=2, label='True Relationship')
plt.xlabel('Feature (X)', fontsize=12)
plt.ylabel('Target (y)', fontsize=12)
plt.title('📊 Synthetic Dataset for Linear Regression', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Dataset shape: X={X.shape}, y={y.shape}")
print(f"True parameters: θ₀={true_theta_0}, θ₁={true_theta_1}")
````

## 3. Implementing Linear Regression from Scratch

Let's build it piece by piece:
1. **Hypothesis function** - makes predictions
2. **Cost function** - measures error
3. **Gradient computation** - finds direction of improvement
4. **Training loop** - iteratively optimizes parameters


````python
class LinearRegressionScratch:
    """
    Linear Regression implemented from scratch using Gradient Descent.
    
    Parameters:
    -----------
    learning_rate : float, default=0.01
        The learning rate for gradient descent
    n_iterations : int, default=1000
        Number of gradient descent iterations
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta = None
        self.cost_history = []
        
    def hypothesis(self, X):
        """Compute h_θ(x) = X @ θ"""
        return X @ self.theta
    
    def compute_cost(self, X, y):
        """Compute Mean Squared Error cost function"""
        m = len(y)
        predictions = self.hypothesis(X)
        cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
        return cost
    
    def compute_gradient(self, X, y):
        """Compute gradients for gradient descent"""
        m = len(y)
        predictions = self.hypothesis(X)
        gradient = (1 / m) * X.T @ (predictions - y)
        return gradient
    
    def fit(self, X, y, verbose=False):
        """
        Train the model using gradient descent.
        
        Parameters:
        -----------
        X : array-like, shape (m, n_features)
            Training features
        y : array-like, shape (m,)
            Training targets
        """
        # Add bias term (column of ones)
        m, n = X.shape
        X_b = np.c_[np.ones((m, 1)), X]  # Add x₀ = 1
        
        # Initialize parameters
        self.theta = np.random.randn(n + 1) * 0.01
        
        # Training loop
        for i in range(self.n_iterations):
            # Compute cost
            cost = self.compute_cost(X_b, y)
            self.cost_history.append(cost)
            
            # Compute gradients
            gradients = self.compute_gradient(X_b, y)
            
            # Update parameters
            self.theta -= self.learning_rate * gradients
            
            if verbose and i % 100 == 0:
                print(f"Iteration {i}: Cost = {cost:.6f}")
        
        return self
    
    def predict(self, X):
        """Make predictions"""
        m = X.shape[0]
        X_b = np.c_[np.ones((m, 1)), X]
        return self.hypothesis(X_b)
    
    def score(self, X, y):
        """Calculate R² score"""
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - (ss_res / ss_tot)

print("✅ LinearRegressionScratch class defined!")
````

````python
# 🚀 Train our from-scratch model
model_scratch = LinearRegressionScratch(learning_rate=0.01, n_iterations=1000)
model_scratch.fit(X, y, verbose=True)

print(f"\n🎯 Learned parameters:")
print(f"θ₀ (bias) = {model_scratch.theta[0]:.4f} (true: {true_theta_0})")
print(f"θ₁ (slope) = {model_scratch.theta[1]:.4f} (true: {true_theta_1})")
print(f"R² Score: {model_scratch.score(X, y):.4f}")
````

````python
# 📉 Visualize cost function convergence
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Cost vs Iterations
axes[0].plot(model_scratch.cost_history, linewidth=2, color='purple')
axes[0].set_xlabel('Iteration', fontsize=12)
axes[0].set_ylabel('Cost (MSE)', fontsize=12)
axes[0].set_title('📉 Cost Function Convergence', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Plot 2: Predictions vs Actual
y_pred_scratch = model_scratch.predict(X)
axes[1].scatter(X, y, alpha=0.6, color='blue', label='Actual Data')
axes[1].plot(X, y_pred_scratch, 'r-', linewidth=3, label='Predicted Line')
axes[1].set_xlabel('Feature (X)', fontsize=12)
axes[1].set_ylabel('Target (y)', fontsize=12)
axes[1].set_title('🎯 Predictions vs Actual', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 4. Evaluating the Model

### 📊 Key Metrics:

**MAE (Mean Absolute Error):** $\frac{1}{m} \sum_{i=1}^{m} |y_i - \hat{y}_i|$
- Average magnitude of errors
- Robust to outliers

**MSE (Mean Squared Error):** $\frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)^2$
- Penalizes large errors
- Differentiable (good for optimization)

**RMSE (Root Mean Squared Error):** $\sqrt{MSE}$
- Same units as target
- Interpretable

**R² Score:** $1 - \frac{SS_{res}}{SS_{tot}}$
- Proportion of variance explained
- 1.0 is perfect prediction


````python
def evaluate_model(y_true, y_pred, model_name="Model"):
    """Calculate and display all regression metrics"""
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    print(f"\n📊 {model_name} Evaluation Metrics:")
    print("=" * 40)
    print(f"MAE  (Mean Absolute Error):    {mae:.4f}")
    print(f"MSE  (Mean Squared Error):     {mse:.4f}")
    print(f"RMSE (Root Mean Squared Error): {rmse:.4f}")
    print(f"R²   (Coefficient of Determination): {r2:.4f}")
    print("=" * 40)
    
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2}

# Evaluate from-scratch model
metrics_scratch = evaluate_model(y, y_pred_scratch, "From-Scratch Linear Regression")
````

````python
# 📊 Residual Analysis
residuals = y - y_pred_scratch

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Residual plot
axes[0].scatter(y_pred_scratch, residuals, alpha=0.6, color='green')
axes[0].axhline(y=0, color='r', linestyle='--', linewidth=2)
axes[0].set_xlabel('Predicted Values', fontsize=12)
axes[0].set_ylabel('Residuals', fontsize=12)
axes[0].set_title('📉 Residual Plot', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Histogram of residuals
axes[1].hist(residuals, bins=20, alpha=0.7, color='orange', edgecolor='black')
axes[1].set_xlabel('Residual Value', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_title('📊 Distribution of Residuals', fontsize=14, fontweight='bold')
axes[1].axvline(x=0, color='r', linestyle='--', linewidth=2)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Mean of residuals: {np.mean(residuals):.6f} (should be ~0)")
print(f"Std of residuals: {np.std(residuals):.4f}")
````

## 7. Feature Scaling and Its Impact

Feature scaling is crucial for gradient descent convergence speed.

**Standardization:** $x' = \frac{x - \mu}{\sigma}$

**Benefits:**
- Faster convergence
- Prevents features with large scales from dominating
- Helps with regularization


````python
# 🔄 Demonstrate feature scaling impact
X_large_scale = X * 1000  # Artificially scale up features

# Without scaling
model_no_scale = LinearRegressionScratch(learning_rate=0.000001, n_iterations=1000)
model_no_scale.fit(X_large_scale, y)

# With scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_large_scale)

model_with_scale = LinearRegressionScratch(learning_rate=0.01, n_iterations=1000)
model_with_scale.fit(X_scaled, y)

# Visualize convergence
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].plot(model_no_scale.cost_history, linewidth=2, color='red')
axes[0].set_title('❌ Without Scaling (LR=0.000001)\nSlow convergence', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Iteration')
axes[0].set_ylabel('Cost')
axes[0].grid(True, alpha=0.3)

axes[1].plot(model_with_scale.cost_history, linewidth=2, color='green')
axes[1].set_title('✅ With StandardScaler (LR=0.01)\nFast convergence', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Iteration')
axes[1].set_ylabel('Cost')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Without scaling - Final cost: {model_no_scale.cost_history[-1]:.4f}")
print(f"With scaling - Final cost: {model_with_scale.cost_history[-1]:.4f}")
print("\n💡 Scaling allows 10,000x larger learning rate and faster convergence!")
````

## 8. Handling Multiple Features (Multivariate Linear Regression)

Real-world problems have multiple features. Let's extend our implementation.


````python
# 🎯 Create multivariate synthetic data
np.random.seed(51)
m = 200
n_features = 3

# Features
X_multi = np.random.randn(m, n_features)
X_multi[:, 0] = X_multi[:, 0] * 2 + 5  # House size (1000s sq ft)
X_multi[:, 1] = X_multi[:, 1] * 1 + 3  # Number of bedrooms
X_multi[:, 2] = X_multi[:, 2] * 5 + 10  # Distance to city center (miles)

# True parameters",
true_theta_multi = np.array([50, 20, -5, 10])  # [bias, size_coef, bedroom_coef, distance_coef]

# Target: House price (in $1000s)
y_multi = (true_theta_multi[0] + 
           true_theta_multi[1] * X_multi[:, 0] + 
           true_theta_multi[2] * X_multi[:, 1] + 
           true_theta_multi[3] * X_multi[:, 2] + 
           np.random.normal(0, 10, m))

print("✅ Multivariate dataset created!")
print(f"Features shape: {X_multi.shape}")
print(f"Target shape: {y_multi.shape}")

# Create DataFrame for visualization
feature_names = ['House Size (1000s sq ft)', 'Bedrooms', 'Distance to City (miles)']
df_multi = pd.DataFrame(X_multi, columns=feature_names)
df_multi['Price ($1000s)'] = y_multi
print("\n📊 First 5 rows:")
print(df_multi.head())
````

````python
# 🚀 Train multivariate models

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_multi, y_multi, test_size=0.2, random_state=51
)

# Scale features
scaler_multi = StandardScaler()
X_train_scaled = scaler_multi.fit_transform(X_train)
X_test_scaled = scaler_multi.transform(X_test)

# From scratch model
model_multi_scratch = LinearRegressionScratch(learning_rate=0.1, n_iterations=2000)
model_multi_scratch.fit(X_train_scaled, y_train)

# Scikit-learn model
model_multi_sklearn = LinearRegression()
model_multi_sklearn.fit(X_train_scaled, y_train)

print("✅ Multivariate models trained!")
print(f"\nFrom Scratch parameters:")
print(f"Bias: {model_multi_scratch.theta[0]:.4f}")
for i, name in enumerate(feature_names):
    print(f"{name}: {model_multi_scratch.theta[i+1]:.4f}")

print(f"\nScikit-Learn parameters:")
print(f"Intercept: {model_multi_sklearn.intercept_:.4f}")
for i, (name, coef) in enumerate(zip(feature_names, model_multi_sklearn.coef_)):
    print(f"{name}: {coef:.4f}")
````

````python
# 📊 Evaluate multivariate models
y_pred_train_scratch = model_multi_scratch.predict(X_train_scaled)
y_pred_test_scratch = model_multi_scratch.predict(X_test_scaled)
y_pred_train_sklearn = model_multi_sklearn.predict(X_train_scaled)
y_pred_test_sklearn = model_multi_sklearn.predict(X_test_scaled)

print("📈 Training Set Performance:")
print("-" * 50)
metrics_train_scratch = evaluate_model(y_train, y_pred_train_scratch, "From Scratch (Train)")
metrics_train_sklearn = evaluate_model(y_train, y_pred_train_sklearn, "Scikit-Learn (Train)")

print("\n📈 Test Set Performance:")
print("-" * 50)
metrics_test_scratch = evaluate_model(y_test, y_pred_test_scratch, "From Scratch (Test)")
metrics_test_sklearn = evaluate_model(y_test, y_pred_test_sklearn, "Scikit-Learn (Test)")
````

````python
# 📉 Feature importance visualization
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Feature importance comparison
x_pos = np.arange(len(feature_names))
width = 0.35

axes[0].bar(x_pos - width/2, model_multi_scratch.theta[1:], width, 
           label='From Scratch', alpha=0.8, color='blue')
axes[0].bar(x_pos + width/2, model_multi_sklearn.coef_, width,
           label='Scikit-Learn', alpha=0.8, color='red')
axes[0].set_xlabel('Features', fontsize=12)
axes[0].set_ylabel('Coefficient Value', fontsize=12)
axes[0].set_title('🔍 Feature Importance Comparison', fontsize=14, fontweight='bold')
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels([name.split('(')[0].strip() for name in feature_names], rotation=15)
axes[0].legend()
axes[0].grid(True, alpha=0.3, axis='y')

# Actual vs Predicted
axes[1].scatter(y_test, y_pred_test_sklearn, alpha=0.6, color='green', s=60)
min_val = min(y_test.min(), y_pred_test_sklearn.min())
max_val = max(y_test.max(), y_pred_test_sklearn.max())
axes[1].plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
axes[1].set_xlabel('Actual Price ($1000s)', fontsize=12)
axes[1].set_ylabel('Predicted Price ($1000s)', fontsize=12)
axes[1].set_title('🎯 Actual vs Predicted (Test Set)', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 9. Applying to House Prices Dataset

Now let's apply everything to the classic House Prices competition dataset!


````python
# 🏠 Load House Prices dataset (using sklearn's California housing as proxy)
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
X_house = pd.DataFrame(housing.data, columns=housing.feature_names)
y_house = housing.target * 100000  # Scale to actual prices

print("✅ California Housing dataset loaded!")
print(f"Dataset shape: {X_house.shape}")
print(f"\nFeatures:")
print(X_house.head())
print(f"\nTarget statistics:")
print(f"Mean price: ${y_house.mean():,.2f}")
print(f"Median price: ${np.median(y_house):,.2f}")
print(f"Price range: ${y_house.min():,.2f} - ${y_house.max():,.2f}")
````

### Solution 2: Mini-Batch Gradient Descent
```python
class MiniBatchLinearRegression(LinearRegressionScratch):
    def __init__(self, learning_rate=0.01, n_iterations=1000, batch_size=32):
        super().__init__(learning_rate, n_iterations)
        self.batch_size = batch_size
    
    def fit(self, X, y):
        m, n = X.shape
        X_b = np.c_[np.ones((m, 1)), X]
        self.theta = np.random.randn(n + 1) * 0.01
        
        for epoch in range(self.n_iterations):
            # Shuffle data
            indices = np.random.permutation(m)
            X_shuffled = X_b[indices]
            y_shuffled = y[indices]
            
            # Mini-batch updates
            for i in range(0, m, self.batch_size):
                X_batch = X_shuffled[i:i+self.batch_size]
                y_batch = y_shuffled[i:i+self.batch_size]
                
                predictions = X_batch @ self.theta
                gradients = (1/len(y_batch)) * X_batch.T @ (predictions - y_batch)
                self.theta -= self.learning_rate * gradients
            
            # Record cost after each epoch
            cost = self.compute_cost(X_b, y)
            self.cost_history.append(cost)
        
        return self
# Mini-batch offers balance between speed (SGD) and stability (batch)
```

### Solution 3: Ridge Regression
```python
class RidgeRegression(LinearRegressionScratch):
    def __init__(self, learning_rate=0.01, n_iterations=1000, lambda_reg=1.0):
        super().__init__(learning_rate, n_iterations)
        self.lambda_reg = lambda_reg
    
    def compute_cost(self, X, y):
        m = len(y)
        predictions = self.hypothesis(X)
        mse = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
        reg_term = (self.lambda_reg / (2 * m)) * np.sum(self.theta[1:] ** 2)  # Exclude bias
        return mse + reg_term
    
    def compute_gradient(self, X, y):
        m = len(y)
        predictions = self.hypothesis(X)
        gradients = (1 / m) * X.T @ (predictions - y)
        gradients[1:] += (self.lambda_reg / m) * self.theta[1:]  # Regularize all except bias
        return gradients
# Regularization prevents overfitting by penalizing large coefficients
```
