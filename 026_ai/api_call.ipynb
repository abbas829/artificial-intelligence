## 1. Introduction to REST APIs and HTTP Methods

**REST (Representational State Transfer)** is an architectural style for designing networked applications. REST APIs use HTTP requests to perform CRUD operations (Create, Read, Update, Delete).

### Common HTTP Methods:
- **GET**: Retrieve data from a server (most common for data fetching)
- **POST**: Send data to create a new resource
- **PUT**: Update an existing resource
- **DELETE**: Remove a resource

For AI/ML data collection, **GET** is your primary tool. When you call an API, you receive a response with:
- **Status code**: Indicates success/failure (200 = OK, 404 = Not Found, 500 = Server Error)
- **Headers**: Metadata about the response
- **Body**: The actual data (usually in JSON format)

````python
# First, let's install and import the required libraries
# Note: requests is usually pre-installed, but pandas might need installation

import requests
import pandas as pd
import json
import time
import numpy as np
from datetime import datetime

print("✅ All libraries imported successfully!")
print(f"Requests version: {requests.__version__}")
print(f"Pandas version: {pd.__version__}")
````

## 2. Using the `requests` Library: GET Requests

The `requests.get()` function is the simplest way to fetch data. Let's start with a basic example using JSONPlaceholder, a free fake API for testing.

💡 **Key concept**: Always check the status code to ensure your request was successful before processing data.

````python
# Basic GET request example
url = "https://jsonplaceholder.typicode.com/posts/1"

response = requests.get(url)

# Check status code
print(f"Status Code: {response.status_code}")

if response.status_code == 200:
    print("✅ Request successful!")
    # View raw text content
    print("\nRaw response text (first 200 chars):")
    print(response.text[:200])
else:
    print(f"❌ Request failed with status code: {response.status_code}")
````

## 3. Parsing JSON Responses

Most modern APIs return data in **JSON (JavaScript Object Notation)** format. The `requests` library provides a convenient `.json()` method to automatically parse JSON into Python dictionaries or lists.

📌 This conversion is essential for AI/ML workflows—you'll typically convert these dictionaries into pandas DataFrames for analysis.

````python
# Parsing JSON response
url = "https://jsonplaceholder.typicode.com/posts/1"
response = requests.get(url)

# Convert JSON to Python dictionary
data = response.json()

print("📊 Parsed JSON data:")
print(json.dumps(data, indent=2))  # Pretty print with indentation

# Access specific fields
print(f"\n📝 Title: {data['title']}")
print(f"👤 User ID: {data['userId']}")
print(f"📄 Body preview: {data['body'][:50]}...")
````

## 4. Working with Query Parameters

Query parameters allow you to filter, sort, or customize your API requests. They're appended to the URL after a `?` and separated by `&`.

Example: `https://api.example.com/data?key1=value1&key2=value2`

With `requests`, you can pass parameters as a dictionary using the `params` argument—this is cleaner and handles URL encoding automatically.

````python
# Using query parameters
url = "https://jsonplaceholder.typicode.com/posts"

# Define parameters
params = {
    "userId": 1,  # Filter by user
    "_limit": 3   # Limit to 3 results (simulated)
}

response = requests.get(url, params=params)

print(f"Final URL: {response.url}")  # Shows the complete URL with parameters
print(f"Status Code: {response.status_code}")

posts = response.json()
print(f"\n📬 Retrieved {len(posts)} posts")

# Display first post
if posts:
    print(f"\nFirst post title: {posts[0]['title']}")
````

## 6. Error Handling for API Calls

Network operations are inherently unpredictable. Servers go down, networks fail, and APIs change. Robust error handling is essential for production ML pipelines.

### Common Exceptions:
- `requests.exceptions.ConnectionError`: Network problem (DNS failure, refused connection)
- `requests.exceptions.Timeout`: Request timed out
- `requests.exceptions.HTTPError`: HTTP error (4xx or 5xx status codes)
- `requests.exceptions.RequestException`: Base class for all exceptions

💡 **Best Practice**: Always wrap API calls in try-except blocks and implement retry logic for critical data pipelines.

````python
# Comprehensive error handling example
def safe_api_call(url, params=None, headers=None, timeout=10):
    """
    Safely make an API call with comprehensive error handling.
    
    Args:
        url: API endpoint URL
        params: Optional query parameters
        headers: Optional request headers
        timeout: Request timeout in seconds
    
    Returns:
        dict: Response data or error information
    """
    try:
        response = requests.get(url, params=params, headers=headers, timeout=timeout)
        
        # Raise HTTPError for bad status codes (4xx, 5xx)
        response.raise_for_status()
        
        return {
            "success": True,
            "data": response.json(),
            "status_code": response.status_code
        }
    
    except requests.exceptions.Timeout:
        return {"success": False, "error": "Request timed out. Try again later."}
    
    except requests.exceptions.ConnectionError:
        return {"success": False, "error": "Connection failed. Check your internet or the API URL."}
    
    except requests.exceptions.HTTPError as e:
        return {"success": False, "error": f"HTTP Error: {e}"}
    
    except requests.exceptions.RequestException as e:
        return {"success": False, "error": f"Request failed: {e}"}
    
    except json.JSONDecodeError:
        return {"success": False, "error": "Invalid JSON response from server"}

# Test with valid URL
result = safe_api_call("https://jsonplaceholder.typicode.com/posts/1")
print("✅ Valid request result:")
print(json.dumps(result, indent=2)[:500])

# Test with invalid URL (simulated error)
result_error = safe_api_call("https://jsonplaceholder.typicode.com/invalid_endpoint")
print("\n❌ Invalid request result:")
print(json.dumps(result_error, indent=2))
````

## 7. Rate Limiting and Respectful Usage

APIs are shared resources. **Rate limiting** restricts how many requests you can make in a given time period (e.g., 100 requests per minute). Exceeding limits can result in temporary blocks or permanent bans.

### Best Practices:
1. ✅ Read API documentation for rate limits
2. ✅ Add delays between requests using `time.sleep()`
3. ✅ Handle 429 (Too Many Requests) status codes gracefully
4. ✅ Use caching to avoid redundant requests
5. ✅ Respect `Retry-After` headers when rate limited

````python
# Rate limiting example: Fetching multiple items with delays
def fetch_multiple_posts(post_ids, delay=1.0):
    """
    Fetch multiple posts with respectful delays between requests.
    
    Args:
        post_ids: List of post IDs to fetch
        delay: Seconds to wait between requests
    """
    results = []
    
    print(f"🔄 Starting batch fetch of {len(post_ids)} posts...")
    
    for i, post_id in enumerate(post_ids, 1):
        url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
        
        try:
            response = requests.get(url, timeout=5)
            
            if response.status_code == 200:
                results.append(response.json())
                print(f"  [{i}/{len(post_ids)}] ✅ Fetched post {post_id}")
            elif response.status_code == 429:
                print(f"  [{i}/{len(post_ids)}] ⚠️ Rate limited! Waiting longer...")
                time.sleep(5)  # Extra wait if rate limited
            else:
                print(f"  [{i}/{len(post_ids)}] ❌ Failed: {response.status_code}")
        
        except requests.exceptions.RequestException as e:
            print(f"  [{i}/{len(post_ids)}] ❌ Error: {e}")
        
        # Respectful delay (except for last item)
        if i < len(post_ids):
            time.sleep(delay)
    
    print(f"\n📊 Completed! Successfully fetched {len(results)} posts.")
    return results

# Fetch 3 posts with 0.5 second delays
posts = fetch_multiple_posts([1, 2, 3], delay=0.5)

# Convert to DataFrame
if posts:
    df = pd.DataFrame(posts)
    print("\n📋 DataFrame preview:")
    print(df[['id', 'title', 'userId']].head())
````

### Example 1: Fetching Weather Data (Open-Meteo - No API Key Required)

Open-Meteo provides free weather data without API keys—perfect for learning and prototyping ML models for weather prediction.

````python
# Real-time weather data fetching (Open-Meteo API)
def get_weather_data(latitude=51.5074, longitude=-0.1278):
    """
    Fetch current weather data for given coordinates (default: London).
    
    Args:
        latitude: Location latitude
        longitude: Location longitude
    
    Returns:
        DataFrame with weather data
    """
    url = "https://api.open-meteo.com/v1/forecast"
    
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current": ["temperature_2m", "relative_humidity_2m", "apparent_temperature", 
                    "precipitation", "weather_code", "wind_speed_10m"],
        "timezone": "auto"
    }
    
    try:
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        data = response.json()
        
        # Extract current weather
        current = data.get('current', {})
        
        weather_df = pd.DataFrame([{
            'timestamp': current.get('time'),
            'temperature_c': current.get('temperature_2m'),
            'humidity_percent': current.get('relative_humidity_2m'),
            'apparent_temp_c': current.get('apparent_temperature'),
            'precipitation_mm': current.get('precipitation'),
            'weather_code': current.get('weather_code'),
            'wind_speed_kmh': current.get('wind_speed_10m'),
            'latitude': latitude,
            'longitude': longitude
        }])
        
        return weather_df
    
    except Exception as e:
        print(f"❌ Error fetching weather: {e}")
        return pd.DataFrame()

# Fetch weather for London
weather_df = get_weather_data(51.5074, -0.1278)

if not weather_df.empty:
    print("🌤️ Current Weather Data (London):")
    print(weather_df.to_string(index=False))
    
    print(f"\n💡 Temperature: {weather_df['temperature_c'].values[0]}°C")
    print(f"💡 Humidity: {weather_df['humidity_percent'].values[0]}%")
````

### Example 2: Cryptocurrency Prices (CoinGecko API - Free Tier)

CoinGecko offers a comprehensive free API for cryptocurrency data—ideal for building crypto price prediction models or portfolio trackers.

````python
# Fetching cryptocurrency prices
def get_crypto_prices(coins=['bitcoin', 'ethereum', 'cardano']):
    """
    Fetch current prices for specified cryptocurrencies.
    
    Args:
        coins: List of coin IDs (as per CoinGecko)
    
    Returns:
        DataFrame with price data
    """
    url = "https://api.coingecko.com/api/v3/simple/price"
    
    params = {
        "ids": ",".join(coins),
        "vs_currencies": "usd,eur",
        "include_24hr_change": "true",
        "include_market_cap": "true",
        "include_24hr_vol": "true"
    }
    
    try:
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        data = response.json()
        
        # Convert nested dict to DataFrame
        records = []
        for coin, info in data.items():
            records.append({
                'coin': coin,
                'price_usd': info.get('usd'),
                'price_eur': info.get('eur'),
                'market_cap_usd': info.get('usd_market_cap'),
                'volume_24h_usd': info.get('usd_24h_vol'),
                'change_24h_percent': info.get('usd_24h_change'),
                'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            })
        
        return pd.DataFrame(records)
    
    except Exception as e:
        print(f"❌ Error fetching crypto data: {e}")
        return pd.DataFrame()

# Fetch prices (be mindful of rate limits - don't run too frequently!)
print("⏳ Fetching cryptocurrency prices...")
crypto_df = get_crypto_prices(['bitcoin', 'ethereum', 'cardano'])

if not crypto_df.empty:
    print("\n💰 Cryptocurrency Prices:")
    print(crypto_df[['coin', 'price_usd', 'change_24h_percent', 'timestamp']].to_string(index=False))
    
    # Quick analysis
    print(f"\n📈 Highest 24h change: {crypto_df.loc[crypto_df['change_24h_percent'].idxmax(), 'coin']}")
    print(f"📉 Lowest 24h change: {crypto_df.loc[crypto_df['change_24h_percent'].idxmin(), 'coin']}")
````

For NLP projects, you need text data. News APIs provide headlines and articles for sentiment analysis, topic modeling, or fake news detection. Below we simulate the structure using a mock endpoint.

### Example 4: Building a Simple Data Collection Function

Let's create a reusable function that encapsulates best practices for API data collection—perfect for building your own ML datasets.

````python
# Production-ready data collection function
class APIDataCollector:
    """
    A reusable class for collecting API data with built-in safety features.
    """
    
    def __init__(self, base_url, api_key=None, rate_limit_delay=1.0):
        self.base_url = base_url
        self.api_key = api_key
        self.rate_limit_delay = rate_limit_delay
        self.session = requests.Session()  # Reuse connections for performance
        
        if api_key:
            self.session.headers.update({"Authorization": f"Bearer {api_key}"})
    
    def fetch(self, endpoint, params=None, retries=3):
        """
        Fetch data with automatic retries and error handling.
        """
        url = f"{self.base_url}/{endpoint}"
        
        for attempt in range(retries):
            try:
                response = self.session.get(url, params=params, timeout=10)
                
                if response.status_code == 200:
                    time.sleep(self.rate_limit_delay)  # Respect rate limits
                    return {"success": True, "data": response.json()}
                
                elif response.status_code == 429:
                    wait_time = (attempt + 1) * 2
                    print(f"⚠️ Rate limited. Waiting {wait_time}s...")
                    time.sleep(wait_time)
                
                else:
                    return {"success": False, "error": f"HTTP {response.status_code}"}
            
            except requests.exceptions.RequestException as e:
                if attempt == retries - 1:
                    return {"success": False, "error": str(e)}
                time.sleep(1)
        
        return {"success": False, "error": "Max retries exceeded"}
    
    def to_dataframe(self, data_list):
        """Convert list of dicts to DataFrame with error handling."""
        try:
            return pd.DataFrame(data_list)
        except Exception as e:
            print(f"❌ DataFrame conversion error: {e}")
            return pd.DataFrame()

# Usage example
collector = APIDataCollector(
    base_url="https://jsonplaceholder.typicode.com",
    rate_limit_delay=0.5
)

# Fetch multiple endpoints
endpoints = ["posts/1", "posts/2", "users/1"]
all_data = []

for endpoint in endpoints:
    result = collector.fetch(endpoint)
    if result["success"]:
        all_data.append(result["data"])
        print(f"✅ Fetched: {endpoint}")
    else:
        print(f"❌ Failed: {endpoint} - {result['error']}")

print(f"\n📊 Collected {len(all_data)} records successfully")

# Convert to DataFrame if data structure allows
if all_data and isinstance(all_data[0], dict):
    df_collected = pd.DataFrame(all_data)
    print("\n📋 Preview:")
    print(df_collected.head())
````

### Exercise 7: Fetch Multiple Pages

Create a function that fetches posts 1 through 15 with a 0.3-second delay between each request. Store all results in a list and convert to a single DataFrame. Handle any failed requests gracefully (skip them and continue).

````python
import requests
import pandas as pd
import time

def fetch_posts_with_delay(post_ids, delay=0.3):
    results = []
    
    for post_id in post_ids:
        try:
            url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
            response = requests.get(url, timeout=5)
            if response.status_code == 200:
                results.append(response.json())
        except Exception as e:
            print(f"⚠️ Skipped post {post_id}: {e}")
        
        time.sleep(delay)
    
    return pd.DataFrame(results)

df = fetch_posts_with_delay(range(1, 16), delay=0.3)
print(f"Fetched {len(df)} posts")
print(df[['id', 'title']].head())
````

### Exercise 10: Build a Crypto Tracker Function

Create a function that accepts a list of cryptocurrency IDs and returns a DataFrame with current prices in USD. Add error handling for invalid coin IDs and include a timestamp column. Test it with at least 3 different coins.

````python
import requests
import pandas as pd
from datetime import datetime

def get_crypto_prices_simple(coin_ids):
    url = "https://api.coingecko.com/api/v3/simple/price"
    params = {
        "ids": ",".join(coin_ids),
        "vs_currencies": "usd",
        "include_24hr_change": "true"
    }
    
    try:
        response = requests.get(url, timeout=10)
        data = response.json()
        
        records = []
        for coin in coin_ids:
            if coin in data:
                records.append({
                    'coin_id': coin,
                    'price_usd': data[coin]['usd'],
                    'change_24h': data[coin].get('usd_24h_change'),
                    'timestamp': datetime.now()
                })
        
        return pd.DataFrame(records)
    except Exception as e:
        print(f"Error: {e}")
        return pd.DataFrame()

# Test
df_crypto = get_crypto_prices_simple(['bitcoin', 'ethereum', 'cardano'])
print(df_crypto)
````

### Exercise 11: Mini Data Pipeline

Build a complete data collection pipeline that:
1. Fetches weather data for your city
2. Fetches crypto prices for Bitcoin and Ethereum
3. Saves both datasets to CSV files with timestamps in filenames
4. Implements a 2-second delay between API calls
5. Logs success/failure for each step

This simulates a real ML data collection workflow!

````python
import requests
import pandas as pd
import time
from datetime import datetime

def run_data_pipeline_simple():
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    logs = []
    
    # Step 1: Weather data (simplified)
    try:
        weather_url = "https://api.open-meteo.com/v1/forecast"
        params = {
            "latitude": 51.5074,
            "longitude": -0.1278,
            "current": "temperature_2m",
            "timezone": "auto"
        }
        response = requests.get(weather_url, params=params, timeout=10)
        weather_data = response.json()
        
        df_weather = pd.DataFrame([{
            'temperature': weather_data['current']['temperature_2m'],
            'timestamp': datetime.now()
        }])
        
        # Save to CSV
        weather_file = f"weather_{timestamp}.csv"
        df_weather.to_csv(weather_file, index=False)
        logs.append(f"✅ Weather data saved to {weather_file}")
    except Exception as e:
        logs.append(f"❌ Weather failed: {e}")
    
    time.sleep(2)  # Delay between APIs
    
    # Step 2: Crypto data
    try:
        crypto_url = "https://api.coingecko.com/api/v3/simple/price"
        params = {
            "ids": "bitcoin,ethereum",
            "vs_currencies": "usd",
            "include_24hr_change": "true"
        }
        response = requests.get(crypto_url, params=params, timeout=10)
        crypto_data = response.json()
        
        records = []
        for coin, data in crypto_data.items():
            records.append({
                'coin': coin,
                'price_usd': data['usd'],
                'change_24h': data.get('usd_24h_change'),
                'timestamp': datetime.now()
            })
        
        df_crypto = pd.DataFrame(records)
        crypto_file = f"crypto_{timestamp}.csv"
        df_crypto.to_csv(crypto_file, index=False)
        logs.append(f"✅ Crypto data saved to {crypto_file}")
    except Exception as e:
        logs.append(f"❌ Crypto failed: {e}")
    
    # Print logs
    print("\n".join(logs))
    return logs

# Run pipeline
run_data_pipeline_simple()
````

### Solution 5: Error Handling

````python
import requests

def safe_fetch(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        return {"success": True, "data": response.json()}
    except requests.exceptions.Timeout:
        return {"success": False, "error": "Request timed out after 5 seconds"}
    except requests.exceptions.ConnectionError:
        return {"success": False, "error": "Failed to connect to server"}
    except requests.exceptions.HTTPError as e:
        if response.status_code == 404:
            return {"success": False, "error": "Resource not found (404)"}
        return {"success": False, "error": f"HTTP error: {e}"}

# Test valid URL
print("Testing valid URL:")
print(safe_fetch("https://jsonplaceholder.typicode.com/posts/1"))

# Test invalid URL (404)
print("\nTesting invalid URL:")
print(safe_fetch("https://jsonplaceholder.typicode.com/nonexistent"))
````

### Solution 7: Fetch Multiple Pages

````python
import requests
import pandas as pd
import time

def fetch_posts_with_delay(post_ids, delay=0.3):
    results = []
    
    for post_id in post_ids:
        try:
            url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
            response = requests.get(url, timeout=5)
            if response.status_code == 200:
                results.append(response.json())
        except Exception as e:
            print(f"⚠️ Skipped post {post_id}: {e}")
        
        time.sleep(delay)
    
    return pd.DataFrame(results)

df = fetch_posts_with_delay(range(1, 16), delay=0.3)
print(f"Fetched {len(df)} posts")
print(df[['id', 'title']].head())
````

### Solution 9: Real-Time Weather Collector

````python
import requests
import pandas as pd

def get_weather(lat, lon, city_name):
    """Fetch weather data with proper error handling"""
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": lat,
        "longitude": lon,
        "current": "temperature_2m",
        "timezone": "auto"
    }
    try:
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        data = response.json()
        return {
            'city': city_name,
            'temperature': data['current']['temperature_2m'],
            'timestamp': data['current']['time']
        }
    except requests.exceptions.Timeout:
        print(f"⚠️ Timeout fetching weather for {city_name}")
        return {
            'city': city_name,
            'temperature': None,
            'timestamp': None,
            'error': 'Timeout'
        }
    except requests.exceptions.RequestException as e:
        print(f"⚠️ Error fetching weather for {city_name}: {e}")
        return {
            'city': city_name,
            'temperature': None,
            'timestamp': None,
            'error': str(e)
        }

cities = [
    (51.5074, -0.1278, "London"),
    (40.7128, -74.0060, "New York"),
    (35.6762, 139.6503, "Tokyo")
]

weather_data = [get_weather(lat, lon, city) for lat, lon, city in cities]
df_weather = pd.DataFrame(weather_data)

print(df_weather)
if 'temperature' in df_weather.columns:
    valid_temps = df_weather['temperature'].dropna()
    if len(valid_temps) > 0:
        print(f"\n🌡️ Average temperature: {valid_temps.mean():.1f}°C")
    else:
        print("\n⚠️ No valid temperature data available")
````

### Solution 11: Mini Data Pipeline

````python
import requests
import pandas as pd
import time
from datetime import datetime

def run_data_pipeline():
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    logs = []
    
    # Step 1: Weather data
    try:
        weather_url = "https://api.open-meteo.com/v1/forecast"
        params = {
            "latitude": 51.5074,
            "longitude": -0.1278,
            "current": ["temperature_2m", "humidity_2m"],
            "timezone": "auto"
        }
        response = requests.get(weather_url, params=params, timeout=10)
        weather_data = response.json()
        
        df_weather = pd.DataFrame([{
            'temperature': weather_data['current']['temperature_2m'],
            'humidity': weather_data['current']['humidity_2m'],
            'timestamp': datetime.now()
        }])
        
        weather_file = f"weather_{timestamp}.csv"
        df_weather.to_csv(weather_file, index=False)
        logs.append(f"✅ Weather data saved to {weather_file}")
    except Exception as e:
        logs.append(f"❌ Weather failed: {e}")
    
    time.sleep(2)  # Delay between APIs
    
    # Step 2: Crypto data
    try:
        crypto_url = "https://api.coingecko.com/api/v3/simple/price"
        params = {
            "ids": "bitcoin,ethereum",
            "vs_currencies": "usd",
            "include_24hr_change": "true"
        }
        response = requests.get(crypto_url, params=params, timeout=10)
        crypto_data = response.json()
        
        records = []
        for coin, data in crypto_data.items():
            records.append({
                'coin': coin,
                'price_usd': data['usd'],
                'change_24h': data.get('usd_24h_change'),
                'timestamp': datetime.now()
            })
        
        df_crypto = pd.DataFrame(records)
        crypto_file = f"crypto_{timestamp}.csv"
        df_crypto.to_csv(crypto_file, index=False)
        logs.append(f"✅ Crypto data saved to {crypto_file}")
    except Exception as e:
        logs.append(f"❌ Crypto failed: {e}")
    
    # Print logs
    print("\n".join(logs))
    return logs

# Run pipeline
run_data_pipeline()
````
