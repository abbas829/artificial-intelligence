```python
# ============================================================
# Generate Realistic Customer Churn Dataset
# ============================================================
print_section("Step 1: Data Loading & Advanced EDA", "📂")

np.random.seed(42)
n_samples = 7000

logger.info(f"Generating synthetic dataset with {n_samples} customers")

# Customer Demographics
customer_id = [f'CUST_{i:06d}' for i in range(n_samples)]
gender = np.random.choice(['Male', 'Female'], n_samples)
senior_citizen = np.random.choice([0, 1], n_samples, p=[0.83, 0.17])
partner = np.random.choice(['Yes', 'No'], n_samples, p=[0.48, 0.52])
dependents = np.random.choice(['Yes', 'No'], n_samples, p=[0.30, 0.70])

# Geographic
city = np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], n_samples)
region = np.random.choice(['Northeast', 'West', 'Midwest', 'South'], n_samples)

# Account Information
tenure = np.random.randint(0, 73, n_samples)
contract = np.random.choice(
    ['Month-to-month', 'One year', 'Two year'],
    n_samples,
    p=[0.55, 0.24, 0.21]
)
paperless_billing = np.random.choice(['Yes', 'No'], n_samples, p=[0.59, 0.41])
payment_method = np.random.choice(
    ['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)'],
    n_samples,
    p=[0.34, 0.16, 0.25, 0.25]
)

# Services
phone_service = np.random.choice(['Yes', 'No'], n_samples, p=[0.90, 0.10])
multiple_lines = np.where(
    phone_service == 'Yes',
    np.random.choice(['Yes', 'No'], n_samples, p=[0.42, 0.58]),
    'No phone service'
)
internet_service = np.random.choice(
    ['DSL', 'Fiber optic', 'No'],
    n_samples,
    p=[0.34, 0.44, 0.22]
)

# Internet add-ons
internet_addons = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
addon_data = {}
for addon in internet_addons:
    addon_data[addon] = np.where(
        internet_service != 'No',
        np.random.choice(['Yes', 'No'], n_samples, p=[0.36, 0.64]),
        'No internet service'
    )

# Financial
monthly_charges = np.random.gamma(7, 9, n_samples) + 18
monthly_charges = np.clip(monthly_charges, 18, 120)
total_charges = monthly_charges * tenure + np.random.normal(0, 150, n_samples)
total_charges = np.clip(total_charges, 0, None)

# Support interactions
support_calls = np.random.poisson(1.2, n_samples)
support_calls = np.clip(support_calls, 0, 10)
satisfaction_score = np.random.choice([1, 2, 3, 4, 5], n_samples, p=[0.08, 0.12, 0.20, 0.35, 0.25])

# Create DataFrame
df = pd.DataFrame({
    'customerID': customer_id,
    'gender': gender,
    'SeniorCitizen': senior_citizen,
    'Partner': partner,
    'Dependents': dependents,
    'city': city,
    'region': region,
    'tenure': tenure,
    'PhoneService': phone_service,
    'MultipleLines': multiple_lines,
    'InternetService': internet_service,
    'Contract': contract,
    'PaperlessBilling': paperless_billing,
    'PaymentMethod': payment_method,
    'MonthlyCharges': monthly_charges,
    'TotalCharges': total_charges,
    'SupportCalls': support_calls,
    'SatisfactionScore': satisfaction_score,
    **addon_data
})

# ============================================================
# Generate Churn with Realistic Patterns
# ============================================================
churn_prob = np.zeros(n_samples)

# Base risk factors
churn_prob += (df['Contract'] == 'Month-to-month').astype(int) * 0.22
churn_prob += (df['InternetService'] == 'Fiber optic').astype(int) * 0.14
churn_prob += (df['PaymentMethod'] == 'Electronic check').astype(int) * 0.11
churn_prob += (df['tenure'] < 12).astype(int) * 0.18
churn_prob += (df['TechSupport'] == 'Yes').astype(int) * -0.09
churn_prob += (df['MonthlyCharges'] > 80).astype(int) * 0.12
churn_prob += df['SeniorCitizen'] * 0.06
churn_prob += (df['Dependents'] == 'No').astype(int) * 0.06
churn_prob += (df['SatisfactionScore'] <= 2).astype(int) * 0.15
churn_prob += (df['SupportCalls'] > 3).astype(int) * 0.10
churn_prob += (df['OnlineSecurity'] == 'No').astype(int) * 0.08

# Normalize and add noise
churn_prob = np.clip(churn_prob + np.random.normal(0, 0.06, n_samples), 0.01, 0.99)
df['Churn'] = np.random.binomial(1, churn_prob)

print_success(f"Dataset generated: {n_samples:,} customers")
print_info(f"Churn Rate: {df['Churn'].mean():.1%}")
print_info(f"Features: {df.shape[1]} columns")
print_info(f"Memory: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

## 4. Step 2: Comprehensive Feature Engineering


```python
print_section("Step 2: Comprehensive Feature Engineering", "🚀")

# ============================================================
# Feature Engineering Pipeline
# ============================================================
class FeatureEngineer:
    """Production-grade feature engineering pipeline."""
    
    def __init__(self):
        self.feature_stats = {}
    
    def fit(self, df: pd.DataFrame) -> 'FeatureEngineer':
        """Compute statistics needed for feature engineering."""
        self.feature_stats['monthly_charges_median'] = df['MonthlyCharges'].median()
        self.feature_stats['monthly_charges_75th'] = df['MonthlyCharges'].quantile(0.75)
        self.feature_stats['tenure_bins'] = [0, 6, 12, 24, 48, float('inf')]
        return self
    
    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        """Apply feature engineering transformations."""
        df = df.copy()
        
        # 1. Financial Features
        df['AvgMonthlySpend'] = df['TotalCharges'] / (df['tenure'] + 1)
        df['ChargesPerMonth'] = df['MonthlyCharges'] / (df['tenure'] + 1)
        df['CLV_Proxy'] = df['MonthlyCharges'] * df['tenure']
        df['PriceSensitivity'] = df['MonthlyCharges'] - self.feature_stats['monthly_charges_median']
        df['HighValueCustomer'] = (df['MonthlyCharges'] > self.feature_stats['monthly_charges_75th']).astype(int)
        
        # 2. Tenure Features
        tenure_filled = df['tenure'].fillna(0).clip(lower=0)
        df['TenureGroup'] = pd.cut(
            tenure_filled,
            bins=self.feature_stats['tenure_bins'],
            labels=False,
            right=False,
            include_lowest=True
        ).fillna(0).astype(int)
        df['IsNewCustomer'] = (df['tenure'] <= 6).astype(int)
        df['IsLoyalCustomer'] = (df['tenure'] >= 48).astype(int)
        
        # 3. Contract Risk Features
        contract_risk = {'Month-to-month': 1.0, 'One year': 0.4, 'Two year': 0.15}
        df['ContractRiskScore'] = df['Contract'].map(contract_risk)
        
        # 4. Service Bundle Features
        service_cols = ['PhoneService', 'OnlineSecurity', 'OnlineBackup',
                        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
        df['ServiceBundleCount'] = 0
        for col in service_cols:
            if col in df.columns:
                df['ServiceBundleCount'] += (df[col] == 'Yes').astype(int)
        
        df['HasInternet'] = (df['InternetService'] != 'No').astype(int)
        df['HasPhoneAndInternet'] = ((df['PhoneService'] == 'Yes') & (df['InternetService'] != 'No')).astype(int)
        
        # 5. Support & Satisfaction Features
        df['HighSupportUsage'] = (df['SupportCalls'] > 3).astype(int)
        df['LowSatisfaction'] = (df['SatisfactionScore'] <= 2).astype(int)
        df['SupportSatisfactionRatio'] = df['SupportCalls'] / (df['SatisfactionScore'] + 1)
        
        # 6. Payment Risk Features
        df['IsElectronicCheck'] = (df['PaymentMethod'] == 'Electronic check').astype(int)
        df['IsAutoPayment'] = df['PaymentMethod'].str.contains('automatic').astype(int)
        
        # 7. Demographic Features
        df['IsSeniorAlone'] = ((df['SeniorCitizen'] == 1) & (df['Partner'] == 'No')).astype(int)
        df['HasFamily'] = ((df['Partner'] == 'Yes') | (df['Dependents'] == 'Yes')).astype(int)
        
        # 8. Interaction Features
        df['Tenure_x_MonthlyCharges'] = df['tenure'] * df['MonthlyCharges']
        df['ContractRisk_x_Tenure'] = df['ContractRiskScore'] * df['tenure']
        
        return df
    
    def fit_transform(self, df: pd.DataFrame) -> pd.DataFrame:
        """Fit and transform in one step."""
        return self.fit(df).transform(df)

# Apply feature engineering
fe = FeatureEngineer()
df_engineered = fe.fit_transform(df)

print_success(f"Engineered {df_engineered.shape[1] - df.shape[1]} new features!")
print(f"📋 Total features: {df_engineered.shape[1]}")

new_features = [col for col in df_engineered.columns if col not in df.columns]
print(f"\n📊 New Features: {new_features}")
```

## 5. Step 3: Model Training with Multiple Algorithms + AutoML Comparison


```python
print_section("Step 3: Model Training & AutoML Comparison", "🤖")

# ============================================================
# Data Preparation
# ============================================================
# Drop non-feature columns
X = df_engineered.drop(['customerID', 'Churn'], axis=1, errors='ignore')
y = df_engineered['Churn']

# One-hot encode categoricals
X_encoded = pd.get_dummies(X, drop_first=True)
feature_names = X_encoded.columns.tolist()
scaled_model_names = {'Logistic Regression', 'SVM (RBF)', 'K-Nearest Neighbors', 'Naive Bayes'}

# Stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

# Scale for algorithms that need it
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print_info(f"Training set: {X_train.shape[0]:,} samples")
print_info(f"Test set: {X_test.shape[0]:,} samples")
print_info(f"Features: {X_train.shape[1]}")
print_info(f"Training churn rate: {y_train.mean():.2%}")
```

```python
# ============================================================
# Define All Models
# ============================================================
models = {
    'Logistic Regression': LogisticRegression(max_iter=2000, random_state=42, class_weight='balanced'),
    'Random Forest': RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced', n_jobs=1),
    'Extra Trees': ExtraTreesClassifier(n_estimators=300, random_state=42, class_weight='balanced', n_jobs=1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=300, random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=300, random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB(),
    'SVM (RBF)': SVC(probability=True, random_state=42, class_weight='balanced')
}

# Add advanced boosting models
if XGBOOST_AVAILABLE:
    models['XGBoost'] = xgb.XGBClassifier(
        n_estimators=300, max_depth=6, learning_rate=0.1,
        subsample=0.8, colsample_bytree=0.8, random_state=42,
        eval_metric='logloss', use_label_encoder=False
    )

if LIGHTGBM_AVAILABLE:
    models['LightGBM'] = lgb.LGBMClassifier(
        n_estimators=300, max_depth=6, learning_rate=0.1,
        subsample=0.8, colsample_bytree=0.8, random_state=42,
        verbose=-1
    )

if CATBOOST_AVAILABLE:
    models['CatBoost'] = cb.CatBoostClassifier(
        iterations=300, depth=6, learning_rate=0.1,
        random_seed=42, verbose=False
    )

print(f"📋 Models to evaluate: {len(models)} algorithms")
print(f"   {list(models.keys())}")
```

```python
# ============================================================
# Cross-Validation Evaluation
# ============================================================
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {}

print("\n🔬 Running 5-Fold Stratified Cross-Validation...\n")

for name, model in models.items():
    logger.info(f"Training {name}...")
    
    # Use scaled data for distance-based and linear models
    if name in scaled_model_names:
        X_cv = X_train_scaled
    else:
        X_cv = X_train.values
    
    # Multiple metrics
    metrics = {}
    metrics['Accuracy'] = cross_val_score(model, X_cv, y_train, cv=cv, scoring='accuracy').mean()
    metrics['Precision'] = cross_val_score(model, X_cv, y_train, cv=cv, scoring='precision').mean()
    metrics['Recall'] = cross_val_score(model, X_cv, y_train, cv=cv, scoring='recall').mean()
    metrics['F1'] = cross_val_score(model, X_cv, y_train, cv=cv, scoring='f1').mean()
    metrics['ROC_AUC'] = cross_val_score(model, X_cv, y_train, cv=cv, scoring='roc_auc').mean()
    metrics['ROC_AUC_std'] = cross_val_score(model, X_cv, y_train, cv=cv, scoring='roc_auc').std()
    
    results[name] = metrics
    print(f"  ✅ {name:25s} | AUC: {metrics['ROC_AUC']:.4f} (+/- {metrics['ROC_AUC_std']*2:.4f})")

print_success("\nCross-validation complete for all models!")
```

```python
# ============================================================
# Results Comparison & Visualization
# ============================================================
results_df = pd.DataFrame(results).T
results_df = results_df.sort_values('ROC_AUC', ascending=False)

print("\n📊 Complete Model Comparison (Sorted by ROC-AUC):")
display(results_df.round(4))

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('AutoML Model Comparison Dashboard', fontsize=16, fontweight='bold')

# 1. ROC-AUC Ranking
ax1 = axes[0, 0]
colors = plt.cm.Spectral(np.linspace(0.2, 0.8, len(results_df)))
bars = ax1.barh(results_df.index, results_df['ROC_AUC'], color=colors,
                xerr=results_df['ROC_AUC_std'], capsize=4, edgecolor='black', linewidth=1)
ax1.set_xlabel('ROC-AUC Score', fontsize=11)
ax1.set_title('Model Ranking by ROC-AUC', fontsize=13, fontweight='bold')
ax1.set_xlim(0.5, 1.0)
for i, (idx, row) in enumerate(results_df.iterrows()):
    ax1.text(row['ROC_AUC'] + 0.01, i, f'{row["ROC_AUC"]:.3f}', va='center', fontsize=10)

# 2. Multi-metric Radar
ax2 = axes[0, 1]
metrics_plot = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC_AUC']
x = np.arange(len(metrics_plot))
width = 0.08
for i, (model, row) in enumerate(results_df.head(6).iterrows()):
    values = [row[m] for m in metrics_plot]
    ax2.bar(x + i*width, values, width, label=model, alpha=0.85)
ax2.set_ylabel('Score', fontsize=11)
ax2.set_title('Top 6 Models - Multi-Metric Comparison', fontsize=13, fontweight='bold')
ax2.set_xticks(x + width * 2.5)
ax2.set_xticklabels(metrics_plot, rotation=30, ha='right')
ax2.legend(fontsize=8, loc='upper left')
ax2.set_ylim(0, 1.05)

# 3. Precision vs Recall
ax3 = axes[1, 0]
ax3.scatter(results_df['Recall'], results_df['Precision'],
           s=200, c=results_df['ROC_AUC'], cmap='RdYlGn', edgecolors='black', linewidth=1.5)
for i, model in enumerate(results_df.index):
    ax3.annotate(model, (results_df.loc[model, 'Recall'], results_df.loc[model, 'Precision']),
                fontsize=8, ha='center', va='bottom')
ax3.set_xlabel('Recall', fontsize=11)
ax3.set_ylabel('Precision', fontsize=11)
ax3.set_title('Precision vs Recall Trade-off', fontsize=13, fontweight='bold')
ax3.grid(True, alpha=0.3)

# 4. Model Complexity vs Performance
ax4 = axes[1, 1]
complexity_scores = {'Logistic Regression': 1, 'Naive Bayes': 1, 'K-Nearest Neighbors': 2,
                     'SVM (RBF)': 3, 'AdaBoost': 4, 'Gradient Boosting': 5,
                     'Random Forest': 5, 'Extra Trees': 5, 'XGBoost': 6, 'LightGBM': 6, 'CatBoost': 6}
results_df['Complexity'] = results_df.index.map(complexity_scores)
scatter = ax4.scatter(results_df['Complexity'], results_df['ROC_AUC'],
                     s=300, c=results_df['F1'], cmap='plasma', edgecolors='black', linewidth=2)
for i, model in enumerate(results_df.index):
    ax4.annotate(model, (results_df.loc[model, 'Complexity'], results_df.loc[model, 'ROC_AUC']),
                fontsize=8, ha='center', va='bottom')
ax4.set_xlabel('Model Complexity', fontsize=11)
ax4.set_ylabel('ROC-AUC', fontsize=11)
ax4.set_title('Complexity vs Performance (Color = F1)', fontsize=13, fontweight='bold')
plt.colorbar(scatter, ax=ax4, label='F1 Score')

plt.tight_layout()
plt.show()

best_model_name = results_df.index[0]
print(f"\n🏆 Best Model: {best_model_name} (ROC-AUC: {results_df.loc[best_model_name, 'ROC_AUC']:.4f})")
```

## 6. Step 4: Advanced Model Evaluation & Cross-Validation


```python
print_section("Step 4: Advanced Model Evaluation", "Eval")

# ============================================================
# Hyperparameter Tuning with Optuna
# ============================================================
tunable_models = {'Random Forest', 'Gradient Boosting', 'XGBoost', 'LightGBM', 'CatBoost'}
if OPTUNA_AVAILABLE and best_model_name in tunable_models:
    print(f"Tuning {best_model_name} with Bayesian Optimization\n")
    
    def objective(trial):
        if best_model_name == 'Random Forest':
            params = {
                'n_estimators': trial.suggest_int('n_estimators', 100, 600),
                'max_depth': trial.suggest_int('max_depth', 3, 25),
                'min_samples_split': trial.suggest_int('min_samples_split', 2, 25),
                'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 15),
                'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None])
            }
            model = RandomForestClassifier(**params, random_state=42, class_weight='balanced', n_jobs=1)
            
        elif best_model_name == 'XGBoost' and XGBOOST_AVAILABLE:
            params = {
                'n_estimators': trial.suggest_int('n_estimators', 100, 600),
                'max_depth': trial.suggest_int('max_depth', 3, 12),
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
                'subsample': trial.suggest_float('subsample', 0.5, 1.0),
                'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
                'gamma': trial.suggest_float('gamma', 0, 10),
                'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
                'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True),
                'min_child_weight': trial.suggest_int('min_child_weight', 1, 10)
            }
            model = xgb.XGBClassifier(**params, random_state=42, eval_metric='logloss', use_label_encoder=False)
            
        elif best_model_name == 'LightGBM' and LIGHTGBM_AVAILABLE:
            params = {
                'n_estimators': trial.suggest_int('n_estimators', 100, 600),
                'max_depth': trial.suggest_int('max_depth', 3, 12),
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
                'num_leaves': trial.suggest_int('num_leaves', 20, 200),
                'subsample': trial.suggest_float('subsample', 0.5, 1.0),
                'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
                'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
                'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True)
            }
            model = lgb.LGBMClassifier(**params, random_state=42, verbose=-1)
            
        elif best_model_name == 'CatBoost' and CATBOOST_AVAILABLE:
            params = {
                'iterations': trial.suggest_int('iterations', 100, 600),
                'depth': trial.suggest_int('depth', 4, 12),
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
                'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-8, 10.0, log=True),
                'border_count': trial.suggest_int('border_count', 32, 255)
            }
            model = cb.CatBoostClassifier(**params, random_seed=42, verbose=False)
        else:
            params = {
                'n_estimators': trial.suggest_int('n_estimators', 100, 600),
                'max_depth': trial.suggest_int('max_depth', 3, 12),
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True)
            }
            model = GradientBoostingClassifier(**params, random_state=42)
        
        X_tune = X_train_scaled if best_model_name in scaled_model_names else X_train.values
        score = cross_val_score(model, X_tune, y_train, cv=cv, scoring='roc_auc').mean()
        return score
    
    study = optuna.create_study(direction='maximize', sampler=TPESampler(seed=42))
    study.optimize(objective, n_trials=50, show_progress_bar=True)
    
    print(f"\nBest Trial Results:")
    print(f"   ROC-AUC: {study.best_value:.4f}")
    print("   Best Parameters:")
    for key, val in study.best_params.items():
        print(f"      {key}: {val}")
    
    # Train final model
    if best_model_name == 'Random Forest':
        final_model = RandomForestClassifier(**study.best_params, random_state=42, class_weight='balanced', n_jobs=1)
    elif best_model_name == 'XGBoost' and XGBOOST_AVAILABLE:
        final_model = xgb.XGBClassifier(**study.best_params, random_state=42, eval_metric='logloss', use_label_encoder=False)
    elif best_model_name == 'LightGBM' and LIGHTGBM_AVAILABLE:
        final_model = lgb.LGBMClassifier(**study.best_params, random_state=42, verbose=-1)
    elif best_model_name == 'CatBoost' and CATBOOST_AVAILABLE:
        final_model = cb.CatBoostClassifier(**study.best_params, random_seed=42, verbose=False)
    else:
        final_model = GradientBoostingClassifier(**study.best_params, random_state=42)
        
else:
    if OPTUNA_AVAILABLE:
        print_warning(f"Optuna tuning is not configured for {best_model_name}. Using the best cross-validated base estimator.")
        final_model = clone(models[best_model_name])
    elif best_model_name == 'Random Forest':
        print("Optuna not available. Using GridSearchCV fallback.")
        param_grid = {
            'n_estimators': [200, 300, 400],
            'max_depth': [6, 10, 15],
            'min_samples_split': [2, 5, 10]
        }
        grid = GridSearchCV(RandomForestClassifier(random_state=42, class_weight='balanced', n_jobs=1),
                            param_grid, cv=cv, scoring='roc_auc', n_jobs=1)
        grid.fit(X_train.values, y_train)
        final_model = grid.best_estimator_
        best_model_name = 'Random Forest'
        print(f"Best CV Score: {grid.best_score_:.4f}")
    else:
        print_warning(f"Optuna is unavailable, so {best_model_name} will be trained with its validated baseline settings.")
        final_model = clone(models[best_model_name])
```

## 8. Step 6: Building a Production Inference Pipeline


```python
print_section("Step 6: Production Inference Pipeline", "????")

# ============================================================
# Production-Grade Prediction Pipeline
# ============================================================
class ChurnPredictionSystem:
    """
    Production-ready Customer Churn Prediction System.
    
    Handles complete ML workflow from raw data to business insights.
    """
    
    VERSION = "2.0.0"
    
    def __init__(self, model, scaler, feature_engineer, feature_names, explainer=None, requires_scaling: bool = False):
        self.model = model
        self.scaler = scaler
        self.feature_engineer = feature_engineer
        self.feature_names = feature_names
        self.explainer = explainer
        self.requires_scaling = requires_scaling
        self.metadata = {
            'version': self.VERSION,
            'created_at': datetime.now().isoformat(),
            'model_type': type(model).__name__,
            'n_features': len(feature_names)
        }
        logger.info(f"Pipeline initialized (v{self.VERSION})")
    
    def preprocess(self, df_raw: pd.DataFrame) -> np.ndarray:
        """Full preprocessing pipeline."""
        # Feature engineering
        df = self.feature_engineer.transform(df_raw)
        
        # One-hot encoding alignment
        df_encoded = pd.get_dummies(df, drop_first=True)
        
        # Ensure feature alignment
        for col in self.feature_names:
            if col not in df_encoded.columns:
                df_encoded[col] = 0
        
        X = df_encoded[self.feature_names].values
        if self.requires_scaling and self.scaler is not None:
            X = self.scaler.transform(X)
        return X
    
    def predict(self, df_raw: pd.DataFrame, return_proba: bool = True) -> np.ndarray:
        """Generate predictions."""
        X = self.preprocess(df_raw)
        if return_proba:
            return self.model.predict_proba(X)[:, 1]
        return self.model.predict(X)
    
    def predict_risk_tier(self, df_raw: pd.DataFrame, probabilities: Optional[np.ndarray] = None) -> List[str]:
        """Classify into risk tiers."""
        proba = probabilities if probabilities is not None else self.predict(df_raw, return_proba=True)
        tiers = []
        for p in proba:
            if p < 0.25:
                tiers.append('Low Risk')
            elif p < 0.50:
                tiers.append('Medium Risk')
            elif p < 0.75:
                tiers.append('High Risk')
            else:
                tiers.append('Critical Risk')
        return tiers
    
    def explain(self, df_raw: pd.DataFrame, idx: int = 0) -> Dict:
        """Generate SHAP explanation for a customer."""
        if self.explainer is None:
            return {'error': 'SHAP explainer not available'}
        
        X = self.preprocess(df_raw)
        shap_vals = self.explainer.shap_values(X)
        if isinstance(shap_vals, list):
            shap_vals = shap_vals[1]
        elif isinstance(shap_vals, np.ndarray) and shap_vals.ndim == 3:
            shap_vals = shap_vals[:, :, 1]
        
        if isinstance(self.explainer.expected_value, (list, np.ndarray)) and len(np.atleast_1d(self.explainer.expected_value)) > 1:
            base_value = np.atleast_1d(self.explainer.expected_value)[1]
        else:
            base_value = self.explainer.expected_value
        
        return {
            'shap_values': shap_vals[idx],
            'feature_values': X[idx],
            'feature_names': self.feature_names,
            'base_value': base_value
        }
    
    def batch_predict(self, df_raw: pd.DataFrame) -> pd.DataFrame:
        """Batch prediction with full reporting."""
        predictions = self.predict(df_raw, return_proba=True)
        risk_tiers = self.predict_risk_tier(df_raw, probabilities=predictions)
        
        return pd.DataFrame({
            'ChurnProbability': predictions,
            'RiskTier': risk_tiers,
            'PredictedChurn': (predictions >= 0.5).astype(int)
        })
    
    def generate_report(self, df_raw: pd.DataFrame) -> Dict:
        """Generate comprehensive business report."""
        results = self.batch_predict(df_raw)
        
        report = {
            'summary': {
                'total_customers': len(df_raw),
                'predicted_churners': int(results['PredictedChurn'].sum()),
                'churn_rate': float(results['PredictedChurn'].mean()),
                'avg_churn_probability': float(results['ChurnProbability'].mean())
            },
            'risk_distribution': results['RiskTier'].value_counts().to_dict(),
            'high_risk_customers': df_raw[results['RiskTier'] == 'Critical Risk']['customerID'].tolist()[:10],
            'metadata': self.metadata
        }
        return report
    
    def save(self, filepath: str) -> None:
        """Serialize pipeline."""
        with open(filepath, 'wb') as f:
            serializer.dump(self, f)
        logger.info(f"Pipeline saved to {filepath}")
    
    @staticmethod
    def load(filepath: str) -> 'ChurnPredictionSystem':
        """Deserialize pipeline."""
        with open(filepath, 'rb') as f:
            pipeline = serializer.load(f)
        logger.info(f"Pipeline loaded from {filepath}")
        return pipeline

# Instantiate pipeline
pipeline = ChurnPredictionSystem(
    model=final_model,
    scaler=scaler,
    feature_engineer=fe,
    feature_names=feature_names,
    explainer=explainer if SHAP_AVAILABLE and 'explainer' in locals() else None,
    requires_scaling=best_model_name in scaled_model_names
)

print_success("Production pipeline created successfully!")
```
