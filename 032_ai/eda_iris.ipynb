## Introduction

The Iris dataset, introduced by British statistician Ronald Fisher in 1936, is the 'Hello World' of machine learning. It contains measurements of 150 iris flowers from three different species: *Iris setosa*, *Iris versicolor*, and *Iris virginica*.

**Why is this dataset perfect for learning EDA?**
- ✅ Small size (150 samples) — fast to process
- ✅ Clean data — minimal preprocessing needed
- ✅ Clear patterns — excellent for visualization
- ✅ Real-world relevance — classification problem
- ✅ Four numerical features + one categorical target

By the end of this notebook, you'll have a complete, professional EDA template that you can adapt for any classification dataset. Let's dive in! 🚀

## 6. Outlier Detection and Analysis

Outliers can significantly impact model performance. Let's identify and analyze them systematically.

````python
# Statistical outlier detection using IQR method
def detect_outliers_iqr(data, feature):
    """Detect outliers using Interquartile Range method"""
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    return outliers, lower_bound, upper_bound

print('🔍 OUTLIER DETECTION REPORT')
print('=' * 60)

outlier_summary = {}
for feature in numerical_features:
    outliers, lower, upper = detect_outliers_iqr(df, feature)
    outlier_summary[feature] = len(outliers)
    print(f"\n📌 {feature.replace('_', ' ').title()}:")
    print(f"   Valid range: [{lower:.2f}, {upper:.2f}]")
    print(f"   Outliers found: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
    if len(outliers) > 0:
        print(f"   Outlier values: {outliers[feature].values}")

print('\n' + '=' * 60)
print('📊 SUMMARY: Total outliers per feature')
for feature, count in outlier_summary.items():
    print(f"   {feature.replace('_', ' ').title()}: {count}")
````

````python
# Visualize outliers with enhanced box plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

for idx, feature in enumerate(numerical_features):
    # Box plot with swarm overlay
    sns.boxplot(data=df, x='species', y=feature, ax=axes[idx],
                palette=['#FF6B6B', '#4ECDC4', '#45B7D1'])
    sns.swarmplot(data=df, x='species', y=feature, ax=axes[idx],
                  color='black', alpha=0.5, size=4)
    axes[idx].set_title(f'Outliers in {feature.replace("_", " ").title()} by Species', 
                        fontsize=12, fontweight='bold')
    axes[idx].tick_params(axis='x', rotation=15)

plt.suptitle('🎯 Outlier Analysis by Species', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

### 💡 Key Insight: Outlier Assessment

**Outlier Findings:**
- Most outliers appear in **Sepal Width** (primarily in *Versicolor* and *Virginica*)
- Outliers represent natural variation, not data entry errors
- Given small dataset size (150 samples), removing outliers might lose valuable information
- **Recommendation**: Keep outliers but consider robust scaling for ML models

## 7. Feature Distributions and Skewness

Understanding skewness helps us determine if transformations are needed for machine learning algorithms.

````python
# Calculate skewness and kurtosis for each feature
from scipy.stats import skew, kurtosis, shapiro

print('📐 DISTRIBUTION ANALYSIS')
print('=' * 70)
print(f"{'Feature':<20} {'Skewness':<12} {'Kurtosis':<12} {'Normal?':<10}")
print('-' * 70)

distribution_stats = []
for feature in numerical_features:
    skewness = skew(df[feature])
    kurt = kurtosis(df[feature])
    _, p_value = shapiro(df[feature])
    is_normal = 'Yes' if p_value > 0.05 else 'No'
    
    distribution_stats.append({
        'feature': feature,
        'skewness': skewness,
        'kurtosis': kurt,
        'shapiro_p': p_value,
        'normal': is_normal
    })
    
    skew_interpretation = '""' if abs(skewness) < 0.5 else ('Moderate' if abs(skewness) < 1 else 'High')
    print(f"{feature.replace('_', ' ').title():<20} {skewness:<12.3f} {kurt:<12.3f} {is_normal:<10}")

print('-' * 70)
print('\n📖 Interpretation Guide:')
print('   Skewness: |value| < 0.5 (approx normal), 0.5-1 (moderate), >1 (high)')
print('   Kurtosis: >0 (heavy tails), <0 (light tails)')
print('   Shapiro-Wilk: p > 0.05 suggests normal distribution')
````

````python
# Q-Q plots to visualize normality
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

from scipy.stats import probplot

for idx, feature in enumerate(numerical_features):
    probplot(df[feature], dist="norm", plot=axes[idx])
    axes[idx].set_title(f'Q-Q Plot: {feature.replace("_", " ").title()}', 
                        fontsize=12, fontweight='bold')
    axes[idx].get_lines()[0].set_markerfacecolor(colors[idx])
    axes[idx].get_lines()[0].set_markersize(6)
    axes[idx].get_lines()[1].set_color('red')
    axes[idx].get_lines()[1].set_linewidth(2)

plt.suptitle('📈 Q-Q Plots: Normality Assessment', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

### 💡 Key Insight: Distribution Characteristics

**Skewness Analysis:**
- **Sepal Width**: Slightly left-skewed (some low outliers)
- **Petal features**: Bimodal distributions (not truly skewed, but multi-modal)
- **No features** follow perfect normal distribution (expected given 3 species mix)

**ML Implications:**
- Tree-based models: No transformation needed ✅
- Linear models/NNs: Consider standardization ✅

````python
# Feature importance visualization based on species separation
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Encode target variable
le = LabelEncoder()
y_encoded = le.fit_transform(df['species'])
X = df[numerical_features]

# Train a simple Random Forest to get feature importance
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y_encoded)

# Create feature importance plot
importance_df = pd.DataFrame({
    'Feature': [f.replace('_', ' ').title() for f in numerical_features],
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=True)

plt.figure(figsize=(10, 6))
bars = plt.barh(importance_df['Feature'], importance_df['Importance'], 
                color=['#96CEB4', '#45B7D1', '#4ECDC4', '#FF6B6B'])

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.005, bar.get_y() + bar.get_height()/2,
             f'{width:.3f}', ha='left', va='center', fontweight='bold')

plt.title('🎯 Feature Importance for Species Classification', 
          fontsize=16, fontweight='bold')
plt.xlabel('Importance Score', fontsize=12)
plt.xlim(0, max(importance_df['Importance']) * 1.2)
plt.tight_layout()
plt.show()

print('\n📊 Feature Importance Ranking:')
for idx, row in importance_df.sort_values('Importance', ascending=False).iterrows():
    print(f"   {row['Feature']}: {row['Importance']:.3f}")
````
