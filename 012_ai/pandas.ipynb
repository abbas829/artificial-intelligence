## 2. Creating Series

A **Series** is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, Python objects, etc.). Think of it as a single column of data with an index.

**Key characteristics:**
- One-dimensional
- Homogeneous data type (all elements are the same type)
- Has an index (labels for each element)
- Can be created from lists, dictionaries, numpy arrays, or scalars

````python
# 📌 Creating Series from different sources

# 1. From a Python list
fruits = pd.Series(['apple', 'banana', 'cherry', 'date'])
print("From list:")
print(fruits)
print()

# 2. From a dictionary (keys become index)
population = pd.Series({
    'New York': 8.4,
    'Los Angeles': 3.9,
    'Chicago': 2.7,
    'Houston': 2.3
})
print("From dictionary:")
print(population)
print()

# 3. From a NumPy array with custom index
temps = pd.Series(
    np.array([72, 75, 78, 71, 69]),
    index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    name='Temperature'
)
print("From NumPy array with custom index:")
print(temps)
print()

# 4. From a scalar value (broadcasted to all indices)
constant = pd.Series(5, index=['a', 'b', 'c', 'd'])
print("From scalar:")
print(constant)
````

## 3. Series Attributes and Methods

Series come with powerful built-in attributes and methods for inspection and analysis.

````python
# 📌 Exploring Series attributes and methods

# Create a sample Series
scores = pd.Series([85, 92, 78, 95, 88, 73, 90], name='Exam Scores')

print("📊 Series Attributes:")
print(f"Values: {scores.values}")
print(f"Index: {scores.index}")
print(f"Data type: {scores.dtype}")
print(f"Name: {scores.name}")
print(f"Size: {scores.size}")
print()

print("🔍 Series Methods:")
print("First 3 elements (head):")
print(scores.head(3))
print()

print("Last 2 elements (tail):")
print(scores.tail(2))
print()

print("Statistical summary (describe):")
print(scores.describe())
print()

print(f"Mean: {scores.mean():.2f}")
print(f"Median: {scores.median():.2f}")
print(f"Standard deviation: {scores.std():.2f}")
print(f"Min: {scores.min()}, Max: {scores.max()}")
````

## 4. Indexing and Selection in Series

Pandas provides multiple ways to access data in a Series:

- **Label-based indexing** (`.loc[]`): Uses the index labels
- **Position-based indexing** (`.iloc[]`): Uses integer positions (0, 1, 2...)
- **Boolean indexing**: Filter using conditions

````python
# 📌 Indexing and selecting data in a Series

# Create a Series with meaningful index
sales = pd.Series(
    [120, 150, 90, 200, 175, 130],
    index=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    name='Monthly Sales'
)

print("Original Series:")
print(sales)
print()

# .loc[] - Label-based indexing
print("Using .loc[] (label-based):")
print(f"sales.loc['Mar'] = {sales.loc['Mar']}")
print(f"sales.loc['Feb':'May'] = ")
print(sales.loc['Feb':'May'])  # Inclusive of both endpoints
print()

# .iloc[] - Position-based indexing
print("Using .iloc[] (position-based):")
print(f"sales.iloc[2] = {sales.iloc[2]}")
print(f"sales.iloc[1:4] = ")  # Exclusive of the last endpoint (like Python lists)
print(sales.iloc[1:4])
print()

# Boolean indexing (filtering)
print("Boolean indexing (sales > 140):")
high_sales = sales[sales > 140]
print(high_sales)
print()

# Complex condition
print("Complex condition (sales between 100 and 180):")
mid_sales = sales[(sales >= 100) & (sales <= 180)]
print(mid_sales)
````

## 5. Creating DataFrames

A **DataFrame** is a two-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dictionary of Series objects.

**Key characteristics:**
- Two-dimensional (rows and columns)
- Each column can be a different data type
- Has both row index and column index
- The most commonly used pandas object

````python
# 📌 Creating DataFrames from different sources

# 1. From a dictionary of lists (the most common approach)
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [50000, 60000, 75000, 65000, 70000]
}
df1 = pd.DataFrame(data_dict)
print("From dictionary of lists:")
print(df1)
print()

# 2. From a list of dictionaries
data_list = [
    {'Product': 'Laptop', 'Price': 1200, 'Stock': 50},
    {'Product': 'Mouse', 'Price': 25, 'Stock': 200},
    {'Product': 'Keyboard', 'Price': 100, 'Stock': 150},
    {'Product': 'Monitor', 'Price': 300, 'Stock': 80}
]
df2 = pd.DataFrame(data_list)
print("From a list of dictionaries:")
print(df2)
print()

# 3. From a NumPy array with custom index and columns
np.random.seed(42)
data_array = np.random.randn(5, 4)  # 5 rows, 4 columns of random numbers
df3 = pd.DataFrame(
    data_array,
    index=['A', 'B', 'C', 'D', 'E'],
    columns=['Feature1', 'Feature2', 'Feature3', 'Feature4']
)
print("From a NumPy array:")
print(df3)
print()

# 4. Create an empty DataFrame and then populate it
df4 = pd.DataFrame(columns=['Date', 'Temperature', 'Humidity'])
df4.loc[0] = ['2024-01-01', 72, 45]
df4.loc[1] = ['2024-01-02', 75, 50]
print("Populated empty DataFrame:")
print(df4)
````

## 7. Basic Selection: Columns and Rows

Selecting data is one of the most common operations in Pandas. You can select columns by name and rows by position or label.

````python
# 📌 Selecting columns and rows in DataFrames

# Create a sample DataFrame
students = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math': [85, 92, 78, 95, 88],
    'Science': [90, 85, 88, 92, 85],
    'English': [88, 90, 85, 89, 92]
})
print("Original DataFrame:")
print(students)
print()

# Selecting a single column (returns a Series)
print("Selecting single column 'Name':")
print(students['Name'])
print(f"Type: {type(students['Name'])}")
print()

# Selecting multiple columns (returns a DataFrame)
print("Selecting multiple columns ['Name', 'Math']:")
print(students[['Name', 'Math']])
print(f"Type: {type(students[['Name', 'Math']])}")
print()

# .loc[] - Label-based selection (row, column)
print("Using .loc[] - Label-based:")
print("students.loc[0, 'Name']:", students.loc[0, 'Name'])
print("students.loc[1:3, ['Name', 'Math']]:" )
print(students.loc[1:3, ['Name', 'Math']])  # Rows 1-3 inclusive
print()

# .iloc[] - Position-based selection
print("Using .iloc[] - Position-based:")
print("students.iloc[0, 1]:", students.iloc[0, 1])  # Row 0, Column 1
print("students.iloc[1:4, 0:3]:")  # Rows 1-3 (exclusive of 4), Columns 0-2
print(students.iloc[1:4, 0:3])
print()

# Selecting all rows for specific columns
print("All rows, specific columns with .loc[]:")
print(students.loc[:, ['Name', 'Math', 'Science']])
````

## 8. Adding, Dropping, and Renaming Columns

Data manipulation often requires modifying the structure of your DataFrame.

````python
# 📌 Modifying a DataFrame structure

# Start with a fresh DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})
print("Original DataFrame:")
print(df)
print()

# Add a new column
df['D'] = [1000, 2000, 3000, 4000, 5000]
print("After adding column 'D':")
print(df)
print()

# Add a calculated column
df['Sum'] = df['A'] + df['B'] + df['C'] + df['D']
print("After adding calculated column 'Sum':")
print(df)
print()

# Rename columns
df_renamed = df.rename(columns={'A': 'Column_A', 'B': 'Column_B', 'Sum': 'Total'})
print("After renaming columns:")
print(df_renamed)
print()

# Drop columns (axis=1 refers to columns)
df_dropped = df.drop(['C', 'D'], axis=1)
print("After dropping columns 'C' and 'D':")
print(df_dropped)
print()

# Drop rows (axis=0 refers to rows and is the default)
df_dropped_rows = df.drop([2, 4])  # Drop rows with index 2 and 4
print("After dropping rows 2 and 4:")
print(df_dropped_rows)
print()

# Note: drop() returns a new DataFrame unless you set inplace=True
print("Original df unchanged (drop creates a copy by default):")
print(df.columns.tolist())
````

## 10. Simple Operations and Calculated Columns

Pandas makes it easy to perform vectorized operations on entire columns — no loops needed!


````python
# 📌 Arithmetic operations and calculated columns

# Create a sample dataset
operations_df = pd.DataFrame({
    'Item': ['A', 'B', 'C', 'D', 'E'],
    'Price': [10.5, 25.0, 8.75, 45.0, 12.0],
    'Quantity': [100, 50, 200, 30, 150],
    'Discount_Rate': [0.1, 0.05, 0.15, 0.0, 0.2]
})

print("Original DataFrame:")
print(operations_df)
print()

# Calculate total value before discount
operations_df['Total_Value'] = operations_df['Price'] * operations_df['Quantity']
print("After adding 'Total_Value' (Price * Quantity):")
print(operations_df)
print()

# Calculate discount amount
operations_df['Discount_Amount'] = operations_df['Total_Value'] * operations_df['Discount_Rate']
print("After adding 'Discount_Amount':")
print(operations_df)
print()

# Calculate final price after discount
operations_df['Final_Value'] = operations_df['Total_Value'] - operations_df['Discount_Amount']
print("After adding 'Final_Value':")
print(operations_df)
print()

# Statistical operations across columns
print("📊 Summary statistics:")
print(f"Total revenue (before discount): ${operations_df['Total_Value'].sum():,.2f}")
print(f"Total revenue (after discount): ${operations_df['Final_Value'].sum():,.2f}")
print(f"Average discount rate: {operations_df['Discount_Rate'].mean():.1%}")
print(f"Highest value item: {operations_df.loc[operations_df['Final_Value'].idxmax(), 'Item']}")
````

## 11. Real AI/ML Context: Working with a Dataset

Now let's apply everything we've learned to a realistic ML scenario. We'll create a synthetic "Iris-like" dataset and prepare it for machine learning.

````python
# 📌 Creating a synthetic dataset for ML practice

np.random.seed(42)
n_samples = 150

# Create a synthetic Iris-like dataset
ml_df = pd.DataFrame({
    'sepal_length': np.concatenate([
        np.random.normal(5.0, 0.3, 50),  # setosa
        np.random.normal(6.0, 0.4, 50),  # versicolor
        np.random.normal(6.5, 0.5, 50)   # virginica
    ]).round(2),
    'sepal_width': np.concatenate([
        np.random.normal(3.4, 0.3, 50),
        np.random.normal(2.8, 0.3, 50),
        np.random.normal(3.0, 0.3, 50)
    ]).round(2),
    'petal_length': np.concatenate([
        np.random.normal(1.5, 0.2, 50),
        np.random.normal(4.5, 0.4, 50),
        np.random.normal(5.5, 0.5, 50)
    ]).round(2),
    'petal_width': np.concatenate([
        np.random.normal(0.2, 0.05, 50),
        np.random.normal(1.3, 0.2, 50),
        np.random.normal(2.0, 0.3, 50)
    ]).round(2),
    'species': ['setosa'] * 50 + ['versicolor'] * 50 + ['virginica'] * 50
})

print("🌸 Synthetic Iris dataset (first 10 rows):")
print(ml_df.head(10))
print()

print("📊 Dataset shape:", ml_df.shape)
print("📊 Species distribution:")
print(ml_df['species'].value_counts())
print()

print("📊 Statistical summary of features:")
print(ml_df.describe())
````

````python
# 📊 Visualization 1: Histogram of a numeric column

plt.figure(figsize=(10, 6))
plt.hist(ml_df['sepal_length'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Sepal Length', fontsize=14, fontweight='bold')
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("💡 This histogram shows the distribution of sepal lengths across all species.")
````

````python
# 📊 Visualization 2: Bar chart of a categorical column

species_counts = ml_df['species'].value_counts()

plt.figure(figsize=(8, 6))
bars = plt.bar(species_counts.index, species_counts.values, 
               color=['#FF6B6B', '#4ECDC4', '#45B7D1'], edgecolor='black')
plt.title('Distribution of Species', fontsize=14, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Count', fontsize=12)
# Add value labels to the bars
plt.grid(axis='y', alpha=0.3)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print("💡 This bar chart shows that the dataset is balanced, with 50 samples for each species.")
````

````python
# 📊 Visualization 3: Scatter plot of two features

plt.figure(figsize=(10, 8))

# Create a scatter plot with a different color for each species
colors = {'setosa': '#FF6B6B', 'versicolor': '#4ECDC4', 'virginica': '#45B7D1'}

for species in ml_df['species'].unique():
    subset = ml_df[ml_df['species'] == species]
    plt.scatter(subset['petal_length'], subset['petal_width'], 
                c=colors[species], label=species, s=100, alpha=0.7, edgecolors='black')

plt.title('Petal Length vs Petal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.legend(title='Species', fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("💡 This scatter plot shows that petal measurements separate the species very well.")
print("   Setosa is clearly distinct, while Versicolor and Virginica overlap slightly.")
````

````python
# 📊 Visualization 4: Box plot for outlier detection

plt.figure(figsize=(12, 6))

# Create boxplots for all numeric columns
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
data_to_plot = [ml_df[col] for col in numeric_cols]

box_plot = plt.boxplot(data_to_plot, labels=numeric_cols, patch_artist=True)

# Color the boxes
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

plt.title('Boxplots of All Features (Outlier Detection)', fontsize=14, fontweight='bold')
plt.xlabel('Features', fontsize=12)
plt.ylabel('Measurement (cm)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("💡 Box plots show the distribution and help identify outliers (points outside the whiskers).")
print("   Sepal width appears to have some outliers on both ends.")
````

````python
# 📌 Preparing data for machine learning

# Step 1: Separate features (X) and target (y)
# Features are all columns except 'species'
X = ml_df.drop('species', axis=1)  # Features
y = ml_df['species']                 # Target variable

print("✅ Features (X) - First 5 rows:")
print(X.head())
print(f"   Shape: {X.shape}")
print()

print("✅ Target (y) - First 5 values:")
print(y.head())
print(f"   Shape: {y.shape}")
print()

# Step 2: Check for missing values (critical for ML)
print("🔍 Missing values check:")
print(ml_df.isnull().sum())
print("✅ No missing values found!" if ml_df.isnull().sum().sum() == 0 else "⚠️ Missing values detected!")
print()

# Step 3: Basic statistics for features
print("📊 Feature statistics for ML:")
print(X.describe())
print()

# Step 4: Feature correlation (important for feature selection)
print("🔗 Feature correlation matrix:")
correlation_matrix = X.corr()
print(correlation_matrix.round(2))
````

````python
import pandas as pd

# Create the Series
temperatures = pd.Series(
    [68, 72, 75, 71, 69, 73, 70],
    index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
)

print("Temperatures Series:")
print(temperatures)
print(f"\nData type: {temperatures.dtype}")

# Expected output:
# Mon    68
# Tue    72
# Wed    75
# Thu    71
# Fri    69
# Sat    73
# Sun    70
# dtype: int64
````

### Exercise 2 Solution

````python
# Mean temperature
mean_temp = temperatures.mean()
print(f"Mean temperature: {mean_temp:.2f}")

# Maximum and which day
max_temp = temperatures.max()
max_day = temperatures.idxmax()
print(f"Maximum temperature: {max_temp} on {max_day}")

# Temperatures above 70
above_70 = temperatures[temperatures > 70]
print(f"\nTemperatures above 70:\n{above_70}")

# Expected output:
# Mean temperature: 71.14
# Maximum temperature: 75 on Wed
# Temperatures above 70:
# Tue    72
# Wed    75
# Thu    71
# Sat    73
# dtype: int64
````

### Exercise 3 Solution

````python
# Create DataFrame from dictionary
inventory = pd.DataFrame({
    'Product': ['Widget', 'Gadget', 'Thingama', 'Doohickey'],
    'Price': [25.50, 45.00, 12.99, 8.50],
    'Stock': [100, 75, 200, 150],
    'Category': ['A', 'B', 'A', 'C']
})

print("Inventory DataFrame:")
print(inventory)
print(f"\nShape: {inventory.shape}")

# Expected output:
#     Product  Price  Stock Category
# 0    Widget  25.50    100        A
# 1    Gadget  45.00     75        B
# 2  Thingama  12.99    200        A
# 3 Doohickey   8.50    150        C
# Shape: (4, 4)
````

### Exercise 4 Solution

````python
# Column names
print("Column names:")
print(inventory.columns.tolist())

# Data types
print("\nData types:")
print(inventory.dtypes)

# Info summary
print("\nInfo:")
inventory.info()

# Describe
print("\nStatistical summary:")
print(inventory.describe())
````

### Exercise 5 Solution

````python
# 1. Select Price and Stock for rows 1 and 2 using .loc
result1 = inventory.loc[1:2, ['Price', 'Stock']]
print("Using .loc[1:2, ['Price', 'Stock']]:")
print(result1)
print()

# 2. Select first 3 rows, first 2 columns using .iloc
result2 = inventory.iloc[0:3, 0:2]
print("Using .iloc[0:3, 0:2]:")
print(result2)
print()

# 3. Boolean indexing for Category 'A'
result3 = inventory[inventory['Category'] == 'A']
print("Rows where Category is 'A':")
print(result3)
````

### Exercise 6 Solution

````python
# 1. Add Total_Value column
inventory['Total_Value'] = inventory['Price'] * inventory['Stock']
print("After adding Total_Value:")
print(inventory)
print()

# 2. Rename product (update specific value)
inventory.loc[inventory['Product'] == 'Thingama', 'Product'] = 'Thingamajig'
print("After renaming product:")
print(inventory)
print()

# 3. Rename Category column to Dept
inventory = inventory.rename(columns={'Category': 'Dept'})
print("After renaming column:")
print(inventory)
````

### Exercise 7 Solution

````python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Create synthetic DataFrame
data = pd.DataFrame({
    'Age': np.random.randint(18, 65, 50),
    'Income': np.random.randint(30000, 100000, 50),
    'Score': np.random.randint(0, 100, 50)
})

print("First 10 rows:")
print(data.head(10))
print("\nLast 10 rows:")
print(data.tail(10))
print("\nRandom sample of 5 rows:")
print(data.sample(5))
print("\nMean and Std for each column:")
print(data.agg(['mean', 'std']))
````

### Exercise 8 Solution

````python
# Filter rows
filtered = data[(data['Age'] > 30) & (data['Income'] > 50000)]

# Calculate average Score for filtered group
avg_score = filtered['Score'].mean()

# Count rows
count = len(filtered)

print(f"Filtered data (Age > 30 AND Income > 50000):")
print(filtered)
print(f"\nAverage Score for this group: {avg_score:.2f}")
print(f"Number of rows meeting conditions: {count}")
````

### Exercise 9 Solution

````python
# Create synthetic dataset
np.random.seed(42)

ml_data = pd.DataFrame({
    'Feature1': np.random.normal(50, 10, 100),
    'Feature2': np.random.uniform(0, 100, 100),
    'Target': np.random.randint(0, 2, 100)
})

# 1. Separate X and y
X = ml_data[['Feature1', 'Feature2']]
y = ml_data['Target']

# 2. Check for missing values
print("Missing values:")
print(ml_data.isnull().sum())

# 3. Print shapes
print(f"\nX shape: {X.shape}")
print(f"y shape: {y.shape}")

# 4. Correlation
correlation = ml_data['Feature1'].corr(ml_data['Feature2'])
print(f"\nCorrelation between Feature1 and Feature2: {correlation:.4f}")
````

### Exercise 10 Solution

````python
import matplotlib.pyplot as plt

# 1. Histogram of Feature1
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.hist(ml_data['Feature1'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Feature1')
plt.xlabel('Feature1')
plt.ylabel('Frequency')

# 2. Bar chart of Target counts
plt.subplot(1, 3, 2)
target_counts = ml_data['Target'].value_counts()
plt.bar(target_counts.index.astype(str), target_counts.values, color=['coral', 'lightgreen'])
plt.title('Target Distribution')
plt.xlabel('Target')
plt.ylabel('Count')

# 3. Scatter plot colored by Target
plt.subplot(1, 3, 3)
for target in [0, 1]:
    subset = ml_data[ml_data['Target'] == target]
    plt.scatter(subset['Feature1'], subset['Feature2'], 
                label=f'Target={target}', alpha=0.6)
plt.title('Feature1 vs Feature2')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()

plt.tight_layout()
plt.show()
````

### Exercise 11 Solution

````python
# Create student DataFrame
students = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Math': [85, 92, 78, 95, 88, 73],
    'Science': [90, 85, 88, 92, 85, 78],
    'English': [88, 90, 85, 89, 92, 80]
})

# 1. Calculate average
students['Average'] = (students['Math'] + students['Science'] + students['English']) / 3

# 2. Filter students with average > 85
high_achievers = students[students['Average'] > 85]

# 3. Sort by Math score descending
sorted_students = high_achievers.sort_values('Math', ascending=False)

# 4. Display Name and Average
result = sorted_students[['Name', 'Average']]

print("Students with Average > 85, sorted by Math score:")
print(result)

# Expected output shows Bob, Diana, Eve with their averages
````
